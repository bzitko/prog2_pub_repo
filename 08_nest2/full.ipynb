## 5.3. Ugniježđene neindeksirane kolekcije

**Neindeksirane kolekcije** su one koje se ne mogu pristupiti prema rednom broju. Takve kolekcije su **skup** (`set`), **zamrznut skup** (`frozenset`) i **rječnik** (`dict`).

Kada ugnježđujemo neindeksirane kolekcije, možemo ih smještati:
- **Jedan u drugi** (rječnik u rječniku, skup kao vrijednost rječnika, itd.)
- **U indeksirane kolekcije** (skup ili rječnik kao element liste)
- **U mješovitim kombinacijama** s indeksiranim kolekcijama

Važno je biti svjestan **promjenjivosti** tih kolekcija:
- `dict` je **promjenjiv** - možemo dodavati, mijenjati i brisati parove ključeva i vrijednosti
- `set` je **promjenjiv** - možemo dodavati i brisati elemente
- `frozenset` je **nepromjenjiv** - ne možemo mijenjati nakon stvaranja

### 5.3.1. Ugniježđeni rječnici

**Ugniježđeni rječnici** omogućuju organiziranje podataka u više razina, gdje vrijednosti mogu biti drugi rječnici. To je korisno za strukturirane podatke kao što su zaposlenik s detaljima, proizvodi s karakteristikama, itd.

**Primjer jednostavnog ugniježđenog rječnika:**

```python
# Rječnik koji sadrži rječnike

    "ime": "Marko",
    "godine": 30,
    "adresa": {
        "ulica": "Teslina 5",

```

Za pristup vrijednostima u ugniježđenom rječniku koristimo **lanac ključeva**:

```python

print(osoba["adresa"]["ulica"])  # Teslina 5
```

**Više razina ugniježđivanja:**


### 5.3.2. Iteracija kroz ugniježđene rječnike

Za **iteraciju kroz ugniježđene rječnike**, možemo koristiti **ugnježđene petlje** ili **rekurziju** ovisno o strukturi podataka.
 
**Iteracija s `.items()`:**

```python
osoba = {
    "ime": "Marko",
    "adresa": {
        "grad": "Split",
        "ulica": "Teslina 5"
    }
}

for kljuc, vrijednost in osoba.items():
    if isinstance(vrijednost, dict):
        print(f"{kljuc}:")
        for unutarnji_kljuc, unutarnja_vrijednost in vrijednost.items():
            print(f"  {unutarnji_kljuc}: {unutarnja_vrijednost}")
    else:
        print(f"{kljuc}: {vrijednost}")
```

**Ispis:**

```
ime: Marko
adresa:
  grad: Split
  ulica: Teslina 5
```

**Pronalaženje vrijednosti u ugniježđenom rječniku:**

```python
fakultet = {
    "naziv": "Informatički",
    "lokacija": {
        "grad": "Split",
        "broj_studenata": 120
    }
}

# Pronalaženje vrijednosti na bilo kojoj razini
# pomoću rekurzije
def pronadi_vrijednost(rjecnik, trazena_vrijednost):
    for kljuc, vrijednost in rjecnik.items():
        if vrijednost == trazena_vrijednost:
            print(f"Pronađeno: {kljuc} = {vrijednost}")
            return True
        if isinstance(vrijednost, dict):
            if pronadi_vrijednost(vrijednost, trazena_vrijednost):
                return True
    return False

pronadi_vrijednost(fakultet, 120)  # Pronađeno: broj_studenata = 120
```

### 5.3.3. Skupovi (set) kao ugniježđene kolekcije

**Skupovi (`set`)** su neindeksirane, neuređene kolekcije jedinstvenih elemenata. 

Važno je znati da **skupove ne možemo koristiti direktno kao elemente drugih skupova** jer skupovi nisu nepromjenjivi. Međutim, možemo koristiti **`frozenset`** (zamrznut skup) koji je nepromjenjiv.

**Skupovi kao vrijednosti u rječniku:**

```python

    "učenik3": {"biologija", "engleski"}

print(predmeti["učenik1"])  # {'matematika', 'fizika', 'engleski'}
```

**Skupovi kao elementi liste:**

```python
# Lista skupova
grupe = [
    {"Ana", "Marko", "Ivan"},
    {"Petra", "Lana"},
    {"Filip", "Ema", "Marko"}
]

for i, grupa in enumerate(grupe):
    print(f"Grupa {i+1}: {grupa}")
```

### 5.3.4. Zamrzuti skupovi (frozenset) kao ugniježđene kolekcije

**Zamrzuti skup (`frozenset`)** je nepromjenjiva verzija skupa. To znači da ga **ne možemo mijenjati nakon što ga stvorimo**.

Glavna prednost `frozenset`-a je da **može biti element drugog skupa** ili **ključ u rječniku**, što obični `set` ne može biti jer je promjenjiv.

**Frozenset kao element seta:**

```python
# Obični skupovi ne mogu biti elementi drugog skupa
# skup_skupova = {{"A", "B"}, {"C", "D"}}  # Greška!

# Ali frozenset-ovi mogu:
skup_skupova = {frozenset({"A", "B"}), frozenset({"C", "D"})}
print(skup_skupova)  # {frozenset({'A', 'B'}), frozenset({'C', 'D'})}
```

**Frozenset kao ključ u rječniku:**

```python
# Obični skupovi ne mogu biti ključevi
# rjecnik = {{"a", "b"}: 1}  # Greška!

# Ali frozenset-ovi mogu:
grupe = {
    frozenset({"matematika", "fizika"}): "STEM grupa",
    frozenset({"povijest", "geografija"}): "Humanistička grupa"
}

print(grupe[frozenset({"matematika", "fizika"})])  # STEM grupa
```

✏️ **Zadatak**

1. Napravi skup od tri frozenset-a koji predstavljaju grupe interesetanja (barem 3 elementa po grupi).

3. Pronađi sve grupe koje sadrže "sport".

### 5.4.1. Lista od rječnika

**Lista od rječnika** je često korištena struktura za pohranu kolekcije objekata, slično kao baza podataka gdje svaki redak predstavlja zapis.

**Primjer:**

```python
# Baza učenika
ucenici = [
    {"id": 1, "ime": "Ana", "klasa": "1a", "ocjena": 4},
    {"id": 2, "ime": "Marko", "klasa": "1a", "ocjena": 3},
    {"id": 3, "ime": "Ivan", "klasa": "1b", "ocjena": 5}
]

# Pristup elementima
print(ucenici[0]["ime"])  # Ana
print(ucenici[1]["ocjena"])  # 3

# Iteracija
for ucenik in ucenici:
    print(f"{ucenik['ime']} ({ucenik['klasa']}): {ucenik['ocjena']}")

# Pronalaženje po kriteriju
najbolji = [u for u in ucenici if u["ocjena"] >= 4]
print(najbolji)
```

**Dodavanje i brisanje:**

```python
# Dodavanje novog učenika
ucenici.append({"id": 4, "ime": "Petra", "klasa": "1b", "ocjena": 4})

# Brisanje učenika
ucenici.pop(0)  # Briše prvog učenika

# Promjena podataka
ucenici[0]["ocjena"] = 5
```

### 5.4.3. Rječnik sa skupovima kao vrijednostima

**Rječnik s skupovima kao vrijednostima** je koristan kad trebamo jedinstvene podatke grupirane po kategorijama.
 
**Primjer:**

```python

}

if "Hrvatski" in jezici_po_zemlji["Hrvatska"]:
    print("Hrvatski se govori u Hrvatskoj")

print(svi_jezici)  # {'hrvatski', 'engleski', 'francuski', 'njemački'}

    print(f"{zemlja}: {jezici}")
```

### 5.4.5. Operacije na mješovitim ugniježđenim kolekcijama

Često trebamo izvršiti različite operacije na mješovitim kolekcijama: pronalaženje, filtriranje, transformaciju, agregaciju.

**Pronalaženje elemenata:**

```python
# Pronalaženje prvog elementa koji zadovoljava uvjet
podaci = [
    {"ime": "Ana", "ocjena": 4},
    {"ime": "Marko", "ocjena": 3},
    {"ime": "Ivan", "ocjena": 5}
]

najbolji = [u for u in podaci if u["ocjena"] >= 5]
prvi_najbolji = najbolji[0] if najbolji else None
print(prvi_najbolji)  # {'ime': 'Ivan', 'ocjena': 5}
```

**Mapiranje (transformacija):**

```python
# Ekstrakcija samo imena
imena = [u["ime"] for u in podaci]
print(imena)  # ['Ana', 'Marko', 'Ivan']

# Dvostruke ocjene
dvostruke_ocjene = {u["ime"]: u["ocjena"] * 2 for u in podaci}
print(dvostruke_ocjene)
```

**Filtriranje:**

```python
# Učenici s ocjenom >= 4
dobri = [u for u in podaci if u["ocjena"] >= 4]
print(dobri)

# Učenici s imenima koja počinju s 'A'
sa_A = [u for u in podaci if u["ime"].startswith("A")]
print(sa_A)
```

**Agregacija:**

```python
# Zbrajanje ocjena
suma_ocjena = sum(u["ocjena"] for u in podaci)
print(suma_ocjena)  # 12

# Prosječna ocjena
prosijek = suma_ocjena / len(podaci)
print(prosijek)  # 4.0
```

✏️ **Zadatak**

1. Kreiraj listu od 5 rječnika koji predstavljaju proizvode s poljem: naziv, cijena, kategorija.
2. Pronađi sve proizvode kategorije "hrana".
3. Sortiraj proizvode po cijeni (od najjeftinijeg do najskupnijeg).
4. Izračunaj prosječnu cijenu svih proizvoda.
5. Pronađi proizvod s najkraćim nazivom.

````python
print("Zadatak - Operacije na mješovitim kolekcijama:\n")

proizvodi = [
    {"naziv": "Jabuka", "cijena": 2.50, "kategorija": "hrana"},
    {"naziv": "Laptop", "cijena": 899.99, "kategorija": "elektronika"},
    {"naziv": "Kruh", "cijena": 1.99, "kategorija": "hrana"},
    {"naziv": "Kamera", "cijena": 599.99, "kategorija": "elektronika"},
    {"naziv": "Mljeko", "cijena": 1.50, "kategorija": "hrana"}
]

print("Zadatak 2 - Proizvodi kategorije 'hrana':")
hrana = [p for p in proizvodi if p["kategorija"] == "hrana"]
for p in hrana:
    print(f"  {p['naziv']}: {p['cijena']}€")

print("\nZadatak 3 - Sortirani po cijeni:")
sortirani = sorted(proizvodi, key=lambda p: p["cijena"])
for p in sortirani:
    print(f"  {p['naziv']}: {p['cijena']}€")

print("\nZadatak 4 - Prosječna cijena:")
prosijek = sum(p["cijena"] for p in proizvodi) / len(proizvodi)
print(f"  {prosijek:.2f}€")

print("\nZadatak 5 - Proizvod s najkraćim nazivom:")
najkraci = min(proizvodi, key=lambda p: len(p["naziv"]))
print(f"  {najkraci['naziv']} ({len(najkraci['naziv'])} znakova)")
````
