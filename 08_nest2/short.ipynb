## 5.3. Ugniježđene neindeksirane kolekcije

Elementima neindeksiranih kolekcija (`set`, `frozenset`, `dict`) se ne može pristupiti prema indeksu.

Kada ih ugnježđujemo, možemo ih smještati jedan u drugi ili u indeksirane kolekcije (liste, n-torke).

Važno je biti svjestan **promjenjivosti**:
- `dict` i `set` su **promjenjivi**
- `frozenset` je **nepromjenjiv**

### 5.3.1. Ugniježđeni rječnici

Ugniježđeni rječnici omogućuju organiziranje podataka u više razina gdje vrijednosti mogu biti drugi rječnici.

```python
osoba = {
    "ime": "Marko",
    "adresa": {
        "grad": "Split",
        "ulica": "Teslina 5"
    }
}
```

Pristup elementima: `osoba["adresa"]["grad"]` → `Split`

✏️ **Zadatak**

1. Napravi ugniježđeni rječnik učenika s imenima, razredom i ocjenama (barem 3 predmeta).
2. Pristupi i ispiši ocjenu iz matematike.
3. Promijeni ocjenu iz matematike na 5.
4. Dodaj novi predmet "Likovni" s ocjenom 4.

````python
print("Zadatak 1 - Ugniježđeni rječnik učenika:")
ucenik = {
    "ime": "Ana",
    "razred": "1a",
    "ocjene": {
        "matematika": 4,
        "engleski": 5,
        "hrvatski": 4,
        "fizika": 3
    }
}
print(ucenik)

print("\nZadatak 2 - Pristup ocjeni iz matematike:")
print(f"Ocjena iz matematike: {ucenik['ocjene']['matematika']}")

print("\nZadatak 3 - Promjena ocjene iz matematike:")
ucenik["ocjene"]["matematika"] = 5
print(f"Nova ocjena iz matematike: {ucenik['ocjene']['matematika']}")

print("\nZadatak 4 - Dodavanje novog predmeta:")
ucenik["ocjene"]["likovni"] = 4
print(ucenik)
````

### 5.3.2. Iteracija kroz ugniježđene rječnike

Za iteraciju kroz ugniježđene rječnike koriste se **ugnježđene petlje** ili **rekurzija** ovisno o strukturi.

```python
osoba = {
    "ime": "Marko",
    "adresa": {
        "grad": "Split",
        "ulica": "Teslina 5"
    }
}

for k, v in osoba.items():
    if isinstance(v, dict):
        for uk, uv in v.items():
            print(f"  {uk}: {uv}")
    else:
        print(f"{k}: {v}")
```

**Ispis:**

```
ime: Marko
adresa:
  grad: Split
  ulica: Teslina 5
```

✏️ **Zadatak**

1. Kreiraj ugniježđeni rječnik škole s odjelima i brojem učenika.
2. Iteriraj kroz sve odijele i ispiši njihove nazive i broj učenika.
3. Napravi funkciju koja broji ukupan broj učenika u svim odjelima.

````python
print("Zadatak 1 - Ugniježđeni rječnik škole:")
skola = {
    "naziv": "Gimnazija Novi Grad",
    "odijeli": {
        "matematika": 45,
        "informatika": 32,
        "fizika": 28,
        "kemija": 35
    }
}
print(skola)

print("\nZadatak 2 - Iteracija kroz odijele:")
for odjel, broj_ucenika in skola["odijeli"].items():
    print(f"{odjel}: {broj_ucenika} učenika")

print("\nZadatak 3 - Ukupan broj učenika:")
def broji_ucenike(skola):
    return sum(skola["odijeli"].values())

ukupno = broji_ucenike(skola)
print(f"Ukupno učenika: {ukupno}")
````

### 5.3.3. Skupovi (set) kao ugniježđene kolekcije

Skupovi su neindeksirane kolekcije jedinstvenih elemenata.

**Skupovi kao vrijednosti u rječniku:**
 
```python
predmeti = {
    "učenik1": {"matematika", "fizika"},
    "učenik2": {"matematika", "kemija"}
}
```

✏️ **Zadatak**

1. Kreiraj rječnik gdje su ključevi imena učenika, a vrijednosti skupovi ocjena.
2. Pronađi učenika s najboljim ocjenama (sve ≥ 4).
3. Pronađi zajedničke ocjene između dva učenika koristeći `&` operator.

````python
print("Zadatak 1 - Rječnik s ocjenama kao skupovima:")
ocjene = {
    "Ana": {5, 4, 5, 4, 5},
    "Marko": {3, 4, 3, 4, 3},
    "Ivan": {4, 4, 5, 4, 4}
}
print(ocjene)

print("\nZadatak 2 - Učenik s najboljim ocjenama:")
for ucenik, his_ocjene in ocjene.items():
    if all(ocjena >= 4 for ocjena in his_ocjene):
        print(f"{ucenik}: {sorted(his_ocjene)}")

print("\nZadatak 3 - Zajedničke ocjene između Ana i Marko:")
zajednicke = ocjene["Ana"] & ocjene["Marko"]
print(f"Zajedničke ocjene: {zajednicke}")
````

### 5.3.4. Zamrzuti skupovi (frozenset) kao ugniježđene kolekcije

**Frozenset** je nepromjenjiva verzija skupa. Može biti element drugog skupa ili ključ u rječniku.

```python
skup_skupova = {frozenset({"A", "B"}), frozenset({"C", "D"})}
```

✏️ **Zadatak**

1. Napravi skup od tri frozenset-a koji predstavljaju grupe interesiranja.
2. Kreiraj rječnik gdje su ključevi frozenset-i a vrijednosti imena grupa.
3. Pronađi sve grupe koje sadrže "sport".

````python
print("Zadatak 1 - Skup frozenset-a s grupama:")
grupe = {
    frozenset({"sport", "fudbal", "tenis"}),
    frozenset({"glazba", "gitara", "klavir"}),
    frozenset({"je", "programiranje", "python"})
}
print(grupe)

print("\nZadatak 2 - Rječnik s frozenset ključevima:")
grupa_imena = {
    frozenset({"sport", "fudbal", "tenis"}): "Sportska grupa",
    frozenset({"glazba", "gitara", "klavir"}): "Glazbena grupa",
    frozenset({"je", "programiranje", "python"}): "IT grupa"
}

for grupa, naziv in grupa_imena.items():
    print(f"{naziv}: {grupa}")

print("\nZadatak 3 - Grupe koje sadrže 'sport':")
for grupa, naziv in grupa_imena.items():
    if "sport" in grupa:
        print(f"{naziv}: {grupa}")
````

## 5.4. Ugniježđene kolekcije različitih tipova

### 5.4.1. Lista od rječnika

Lista od rječnika je često korištena struktura za pohranu kolekcije objekata, slično kao baza podataka.
 
```python
ucenici = [
    {"id": 1, "ime": "Ana", "razred": "1a"},
    {"id": 2, "ime": "Marko", "razred": "1a"}
]
```

✏️ **Zadatak**

1. Kreiraj listu od najmanje 4 rječnika koje predstavljaju knjige s: naslov, autor, godina, broj stranica.
2. Ispiši sve knjige starije od 2000.
3. Pronađi najkraću i najdulju knjigu.
4. Izračunaj prosječan broj stranica po knjizi.

````python
print("Zadatak 1 - Lista od rječnika (knjige):")
knjige = [
    {"naslov": "Python 101", "autor": "John Doe", "godina": 2015, "stranice": 350},
    {"naslov": "Clean Code", "autor": "Robert Martin", "godina": 2008, "stranice": 464},
    {"naslov": "Design Patterns", "autor": "Erich Gamma", "godina": 1994, "stranice": 395},
    {"naslov": "The Pragmatic Programmer", "autor": "Hunt & Thomas", "godina": 1999, "stranice": 321}
]

for knjiga in knjige:
    print(f"{knjiga['naslov']} - {knjiga['autor']} ({knjiga['godina']})")

print("\nZadatak 2 - Knjige starije od 2000:")
stare = [k for k in knjige if k["godina"] < 2000]
for knjiga in stare:
    print(f"{knjiga['naslov']} ({knjiga['godina']})")

print("\nZadatak 3 - Najkraća i najdulja knjiga:")
najkraca = min(knjige, key=lambda k: k["stranice"])
najdulja = max(knjige, key=lambda k: k["stranice"])
print(f"Najkraća: {najkraca['naslov']} ({najkraca['stranice']} stranica)")
print(f"Najdulja: {najdulja['naslov']} ({najdulja['stranice']} stranica)")

print("\nZadatak 4 - Prosječan broj stranica:")
prosijek = sum(k["stranice"] for k in knjige) / len(knjige)
print(f"Prosječno: {prosijek:.1f} stranica")
````

✏️ **Zadatak**

1. Kreiraj rječnik gdje su ključevi imena učenika, a vrijednosti skupovi njihovih hobija.
2. Pronađi hobije koje dijele svi učenici.
3. Pronađi sve hobije koje ima barem jedan učenik.

````python
print("Zadatak 1 - Hobiji učenika:")
hobiji = {
    "Ana": {"čitanje", "sport", "likovni"},
    "Marko": {"glazba", "sport", "igranje"},
    "Ivan": {"čitanje", "glazba", "programiranje"},
    "Petra": {"sport", "čitanje", "likovni"}
}

for ucenik, hobi_set in hobiji.items():
    print(f"{ucenik}: {hobi_set}")

print("\nZadatak 2 - Zajednički hobiji svih učenika:")
zajednicki = set.intersection(*hobiji.values())
print(f"Zajednički: {zajednicki if zajednicki else 'Nema zajedničkih'}")

print("\nZadatak 3 - Svi hobiji:")
svi_hobiji = set()
for hobi_set in hobiji.values():
    svi_hobiji = svi_hobiji | hobi_set
print(f"Svi hobiji: {sorted(svi_hobiji)}")
````

### 5.4.5. Operacije na mješovitim ugniježđenim kolekcijama

Često trebamo izvršiti operacije: pronalaženje, filtriranje, transformaciju, agregaciju na mješovitim kolekcijama.

- **Pronalaženje:** filtriranje sa skraćenim zapisom
- **Mapiranje:** ekstrakcija ili transformacija podataka
- **Agregacija:** zbrajanje, brojanje, pronalaženje min/max

**Pronalaženje elemenata:**

```python
# Pronalaženje prvog elementa koji zadovoljava uvjet
podaci = [
    {"ime": "Ana", "ocjena": 4},
    {"ime": "Marko", "ocjena": 3},
    {"ime": "Ivan", "ocjena": 5}
]

najbolji = [u for u in podaci if u["ocjena"] >= 5]
prvi_najbolji = najbolji[0] if najbolji else None
print(prvi_najbolji)  # {'ime': 'Ivan', 'ocjena': 5}
```

**Mapiranje (transformacija):**

```python
# Ekstrakcija samo imena
imena = [u["ime"] for u in podaci]
print(imena)  # ['Ana', 'Marko', 'Ivan']

# Dvostruke ocjene
dvostruke_ocjene = {u["ime"]: u["ocjena"] * 2 for u in podaci}
print(dvostruke_ocjene)
```

**Filtriranje:**

```python
# Učenici s ocjenom >= 4
dobri = [u for u in podaci if u["ocjena"] >= 4]
print(dobri)

# Učenici s imenima koja počinju s 'A'
sa_A = [u for u in podaci if u["ime"].startswith("A")]
print(sa_A)
```

**Agregacija:**

```python
# Zbrajanje ocjena
suma_ocjena = sum(u["ocjena"] for u in podaci)
print(suma_ocjena)  # 12

# Prosječna ocjena
prosijek = suma_ocjena / len(podaci)
print(prosijek)  # 4.0
```

✏️ **Zadatak**

1. Kreiraj listu od 5 rječnika koji predstavljaju proizvode s: naziv, cijena, kategorija.
2. Pronađi sve proizvode kategorije "hrana".
3. Sortiraj proizvode po cijeni (od najjeftinijeg do najskupnijeg).
4. Izračunaj prosječnu cijenu svih proizvoda.
5. Pronađi proizvod s najkraćim nazivom.

````python
print("Zadatak - Operacije na mješovitim kolekcijama:\n")

proizvodi = [
    {"naziv": "Jabuka", "cijena": 2.50, "kategorija": "hrana"},
    {"naziv": "Laptop", "cijena": 899.99, "kategorija": "elektronika"},
    {"naziv": "Kruh", "cijena": 1.99, "kategorija": "hrana"},
    {"naziv": "Kamera", "cijena": 599.99, "kategorija": "elektronika"},
    {"naziv": "Mljeko", "cijena": 1.50, "kategorija": "hrana"}
]

print("Zadatak 2 - Proizvodi kategorije 'hrana':")
hrana = [p for p in proizvodi if p["kategorija"] == "hrana"]
for p in hrana:
    print(f"  {p['naziv']}: {p['cijena']}€")

print("\nZadatak 3 - Sortirani po cijeni:")
sortirani = sorted(proizvodi, key=lambda p: p["cijena"])
for p in sortirani:
    print(f"  {p['naziv']}: {p['cijena']}€")

print("\nZadatak 4 - Prosječna cijena:")
prosijek = sum(p["cijena"] for p in proizvodi) / len(proizvodi)
print(f"  {prosijek:.2f}€")

print("\nZadatak 5 - Proizvod s najkraćim nazivom:")
najkraci = min(proizvodi, key=lambda p: len(p["naziv"]))
print(f"  {najkraci['naziv']} ({len(najkraci['naziv'])} znakova)")
````
