## 4.6 Ugniježđeni pozivi funkcija

Ugniježđeni pozivi funkcija se javljaju kada **jedna funkcija poziva drugu funkciju**, koja pak može pozvati dodatne funkcije. 

Primjerice, ako imamo tri funkcije `f1()`, `f2()` i `f3()` gdje se:
* unutar funkcije `f1()` poziva funkcija `f2()`, 
* unutar funkcije `f2()` poziva funkcija `f3()`, 
* funkcija `f3()` ne radi ništa i 

u glavnom programu se poziva funkcija `f1()`.

```Python 
def f3():
    pass    # f3() me radi ništa

def f2():
    f3()    # unutar f2() se poziva f3()


def f1():
    f2()    # unutar f1() se poziva f2()

f1()        # u glavnom programu se poziva f1()
```


Slijed izvršavanja:
1. u glavnom programu se poziva `f1()` i počinje izvršavanje od `f1()`.
2. funkcija `f1()` se izvršava i pozove funkciju `f2()`, **pauzira** se izvršavanje `f1()` i počinje izvršavanje `f2()` 
3. funkcija `f2()` se izvršava i pozove funkciju `f3()`, **pauzira** se izvršavanje `f2()` i počinje izvršavajne `f3()`
4. funkcija `f3()` se izvšava i vrati vrijednost u `f2()`, te se izvršavanje od `f2()` **nastavlja**
5. funkcija `f2()` se nastavi izvšavati i vrati vrijednost u `f1()`, te se izvršavanje od `f1()` **nastavlja**
6. funkcija `f1()` se nastavi izvršavati i vrati vrijednost u glavni program

Ovaj proces je moguć zahvaljujući **stogu poziva** (call stack) — posebnoj memorijskoj strukturi koja čuva redoslijed poziva. 
* **Početkom** izvršavanja funkcije `f()` ona se stavlja na stog poziva.
* **Vraćanjem** vrijednosti funkcije `f()` ona se skida sa stoga.


1. Poziva se `f1()` (`f1()` se stavlja na stog poziva)
2. `f1()` se izvršava i poziva `f2()` (`f2()` se stavlja na stog poziva)
3. `f2()` se izvršava i poziva `f3()` (`f3()` se stavlja na stog poziva)
4. `f3()` se izvršava i vraća rezultat (`f3()` se skida sa stoga poziva)
5. `f2()` se nastavlja izvršavati i vraća rezultat (`f2()` se skida sa stoga poziva)
6. `f1()` se nastavlja izvršava i vraća rezultat (`f1()` se skida sa stoga poziva)


```Python 
                       | Stog       | 
 → f1()                | f1         |
     └→ f2()           | f1, f2     |
          └→ f3()      | f1, f2, f3 |
          ←─           | f1, f2     |
       ←─              | f1         |
  ←─                   |            |
```
✏️ **Zadatak**

1. Definiraj funkcije `f3()` koja:
  - ispisuje poruku `"Ulazim u f3()"`
  - ispisuje poruku `"Izlazim iz f3()"`
2. Definiraj funkciju `f2()` koja:
  - ispisuje poruku `"Ulazim u f2()"`
  - poziva funkciju `f3()`
  - ispisuje poruku `"Izlazim iz f2()"`
3. Definiraj funkciju `f1()` koja:
  - ispisuje poruku `"Ulazim u f1()"`
  - poziva funkciju `f2()`
  - ispisuje poruku `"Izlazim iz f1()"`
4. U glavnom programu pozovi funkciju `f1()`

````python
def f3():
    print("     └→ Ulazim u f3()")
    print("     ←─ Izlazim iz f3()")
    
def f2():
    print("   └→ Ulazim u f2()")
    f3()
    print("   ←─ Izlazim iz f2()")

def f1():
    print(" └→ Ulazim u f1()")
    f2()
    print(" ←─ Izlazim iz f1()")

rez = f1()
````

### 4.6.1 Ugniježđeni pozivi funkcija i povratne vrijednosti

Funkcija koja je pozvala drugu funkciju može iskoristiti povratne vrijednosti funkcije koje je pozvala. 

Primjerice, ako imamo funkciju `f1()` i `f2()` gdje se:
* funkcija `f2()` pridruži varijabli vrijednost `2` i vrati vrijednost te varijable, 
* funkcija `f1()` pridruži varijabli vraćenu vrijednost poziva funkcije `f2()`,
* funkcija `f1()` poveća tu varijablu za `1` te vrati vrijednost te varijable i

u glavnom programu se pridruži varijabli vrijednost vraćene vrijednosti funkcije `f1()`, te ispisuje vrijednost te varijable.

```Python
def f2():
    var2 = 2
    return var2

def f1():
    var2 = f2()
    var1 = var2 + 1
    return var1

var1 = f1()
print(var1)
```

```Python
                       | Stog       | 
 → f1()                | f1         |
     └→ f2()           | f1, f2     |
     ←─ return 2       | f1         |
  ←─ return 3          |            |
```

Slijed izvršavanja:
1. u glavnom programu se prilikom pridruživanja vrijednosti varijable `var1` pozove funkcija `f1()` i počinje izvršavanje od `f1()`.
2. funkcija `f1()` se izvršava, prilikom pridruživanja vrijednosti varijable `var2` pozove funkciju `f2()`, **pauzira** se izvršavanje `f1()` i počinje izvršavanje `f2()` 
3. funkcija `f2()` se izvršava, postavi vrijednost `2` u `var2`, vrati `var2`  te se izvršavanje od `f1()` **nastavlja**
5. funkcija `f1()` se nastavi izvršavati i pridruži vraćenu vrijednost od `f2()` u varijablu `var2`
6. funkcija `f1()` se nastavi izvršavati i postavi vrijednost od `var1` na `var2 + 1`
7. funkcija `f1()` vrati `var1`
8. glavni program se nastavi izvršavati i pridruž vraćenu vrijednost od `f1()` u varijablu `var1`
9. glavni program ispisuje vrijednost varijable `var1`.

> 💡 **Napomena**
> 
> Možemo napraviti verzije funkcija `f1()` i `f2()` tako da ne koriste varijable, odnosno:
>
> ```Python
> def f2():
>     return 2
>
> def f1():
>    return f2() + 1
> 
> print(f1())
> ```

✏️ **Zadatak**

1. Definiraj funkcije `f3()` koja:
  - ispisuje poruku `"Ulazim u f3()"`
  - u varijablu `v` postavlja broj `100`
  - ispisuje poruku `f"Izlazim iz f1() i vraćam: {v}"`
  - vraća `v`
2. Definiraj funkciju `f2()` koja:
  - ispisuje poruku `"Ulazim u f2()"`
  - u varijablu `v` postavlja vrijednost dobivenu pozivom funkcije `f3()` povečanu za `10`
  - ispisuje poruku `f"Izlazim iz f2() i vraćam: {v}"`
3. Definiraj funkciju `f1()` koja:
  - ispisuje poruku `"Ulazim u f1()"`
  - u varijablu `v` postavlja vrijednost dobivenu pozivom funkcije `f2()` povečanu za `1`
  - ispisuje poruku `f"Izlazim iz f1() i vraćam: {v}"`
4. U glavnom programu pozovi funkciju `f1()`

````python
def f3():
    print("     └→ Ulazim u f3()")
    v = 100
    print(f"     ←─ Izlazim iz f3() i vraćam: {v}")
    return v

def f2():
    print("   └→ Ulazim u f2()")
    v = f3() + 10
    print(f"   ←─ Izlazim iz f2() i vraćam: {v}")
    return v

def f1():
    print(" └→ Ulazim u f1()")
    v = f2() + 1
    print(f" ←─ Izlazim iz f1() i vraćam: {v}")
    return v


v = f1()
print(f"Rezultat: {v}")
````

### 4.6.2 Ugniježđeni pozivi funkcija s parametrima

U prethodnim lekcijama funkcije su se pozivale bez prosljeđivanja vrijednosti.
Ugniježđene funkcije mogu međusobno razmjenjivati podatke **putem parametara** (argumenata). 

Ugniježđeni pozivi funkcija s parametrima javljaju se kada:

* jedna funkcija poziva drugu funkciju i
* **prosljeđuje joj argumente (vrijednosti)**,
* a pozvana funkcija može dalje proslijediti te vrijednosti drugim funkcijama.

**Primjer:**

Imamo tri funkcije `f1(x)`, `f2(y)` i `f3(z)` gdje se:

* `f1(x)` poziva s nekim argumentom i prosljeđuje ga funkciji `f2(y)`
* `f2(y)` prosljeđuje vrijednost funkciji `f3(z)`
* `f3(z)` koristi tu vrijednost i ispisuje je

```python
def f3(z):
    print(f"f3 prima: {z}")

def f2(y):
    f3(y)

def f1(x):
    f2(x)

f1(5)
```

**Slijed izvršavanja**

1. u glavnom programu se poziva `f1(5)` i počinje izvršavanje funkcije `f1`
2. vrijednost `5` se pridružuje parametru `x`
3. funkcija `f1` poziva `f2(x)` i prosljeđuje vrijednost `5`
4. izvršavanje `f1` se **pauzira**, počinje izvršavanje `f2`
5. vrijednost `5` se pridružuje parametru `y`
6. funkcija `f2` poziva `f3(y)` i prosljeđuje vrijednost `5`
7. izvršavanje `f2` se **pauzira**, počinje izvršavanje `f3`
8. vrijednost `5` se pridružuje parametru `z`
9. funkcija `f3` ispisuje vrijednost i završava
10. izvršavanje `f2` se nastavlja i završava
11. izvršavanje `f1` se nastavlja i završava


**Stog poziva i prijenos vrijednosti**

```python
                       | Stog                      |
 → f1(5)               | f1(x=5)                   |
     └→ f2(x)          | f1(x=5), f2(y=5)          |
          └→ f3(y)     | f1(x=5), f2(y=5), f3(z=5) |
          ←─ return    | f1(x=5), f2(y=5)          |
       ←─ return       | f1(x=5)                   |
  ←─ return            |                           |
```

* svaki poziv funkcije ima **svoje lokalne parametre**
* vrijednosti se **kopiraju (prosljeđuju)** u parametre funkcije
* promjena parametra unutar funkcije ne mijenja izvornu varijablu (za osnovne tipove)

> 💡 **Napomena**
>
> Ako je parametar promjenjivi tip podataka (`list`, `set`, `dict`), promjena unutar funkcije mijenja izvornu varijablu.
>
> Ako se želi sačuvati izvorna varijabla, potrebno je stvoriti njenu kopiju i proslijediti kopiju kao parametar funkcije.
>


**Primjer s obradom vrijednosti**

```python
def f3(z):
    return z * 2

def f2(y):
    return f3(y) + 3

def f1(x):
    return f2(x) + 1

print(f1(5))
```

**Slijed vrijednosti**

1. `f1(5)` → `x = 5`
2. `f2(5)` → `y = 5`
3. `f3(5)` → vraća `10`
4. `f2` vraća `10 + 3 = 13`
5. `f1` vraća `13 + 1 = 14`
6. ispisuje se `14`

> 💡 **Napomena**
>
> Prosljeđivanje se može i direktno ugnijezditi:
>
> ```python
> def f3(z):
>     return z * 2
> 
> def f2(y):
>     return y + 3
> 
> def f1(x):
>     return x + 1
> 
> print(f1(f2(f3(5))))
> ```
>
> ali takav zapis je teže pratiti i koristi se rjeđe.

✏️ **Zadatak**

1. Definiraj funkciju `f3(z)` koja:

   * ispisuje `f"Ulazim u f3({z})"`
   * vraća vrijednost `z * 2`
   * ispisuje `f"Izlazim iz f3({z}) i vraćam: {z * 2}"`

2. Definiraj funkciju `f2(y)` koja:

   * ispisuje `f"Ulazim u f2({y})"`
   * u varijablu `v` postavlja vrijednost dobivenu pozivom `f3(y)` povećanu za `5`
   * ispisuje `f"Izlazim iz f2({y}) i vraćam: {v}"`
   * vraća `v`

3. Definiraj funkciju `f1(x)` koja:

   * ispisuje `f"Ulazim u f1({x})"`
   * u varijablu `v` postavlja vrijednost dobivenu pozivom `f2(x)` povećanu za `1`
   * ispisuje `f"Izlazim iz f1({x}) i vraćam: {v}"`
   * vraća `v`

4. U glavnom programu:

   * pozovi funkciju `f1(10)`
   * ispiši rezultat

````python
def f3(z):
    print(f"     └→ Ulazim u f3({z})")
    print(f"     ←─ Izlazim iz f3({z}) i vraćam: {z * 2}")
    return z * 2

def f2(y):
    print(f"   └→ Ulazim u f2({y})")
    v = f3(y) + 5
    print(f"   ←─ Izlazim iz f2({y}) i vraćam: {v}")
    return v

def f1(x):
    print(f" └→ Ulazim u f1({x})")
    v = f2(x) + 1
    print(f" ←─ Izlazim iz f1({x}) i vraćam: {v}")
    return v


v = f1(10)
print(f"Rezultat: {v}")
````

## 4.7. Rekurzija

Rekurzija nastaje kada funkcija izravno ili neizravno poziva samu sebe. To znači da se isti blok koda izvršava više puta.

**Primjer najjednostavnije rekurzije**
```Python
def f():
    print("└→ Ulazim u f()")
    f()
    print("←─ Izlazim iz f()")

f()
```

Slijed poziva sa stogom poziva:

```python
                       | Stog    |
 → f()                 | f       |
     └→ f()            | f, f    |
          └→ f()       | f, f, f |
             ...       | ...     |
```

Prilikom pokretanja ovog primjera ispisat će se

```
└→ Ulazim u f()
└→ Ulazim u f()
└→ Ulazim u f()
└→ Ulazim u f()
└→ Ulazim u f()
...
└→ Ulazim u f()
RecursionError: maximum recursion depth exceeded
```

Nikad se neće ispisati `←─ Izlazim iz f()` i pojavit će se greška `RecursionError: maximum recursion depth exceeded` jer je stog poziva ograničene veličine i u Pythonu može sadržavati maksimalno 1000 poziva funkcija.




--- 

#### Rekurzija s parametrima

Kako bi rekurzija imala smisla, moramo joj prenijeti parametre, unutar funkcije promijeniti barem jednu njihovu vrijednost i pozivati istu funkciju kojoj ćemo prenijeti promijenjene parametre.

**Primjer rekurzije s parametrima**
```Python
def f(n):
    print(f"└→ Ulazim u f({n})")
    f(n - 1)
    print(f"←─ Izlazim iz f({n})")

f(3)
```

Slijed poziva sa stogom poziva:

```python
                        | Stog             |
 → f(3)                 | f(3)             |
     └→ f(2)            | f(3), f(2)       |
          └→ f(1)       | f(3), f(2), f(1) |
             ...        | ...              |
```


Prilikom pokretanja ovog primjera ispisat će se

```
└→ Ulazim u f(3)
└→ Ulazim u f(2)
└→ Ulazim u f(1)
└→ Ulazim u f(0)
└→ Ulazim u f(-1)
...
└→ Ulazim u f(-996)
RecursionError: maximum recursion depth exceeded
```

#### Uvjet prekida rekurzije

Rekurzija obavezno mora imati uvijet prekida, kada nemamo prekid rekurzija će se (teoretski) ponavljati do beskonačnosti. 

Ovim smo tijelo rekurzivne funkcije podijelili na:
* **Bazni slučaj** - dio u kojem se zaustavlja rekurzivni slijed poziva, sastoji se od uvjeta prekida i vraćanja bazne vrijednosti.
* **Rekurzivni slučaj** - dio u kojem se događa rekurzivni poziv

**Primjer rekurzije s uvjetom prekida**
```Python
def f(n):
    print(f"└→ Ulaz u f({n})")
    # bazni slučaj
    if n == 0:
        print(f"←─ Bazni izlaz iz f({n})")
        return

    # rekurzivni slučaj
    f(n - 1)
    print(f"←─ Rekurzivni izlaz iz f({n})")
    return

f(3)
```

Slijed poziva sa stogom poziva:

```python
                          | Stog                   |
 → f(3)                   | f(3)                   |
     └→ f(2)              | f(3), f(2)             |
          └→ f(1)         | f(3), f(2), f(1)       |
               └→ f(0)    | f(3), f(2), f(1), f(0) |
               ←─ return  | f(3), f(2), f(1)       |
          ←─ return       | f(3), f(2)             |
      ←─ return           | f(3).                  |          
 ←─ return                |                        |               
```

Prilikom pokretanja ovog primjera ispisat će se

```
└→ Ulaz u f(3)
└→ Ulaz u f(2)
└→ Ulaz u f(1)
└→ Ulaz u f(0)
←─ Bazni izlaz iz f(0)
←─ Rekurzivni izlaz f(1)
←─ Rekurzivni izlaz f(2)
←─ Rekurzivni izlaz f(3)
```


## 4.7.1 Linearna rekurzija

Rekurzivna funkcija može imati više rekurzivnih poziva, a kada ima samo jedan, radi se o **linearnoj rekurziji**.

Kako bi rekurzija bila korisna potrebno je da vraća smislene vrijednosti. 

Pri tome se rezultat rekurzivnog poziva iskorištava za dobivanje krajnjeg rezultata.

Opća struktura linearne rekurzije je sljedeća:

* Bazni slučaj rekurzije prekida tijek rekurzivnih poziva i sastoji se od dva dijela:
    1. **provjere** je li ulazni parametar **bazni**
    2. **vraćanje** baznog rezultata
* Rekurzivni slučaj se sastoji od tri dijela:
    4. **redukcija** ulaznog parametra na **sljedeći manji ulazni parametar** (reducirani ulazni parametar)
    5. **pozivanje** rekurzije nad **reduciranom ulaznim parametrom**, dobiva se rekurzivni rezultat
    6. **kombiniranje** rezultata rekurzije s **ulaznim parametrom**, dobiva se ukupni rezultat

Pseudokod rekurzije je:
```Python
def funkcija(ulaz):
    if je_bazni(ulaz):               # 1. provjera
        return bazni_rezultat        # 2. vraćanje
    
    reducirani_ulaz = reduciraj(ulaz)                        # 3. redukcija
    rekurzivni_rezultat = funkcija(reducirani_ulaz)          # 4. pozivanje 
    ukupni_rezultat = kombiniraj(ulaz, rekurzivni_rezultat)  # 5. kombiniranje
    return ukupni_rezultat
```

> 💡 **Napomena**
>
> Za uspješno definiranje rekurzivne funkcije potrebno je za ulazni parametar `ulaz` odrediti:
> * provjeru baznog slučaja `je_bazni(ulaz)`, 
> * bazni rezultat `bazni_rezultat`
> * redukciju ulaznog parametra `rekurzivni_rezultat=reduciraj(ulaz)`
> * kombinaciju rezultata rekurzije s ulaznim parametrom `kombiniraj(ulaz, rekurzivni_rezultat)`
> 
> **Vrlo važno** je shvatiti da se rekurzivnim pozivom nad reduciranim ulazom dobiva rezultat za reducirani ulaz.
> 
> Taj rezultat se onda kombinira s ulazom za dobivanje ukupnog rezultata.

Kod jednostavnih rekurzija često se reduciranje ulaza, rekurzivi poziv i kombiniranje izvršava u jednoj liniji odnosno:

```Python
def funkcija(ulaz):
    if je_bazni(ulaz):
        return bazni_rezultat

    return kombiniraj(ulaz, funkcija(reduciraj(ulaz)))
```

#### Primjer: suma brojeva

Za dani pozitivni cijeli broj `n` rekurzijom izračunaj sumu pozitivnih cijelih brojeva do uključivo `n`.

> 💡 **Napomena**
>
> Kako bi lakše osmislili rekurziju, potrebno je odrediti vrijednosti i način dobivanja vrijednosti za dijelove rekurzije.
> * funkcija: `suma` 
> * ulaz: `n`
> * je bazni: `n == 0`
> * bazni izlaz: `0`
> * redukcija ulaza: `n - 1`
> * rekurzivni rezultat: `suma(n - 1)` vraća `1 + 2 + 3 + ... + (n - 1)`
> * kombinacija: `n + rekurzivni_rezultat` (ako znamo sumu do uključivo `n-1` potrebno joj je dodati `n` da bi dobili sumu do uključivo `n`).

**Rješenje:**

```Python
def suma(n):
    if n == 0:
        return 0
    reducirani = n - 1
    rekuzivni = suma(reducirani)
    ukupni = n + rekurzivni
    return ukupni
```

**Slijed izvršavanja:**

```python
 → suma(3)
     reducirani = 3 - 1 = 2
     rekuzivni = suma(reducirani) = suma(2)
     └→ suma(2)
          reducirani = 2 - 1 = 1
          rekuzivni = suma(reducirani) = suma(1)
          └→ suma(1)
               reducirani = 1 - 1 = 0
               rekuzivni = suma(reducirani) = suma(0)
               └→ suma(0)    
               ←─ return 0  
               rekuzivni = 0
               ukupni = 1 + rekurzivni = 1 + 0 = 1
          ←─ return 1
          rekuzivni = 1
          ukupni = 2 + rekurzivni = 2 + 1 = 3
      ←─ return 3
      rekuzivni = 3
      ukupni = 3 + rekurzivni = 3 + 3 = 6
 ←─ return 6
```

**Skraćena verzija rješenja:**

```Python
def suma(n):
    if n == 0:
        return 0
    return n + suma(n - 1)
```

**Slijed izvršavanja:**

```python
 → suma(3)                   
     └→ suma(2)              
          └→ suma(1)         
               └→ suma(0)    
               ←─ return 0  
          ←─ return 1 + suma(0) = 1 + 0 = 1
```

#### Primjer: suma liste

Za danu listu brojeva rekurzijom izračunaj sumu svih brojeva u njoj.
Na primjer, za listu `[4, 1, 3, 2]` se vraća `10`.

> * funkcija: `sumaliste` 
> * ulaz: `lista`
> * je bazni: `lista == []`
> * bazni izlaz: `0`
> * redukcija ulaza: `lista[:-1]` - lista bez zadnjeg elementa
> * rekurzivni rezultat: `sumaliste(lista[-1])` vraća `4 + 1 + 3 = 8`
> * kombinacija: `lista[-1] + rekurzivni_rezultat` (ako znamo sumu liste bez zadnjeg elementa `8`, potrebno joj je dodati zadnji element `2` da bi dobili konačnu sumu).

**Rješenje:**

```Python
def sumaliste(lista):
    if lista == []:
        return 0
    reducirani = lista[:-1]
    rekurzivni = sumaliste(reducirani)
    ukupni = lista[-1] + rekurzivni
    return ukupni
```

**Slijed izvršavanja:**

```Python
 → sumaliste([4, 1, 3, 2])
     reducirani = [4, 1, 3]                               # [4, 1, 3, 2][:-1]
     rekuzivni = sumaliste([4, 1, 3])                     # sumaliste(reducirani) 
     └→ sumaliste([4, 1, 3])
          reducirani = [4, 1]                             # [4, 1, 3][:-1]
          rekuzivni = sumaliste([4, 1])                   # sumaliste(reducirani)
          └→ sumaliste([4, 1])
               reducirani = [4]                           # [4, 1][:-1]
               rekuzivni =  sumaliste([4])                # sumaliste(reducirani)
               └→ sumaliste([4])    
                    reducirani = []                       # [4][:-1]
                    rekurzivni = sumaliste([])            # sumaliste(reducirani) 
                    └→ sumaliste([])
                    ←─ return 0  
                    rekuzivni = 0
                    ukupni = 4 + rekurzivni = 4 + 0 = 4  # [4][-1] + rekurzivni
               ←─ return 4
               rekuzivni = 4
               ukupni = 1 + rekurzivni = 1 + 4 = 5       # [4, 1][-1] + rekurzivni
          ←─ return 5
          rekuzivni = 5
          ukupni = 3 + rekurzivni = 3 + 5 = 8            # [4, 1, 3][-1] + rekurzivni
     ←─ return 8
     rekurzivni = 8
     ukupni = 2 + rekurzivni = 2 + 8 = 10                # [4, 1, 3, 2][-1] + rekurzivni
←─ return 10
```

**Skraćena verzija rješenja:**

```Python
def sumaliste(lista):
    if not lista:
        return 0
    return lista.pop() + sumaliste(lista)
```

**Slijed izvršavanja:**

```Python

   └→ sumaliste([4, 1, 3])

               ←─ return 0  
           ←─ return 4 + 0 = 4
       ←─ return 1 + 4 = 5

←─ return 2 + 8 = 10
```

✏️ **Zadatak**

1.  Za zadanu listu brojeva napiši rekurzivnu funkciju `prebroji` koja vraća koliko puta se zadani broj pojavljuje u listi. 
2.  Provjeri vrijednost rekurzivne funkcije za listu `[1, 2, 3, 2, 4, 2]` i za brojeve `1`, `2`, `3`, `4`, `5`. 

````python
def prebroji(lista, broj):
    if not lista:
        return 0
    
    if lista[-1] == broj:
        return 1 + prebroji(lista[:-1], broj)
    else:
        return prebroji(lista[:-1], broj)

for broj in range(1, 6):
    koliko = prebroji([1, 2, 3, 2, 4, 2], broj)
    print(f"Broj {broj} se pojavljuje {koliko} puta u listi {[1, 2, 3, 2, 4, 2]}.")
    
````
