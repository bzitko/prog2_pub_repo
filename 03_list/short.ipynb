# 2. Tipovi podataka


## 2.5. Lista: `list`

Lista (`list`) je **uređena kolekcija objekata**.
Elementi mogu biti različitih tipova.
Pristupa im se preko **indeksa (od 0)**.
Liste su **promjenjive (mutable)**.

---

#### Definiranje liste

Definira se pomoću **`[]`**, elementi se odvajaju zarezima.

**Primjeri:**

```python
prazna = []
brojevi = [1, 2, 3, 4, 5]
imena = ["Ana", "Marko", "Ivana"]
razno = [10, "Python", True, 3.14]
lista_u_listi = [[1, 2], 3, [4, [5, 6]]]
```

✏️ **Zadatak:**

1. Definiraj listu `boje` (3 boje).
2. Definiraj listu `brojevi` (prvih 5 prirodnih brojeva).
3. Definiraj listu `podaci` (ime, godina, `True`).
4. Ispiši sve liste pomoću `print()`.

````python
boje = ["crvena", "bijela", "plava"]
brojevi = [1, 2, 3, 4, 5]
podaci = ("Branko", 50, True)

print(boje)
print(brojevi)
print(podaci)
````

#### Indeksiranje

Svaki element ima **indeks (od 0)**.
Omogućuje dohvat i promjenu elemenata.

* **Pozitivni**: od početka (0, 1, …)
* **Negativni**: od kraja (-1, -2, …)

**Primjer:**

```python
elementi = [123, "abc", True, None]
print(elementi[0])
print(elementi[2])
print(elementi[-1])
print(elementi[-2])
```

Elemente možemo i **mijenjati preko indeksa**.

**Primjer:**

```python
elementi = [123, "abc", True, None]
elementi[1] = "xy"
elementi[-2] = False
print(elementi)
```

✏️ **Zadatak:**

1. Definiraj `objekti = ["abc", 100, [2.718, 3.14]]`
2. Ispiši prvi element (pozitivni indeks).
3. Ispiši zadnji element (negativni indeks).
4. Promijeni prvi element u `"xy"`.
5. Povećaj drugi element za 10.
6. Iz zadnjeg elementa ispiši drugi element.


````python
objekti = ["abc", 100, [2.718, 3.14]]
print(objekti[0])
print(objekti[-1])
objekti[0] = "xy"
objekti[1] += 10
print(objekti)
print(objekti[-1][1])
````

#### Rezanje (slicing)

Dohvat dijela liste pomoću:

```
lista[start:stop:step]
```

* `start` – početni indeks
* `stop` – indeks iza zadnjeg
* `step` – korak (default 1)

Ako izostavimo:

* `start` → od početka
* `stop` → do kraja
* `step` → 1

**Primjeri:**

```python
brojevi = [10, 20, 30, 40, 50, 60]

print(brojevi[::2])
print(brojevi[1::2])
print(brojevi[::-1])
```

Rezanje može i **mijenjati više elemenata odjednom**.

---

✏️ **Zadatak:**

1. Definiraj `brojevi = [1, 2, 3, 4, 5, 6, 7, 8]`.
2. Promijeni prvi element u 1000.
3. Promijeni elemente 2–4 u `[99, 100, 101]`.
4. Postavi svaki drugi element na 0.
5. Obriši zadnja dva elementa (rezanjem).
6. Ispiši listu nakon svake promjene.


````python
brojevi = [1, 2, 3, 4, 5, 6, 7, 8]
brojevi[0] = 1000
print(brojevi)
brojevi[2:5] = [99, 100, 101]
print(brojevi)
brojevi[::2] = [0] * 4
print(brojevi)
brojevi[-2:] = []
print(brojevi)
````

#### Operatori spajanja i ponavljanja

Liste se mogu **spajati (`+`)** i **ponavljati (`*`)**.

**Primjeri:**

```python
a = [1, 2, 3]
b = [4, 5, 6]
print(a + b)

boje = ["crvena", "zelena"]
print(boje * 3)
```

> **Napomena:** 
>
>`+` i `*` vraćaju novu listu.

---

✏️ **Zadatak:**

1. Definiraj `a = [1, 2, 3]` i `b = [4, 5]`.
2. Napravi `c = A + B`.
3. Napravi `d = B * 4`.
4. Ispiši `c` i `d`.
5. Usporedi `b + a * 2` i `(b + a) * 2`.


````python
a = [1, 2, 3]
b = [4, 5]
c = a + b
d = b * 4
print(c)
print(d)
print(b + a * 2)
print((b + a) * 2)
````

Provjera pripadnosti pomoću **`in`** i **`not in`**.
Vraća `True` ili `False`.

**Primjer:**

```python

20 in brojevi      # False

```

> **Napomena:**
>
> Provjerava se **cijeli element**, ne dio.

---

✏️ **Zadatak:**

1. Definiraj `gradovi = ["Zagreb", "Split", "Rijeka", "Osijek"]`.
2. Provjeri:

   * `"Split"` u listi,
   * `"Zadar"` nije u listi.
3. Ako `"Osijek"` postoji, ispiši poruku.

#### Ugrađene funkcije za liste

Python nudi **ugrađene funkcije** za dohvat informacija ili stvaranje novih lista.
One **ne mijenjaju listu** osim ako se rezultat ne dodijeli.

| Funkcija           | Opis                            | Primjer                             |
| ------------------ | ------------------------------- | ----------------------------------- |
| `len(lista)`       | Broj elemenata                  | `len([1,2,3]) → 3`                  |
| `sum(lista)`       | Zbroj svih brojeva              | `sum([1,2,3]) → 6`                  |
| `max(lista)`       | Najveći element                 | `max([4,1,9]) → 9`                  |
| `min(lista)`       | Najmanji element                | `min([4,1,9]) → 1`                  |
| `sorted(lista)`    | Nova sortirana lista            | `sorted([3,1,2]) → [1,2,3]`         |
| `reversed(lista)`  | Iterator u obrnutom redoslijedu | `list(reversed([1,2,3])) → [3,2,1]` |
| `all(lista)`       | `True` ako svi elementi True    | `all([1,2,3]) → True`               |
| `any(lista)`       | `True` ako barem jedan True     | `any([0,0,1]) → True`               |
| `list(iterabilno)` | Pretvara u listu                | `list("abc") → ['a','b','c']`       |

> **Napomena:**
>
> `reversed()` vraća iterator → `list(reversed(...))` za listu.  
> `list(range(...))` pretvara iterator u listu.  
> `list()` bez argumenata → prazna lista `[]`.  
> `sorted()` može koristiti `key` i `reverse`.  

---

✏️ **Zadatak:**

1. Definiraj `brojevi = [4, 7, 1, 9, 3]`.
2. Ispiši: duljinu, zbroj, najveći i najmanji element.
3. Napravi `sortirani = sorted(brojevi)`.
4. Ispiši listu obrnutog redoslijeda (`reversed()`).
5. Provjeri `all()` i `any()` za elemente različite od 0.
6. Napravi listu `[10, 20, 30, 40, 50]` koristeći `range()` i `list()`.


````python
brojevi = [4, 7, 1, 9, 3]
print("Duljina liste:", len(brojevi))
print("Zbroj svih brojeva", sum(brojevi))
print("Najmanji broj:", min(brojevi))
print("Najveći broj:", max(brojevi))
sortirani = sorted(brojevi)
print("Sortirano:", sortirani)
print("Svi brojevi različiti od 0:", all(brojevi))
print("Barem jedan broj različit od 0:", any(brojevi))
print("Lista iz range:", list(range(10, 51, 10)))


````

#### Metode za dodavanje elemenata

Liste su **promjenjive**, možemo im **dodavati elemente**.

**Metode:**

| Metoda               | Opis                     | Primjer                | Rezultat     |
| -------------------- | ------------------------ | ---------------------- | ------------ |
| `append(x)`          | Dodaje na kraj           | `[1,2,3].append(4)`    | `[1,2,3,4]`  |
| `insert(i, x)`       | Umeće na poziciju `i`    | `[1,2,3].insert(1,10)` | `[1,10,2,3]` |
| `extend(iterabilno)` | Dodaje elemente iz liste | `[1,2].extend([3,4])`  | `[1,2,3,4]`  |

> **Napomena:**
>
> `append()` dodaje jedan element,  
> `insert()` umetne bilo gdje,  
> `extend()` dodaje element po element.

---

✏️ **Zadatak:**

1. Definiraj `boje = ["crvena", "zelena"]`.
2. Dodaj `"plava"` (`append()`).
3. Umetni `"žuta"` na početak (`insert()`).
4. Dodaj `["ljubičasta", "narančasta"]` (`extend()`).
5. Ispiši listu nakon svake promjene.


````python
boje = ["crvena", "zelena"]
print(boje)
boje.append("plava")
print(boje)
boje.insert(0, "žuta")
print(boje)
boje.extend(["bijela", "crna"])
print(boje)
````

#### Metode za uklanjanje elemenata

Liste su **promjenjive**, elemente možemo uklanjati.

| Metoda      | Opis                       | Primjer               | Rezultat             |
| ----------- | -------------------------- | --------------------- | -------------------- |
| `remove(x)` | Uklanja prvo pojavljivanje | `[1,2,3,2].remove(2)` | `[1,3,2]`            |
| `pop(i=-1)` | Uklanja i vraća element    | `[10,20,30].pop(1)`   | `[10,30]` (vraća 20) |
| `clear()`   | Briše sve elemente         | `[1,2,3].clear()`     | `[]`                 |

> **Napomena:**
>
> `remove()` uklanja prvo pojavljivanje,  
> `pop()` vraća element,  
> `clear()` prazni listu ali ne briše varijablu.

---

✏️ **Zadatak:**

1. Definiraj `brojevi = [1, 2, 3, 2, 4]`.
2. Ukloni prvo `2` (`remove()`).
3. Ukloni zadnji element (`pop()`).
4. Ispiši listu nakon svake promjene.
5. Prazni listu (`clear()`) i provjeri sadržaj.


````python
brojevi = [1, 2, 3, 2, 4]
brojevi.remove(2)
print(brojevi)
brojevi.pop()   # brojevi.pop(-1)
print(brojevi)
brojevi.clear()
print(brojevi)
````

#### Metode za pretraživanje i brojenje

Liste omogućuju **pronalazak i brojanje elemenata**.

| Metoda      | Opis                        | Primjer                   | Rezultat |
| ----------- | --------------------------- | ------------------------- | -------- |
| `.index(x)` | Indeks prvog pojavljivanja  | `[10,20,30,20].index(20)` | `1`      |
| `.count(x)` | Broj pojavljivanja elementa | `[1,2,2,3].count(2)`      | `2`      |

> **Napomena:**
>
> `.index(x)` → ValueError ako element ne postoji
> `.count(x)` → vraća `0` ako element ne postoji

---

✏️ **Zadatak:**

4. Pronađi indeks drugog `3`.

#### Metode za sortiranje i obrtanje

Liste se mogu **sortirati** i **obrnuti**.
Obje metode **mijenjaju originalnu listu**.

| Metoda            | Opis                   | Primjer             | Rezultat  |
| ----------------- | ---------------------- | ------------------- | --------- |
| `lista.sort()`    | Sortira elemente liste | `[3,1,2].sort()`    | `[1,2,3]` |
| `lista.reverse()` | Obrće redoslijed       | `[1,2,3].reverse()` | `[3,2,1]` |

> **Napomena:**
>
> `sort()` i `reverse()` ne vraćaju novu listu, već mijenjaju postojuću ; ako treba original, kopiraj listu.

---

✏️ **Zadatak:**

1. Definiraj `brojevi = [8, 2, 5, 1, 9, 4]`.
2. Sortiraj listu (`sort()`).
3. Obrni redoslijed (`reverse()`).
4. Provjeri efekt dvostrukog `reverse()`.


````python
brojevi = [8, 2, 5, 1, 9, 4]
print(brojevi)
brojevi.sort()
print(brojevi)
brojevi.reverse()
print(brojevi)
brojevi.reverse()
brojevi.reverse()
print(brojevi)
````

#### Skraćeni oblik liste (list comprehension)

Brzo stvaranje lista pomoću **list comprehension**:

```python
[izraz for element in iterator if uvjet]
```

* `izraz` – vrijednost u novu listu
* `element` – trenutni element
* `iterator` – iterabilni objekt
* `if uvjet` – **neobavezno** filtriranje

**Primjeri:**

```python
kvadrati = [x**2 for x in range(6)]
parni = [x for x in range(11) if x % 2 == 0]
rijeci = ["python", "je", "super"]
velika = [r.upper() for r in rijeci]
```

> **Napomena:**
>
> Može uključivati uvjete i transformacije; pogodno za jednostavne i srednje složene liste.

---

✏️ **Zadatak:**

1. Napravi lsitu `brojevi` od 0 do 20.
2. Napravi listu neparnih brojeva iz `brojevi`.
3. Napravi listu kvadrata tih neparnih brojeva.
4. Iz `['jabuka','limun','banana','kokos']` napravi listu voća u velikim slovima koja sadrži `"a"`.


````python
brojevi = [br for br in range(20)]
print(brojevi)
neparni = [br for br in brojevi if br % 2 == 1]
print(neparni)
kvadrati = [br ** 2 for br in neparni]
print(kvadrati)
voca = ['jabuka','limun','banana','kokos']
velika = [v.upper() for v in voca if "a" in v]
print(velika)
````

#### Sortiranje i obrtanje

Python nudi **sortiranje i obrnut redoslijed** pomoću metoda i funkcija:

* `.sort()` – mijenja originalnu listu
* `sorted()` – vraća novu sortiranu listu
* `.reverse()` – obrće redoslijed liste
* `reversed()` – vraća iterator obrnutog redoslijeda

```python
brojevi = [5,2,9,1,7]
brojevi.sort()             # mijenja listu
nove = sorted(brojevi)     # nova lista
brojevi.sort(reverse=True) # silazno
```

```python
boje = ["crvena","zelena","plava"]
boje.reverse()                     # obrće redoslijed
obrnute = list(reversed(boje))     # nova lista
```

> **Napomena:**
>
> Elementi moraju biti usporedivi.
> `sort()` i `sorted()` imaju argumente:
>
> * `reverse=True` → silazno
> * `key=...` → kriterij sortiranja (npr. `len` ili vlastita funkcija)


Primjer korištenja ugrađene funkcije.

```python
rijeci = ["AI","programiranje","kod","python"]
rijeci.sort(key=len)  # po duljini
```

Primjer korištenja vlastite funkcije.

```python
def zadnja_znamenka(broj):
    return broj % 10
brojevi = [42,17,93,58,21]
brojevi.sort(key=zadnja_znamenka)
```

---

✏️ **Zadatak:**

1. Definiraj `brojevi = [42, 7, 19, 3, 25, 11]`.

   * Sortiraj uzlazno, zatim silazno.
   * Napravi novu sortiranu listu (`sorted()`).
2. Definiraj `rijeci = ["sunce","snijeg","oblak","kiša","vjetar"]`.

   * Sortiraj po duljini (`key=len`).
   * Obrni redoslijed (`reversed()`).
3. Napiši `zadnji_znak(rijec)` koja vraća zadnji znak i koristi kao `key`.


````python
brojevi = [42, 7, 19, 3, 25, 11]
print("brojevi:", brojevi)
brojevi.sort()
print("uzlazno:", brojevi)
brojevi.sort(reverse=True)
print("silazno:", brojevi)

rijeci = ["sunce", "snijeg", "oblak", "kiša", "vjetar"]
print("riječi:", rijeci)
rijeci.sort(key=len)
print("po duljini:", rijeci)
rijeci = list(reversed(rijeci))
print("obrnuto:", rijeci)

def zadnji_znak(rijec):
    return rijec[-1]

rijeci = sorted(rijeci, key=zadnji_znak)
print("po zadnjem znaku:", rijeci)

````
