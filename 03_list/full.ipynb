# 2. Tipovi podataka


## 2.5. Lista: `list`

Lista (`list`) je **indeksirana kolekcija objekata** u Pythonu, namijenjena za pohranu bilo kojih objekata.  
Elementi liste mogu biti bilo kojeg tipa (brojevi, stringovi, logičke vrijednosti, pa čak i druge liste).  
Elementi u listi su **uređeni**, što znači da imaju svoj **indeks** (počinje od 0).    
Liste su **promjenjive** (mutable), što znači da im elemente možemo mijenjati nakon stvaranja.  

---

#### Definiranje liste

Liste se definiraju **uglatim zagradama `[]`**, a elementi se odvajaju **zarezima**.

**Primjeri:**

```python
# Prazna lista
prazna = []

# Lista cijelih brojeva
brojevi = [1, 2, 3, 4, 5]

# Lista stringova
imena = ["Ana", "Marko", "Ivana"]

# Mješovita lista
razno = [10, "Python", True, 3.14]

# Lista u listi (ugnježđena lista)
lista_u_listi = [[1, 2], 3, [4, [5, 6]]]
```

✏️ **Zadatak:**

1. Definiraj listu `boje` koja sadrži tri boje po tvom izboru.
2. Definiraj listu `brojevi` koja sadrži prvih pet prirodnih brojeva.
3. Napravi listu `podaci` koja sadrži ime (string), godinu (cijeli broj) i informaciju `True` ako voliš programirati.
4. Ispiši sve tri liste pomoću `print()` funkcije.

````python
boje = ["crvena", "bijela", "plava"]
brojevi = [1, 2, 3, 4, 5]
podaci = ("Branko", 50, True)

print(boje)
print(brojevi)
print(podaci)
````

#### Indeksiranje

Svaki element u listi ima svoj **indeks**, počevši od 0 za prvi objekt. Indeksiranje nam omogućuje da dohvatimo ili provjerimo pojedinačne elemente unutar liste.

* **Pozitivni indeksi**: Broje elemente od početka liste (prvi element ima indeks 0, drugi 1, itd.).
* **Negativni indeksi**: Broje elemente od kraja liste (posljednji element ima indeks -1, pretposljednji -2, itd.).

**Primjer:**
```python
elementi = [123, "abc", True, None]
print(elementi[0])    # 123 je prvi element
print(elementi[2])    # 'abc' je treći element
print(elementi[-1])   # None je posljednji element
print(elementi[-2])   # True je pretposljednji element
```

Možemo i mijenjati elemente liste elemente tako da im pristupimo preko indeksa i dodijelimo novu vrijednost


**Primjer:**
```python
elementi = [123, "abc", True, None]
elementi[1] = "xy"     # mijenja drugi element
elementi[-2] = False   # mijenja predzadnji element
print(elementi)        # [123, "xy", False, None]
```

✏️ **Zadatak:**

1. Definiraj listu `objekti = ["abc", 100, [2.718, 3.14]]`
2. Ispiši prvi element liste koristeći pozitivni indeks.
3. Ispiši posljednji element liste koristeći negativni indeks.
4. Promijeni prvi element liste u `"xy"`.
5. Povečaj drugi element liste za 10
6. U posljednjem elementu liste vrati drugi element.

````python
objekti = ["abc", 100, [2.718, 3.14]]
print(objekti[0])
print(objekti[-1])
objekti[0] = "xy"
objekti[1] += 10
print(objekti)
print(objekti[-1][1])
````

#### Rezanje (slicing)

Python omogućuje dohvat dijelova liste pomoću **rezanja** (slicing). Rezanje nam daje podlistu bez mijenjanja izvorne liste. Sintaksa je:

```
lista[start:stop:step]
```

* **start** – indeks prvog elementa koji želimo uključiti
* **stop** – indeks elementa **iza** posljednjeg elementa kojeg želimo uključiti
* **step** – korak kojim prolazimo kroz listu (po defaultu 1)

Ako izostavimo `start`, rezanje počinje od početka liste.
Ako izostavimo `stop`, rezanje ide do kraja liste.
Ako izostavimo `step`, koristi se korak 1.

**Primjeri:**

```python
brojevi = [10, 20, 30, 40, 50, 60]

print(brojevi[::2])   # [10, 30, 50]  -> svaki drugi element
print(brojevi[1::2])  # [20, 40, 60]
print(brojevi[::-1])  # [60, 50, 40, 30, 20, 10] -> obrnuti redoslijed
```

>  **Napomena**  
> lista je promjenjivi tip podatka, stoga se i pomoću rezanja može promijeniti više elemenata odjednom.  
> Na primjer, ako je `brojevi = [10, 20, 30, 40, 50, 60]` onda će `brojevi[1:4] = [200, 300]` promijeniti listu `brojevi`.  
> Sada će `brojevi` imati vrijednost `[10, 200, 400, 50, 60]`.

✏️ **Zadatak:**
1. Definiraj listu brojevi = [1, 2, 3, 4, 5, 6, 7, 8].
2. Promijeni prvi element u 1000.
3. Promijeni srednja tri elementa (na pozicijama 2–4) u [99, 100, 101].
4. Postavi svaki drugi element na vrijednost 0.
5. Obriši zadnja dva elementa pomoću rezanja.
6. Ispiši listu nakon svake promjene.


````python
brojevi = [1, 2, 3, 4, 5, 6, 7, 8]
brojevi[0] = 1000
print(brojevi)
brojevi[2:5] = [99, 100, 101]
print(brojevi)
brojevi[::2] = [0] * 4
print(brojevi)
brojevi[-2:] = []
print(brojevi)
````

#### Operatori spajanja i ponavljanja

Python omogućuje **spajanje (konkatenaciju)** i **ponavljanje** lista pomoću jednostavnih operatora.
Na taj način možemo lako stvarati nove liste bez potrebe za petljama ili dodatnim funkcijama.

**Primjeri:**

```python
# Spajanje lista
a = [1, 2, 3]
b = [4, 5, 6]
spojena = a + b
print(spojena)  # [1, 2, 3, 4, 5, 6]

# Ponavljanje liste
boje = ["crvena", "zelena"]
ponovljene = boje * 3
print(ponovljene)  # ['crvena', 'zelena', 'crvena', 'zelena', 'crvena', 'zelena']
```

>  **Napomena**  
> Operator `+` ne mijenja postojeće liste, nego vraća **novu listu**.  
> Operator `*` stvara **ponovljenu kopiju** liste — elemente jednostavno ponavlja u istom redoslijedu.

✏️ **Zadatak:**

1. Definiraj liste
   `a = [1, 2, 3]` i `b = [4, 5]`.
2. Napravi novu listu `c` koja spaja `a` i `b`.
3. Napravi listu `d` koja ponavlja elemente liste `b` **četiri puta**.
4. Ispiši obje nove liste.
5. Promijeni redoslijed operacija i provjeri rezultat izraza `b + a * 2` i `(b + a) * 2`.

````python
a = [1, 2, 3]
b = [4, 5]
c = a + b
d = b * 4
print(c)
print(d)
print(b + a * 2)
print((b + a) * 2)
````

#### Operatori pripadnosti

Python omogućuje jednostavnu provjeru **pripadnosti elementa nekoj listi** pomoću operatora `in`. Ovaj operator vraća **`True` ili `False`** ovisno o tome postoji li traženi element u listi. Operator `not in` vraća **True** ako element **nije prisutan** u listi.

>  **Napomena**
> Operator `in` provjerava **točno podudaranje elemenata** u listi.
> Za razliku od **stringova**, gdje možemo provjeriti **podstring** (npr. `"an" in "banana"` → True),
> u **listi** se traži **cijeli element**, a ne njegov dio.
>
> Primjer:
>
> ```python
> voce = ["jabuka", "banana", "trešnja"]
> print(["banana", "trešnja"] in voce)      # False
> print("banana" in voce)  # True
> ```

**Primjeri:**

```python
brojevi = [1, 2, 3, 4, 5]

print(20 in brojevi)      # False
print(10 not in brojevi)  # True
```


✏️ **Zadatak:**

1. Definiraj listu `gradovi = ["Zagreb", "Split", "Rijeka", "Osijek"]`.
2. Provjeri:
   * nalazi li se `"Split"` u listi,
   * ne nalazi li se `"Zadar"` u listi.
3. Provjeri postoji li `"Osijek"` u listi i ako postoji, ispiši poruku `"Osijek je već dodan."`

````python
gradovi = ["Zagreb", "Split", "Rijeka", "Osijek"]
print(gradovi)
print("Split" in gradovi)
print("Zadar" not in gradovi)
if "Osijek" in gradovi:
    print("Osijek je već dodan.")
````

#### Ugrađene funkcije za liste

Osim metoda koje su dio samog tipa `list`, Python nudi i **ugrađene funkcije** koje možemo primijeniti na liste. One **ne mijenjaju listu** (osim ako se izričito ne dodijeli rezultat) i olakšavaju dobivanje informacija ili stvaranje novih lista.

| Funkcija           | Opis                                                     | Primjer                             |
| ------------------ | -------------------------------------------------------- | ----------------------------------- |
| `len(lista)`       | Vraća broj elemenata u listi                             | `len([1,2,3]) → 3`                  |
| `sum(lista)`       | Zbraja sve brojeve u listi                               | `sum([1,2,3]) → 6`                  |
| `max(lista)`       | Vraća najveći element                                    | `max([4,1,9]) → 9`                  |
| `min(lista)`       | Vraća najmanji element                                   | `min([4,1,9]) → 1`                  |
| `sorted(lista)`    | Vraća novu listu sortiranu po rastućem redoslijedu       | `sorted([3,1,2]) → [1,2,3]`         |
| `reversed(lista)`  | Vraća iterator za listu u obrnutom redoslijedu           | `list(reversed([1,2,3])) → [3,2,1]` |
| `all(lista)`       | Vraća `True` ako su svi elementi **logički True**        | `all([1,2,3]) → True`               |
| `any(lista)`       | Vraća `True` ako je **barem jedan element logički True** | `any([0,0,1]) → True`               |
| `list(iterabilno)` | Pretvara iterabilni objekt u listu                       | `list("abc") → ['a','b','c']`       |

>  **Napomena**  
> 
> * Funkcija `reversed()` za primljenu indeksiranu kolekciju kao što je `string` i `list` vraća iterator za listu u obrnutom redoslijedu. Kako bi se iterator pretvorio u listu, koristi se funkcija `list()`.  
> * Pošto `range()` vraća iterator, uz pomoć `list()` se dobiveni iterator pretvara u listu. Na primjer, `list(range(5))` daje listu `[0, 1, 2, 3, 4]`.  
> * Funkcija `list()` se također može koristiti i bez argumenata. U tom slućaju se dobije nova prazna lista `[]`.
> * Funkcija `sorted()` ima argumente s zadanim vrijednostima:
>   * `key=None` - 
>   * `reverse=False`

✏️ **Zadatak:**

1. Definiraj listu `brojevi = [4, 7, 1, 9, 3]`.
2. Ispiši: duljinu liste, zbroj svih brojeva, najveći i najmanji broj.
3. Napravi novu listu `sortirani` koja je sortirana verzija liste `brojevi`.
4. Ispiši listu u obrnutom redoslijedu pomoću `reversed()`.
5. Provjeri koristeći `all()` i `any()` je li svi ili barem jedan element različit od 0.
6. Koristeći `range()` i `list()` napravi listu `[10, 20, 30, 40, 50]`.

````python
brojevi = [4, 7, 1, 9, 3]
print("Duljina liste:", len(brojevi))
print("Zbroj svih brojeva", sum(brojevi))
print("Najmanji broj:", min(brojevi))
print("Najveći broj:", max(brojevi))
sortirani = sorted(brojevi)
print("Sortirano:", sortirani)
print("Svi brojevi različiti od 0:", all(brojevi))
print("Barem jedan broj različit od 0:", any(brojevi))
print("Lista iz range:", list(range(10, 51, 10)))


````

#### Metode za dodavanje elemenata

Metoda je **funkcija koja je vezana uz objekt**. Kada se metoda poziva na stringu, ona može vratiti novu listu ili neku vrijednost, a može i promijeniti izvornu listu. Sintaksa je:

```python
objekt.metoda(argumenti)
```

Liste u Pythonu su **promjenjive (mutable)**, što znači da im možemo **dodavati nove elemente**.
Python nudi nekoliko praktičnih metoda za dodavanje elemenata na različite načine.

**Pregled metoda za dodavanje**

| Metoda               | Opis                                                       | Primjer                               | Rezultat        |
| -------------------- | ---------------------------------------------------------- | ------------------------------------- | --------------- |
| `append(x)`          | Dodaje element `x` na kraj liste                           | `lista = [1,2,3]; lista.append(4)`    | `[1, 2, 3, 4]`  |
| `insert(i, x)`       | Umeće element `x` na poziciju `i`                          | `lista = [1,2,3]; lista.insert(1,10)` | `[1, 10, 2, 3]` |
| `extend(iterabilno)` | Dodaje sve elemente iz druge liste ili iterabilnog objekta | `lista = [1,2]; lista.extend([3,4])`  | `[1, 2, 3, 4]`  |

>  **Napomena**
> * `append()` dodaje **jedan element**, čak i ako je to lista (ne spaja je).
> * `insert()` može umetnuti element bilo gdje u listu.
> * `extend()` **proširuje listu** element po element iz druge iterabilne kolekcije.

✏️ **Zadatak:**

1. Definiraj listu `boje = ["crvena", "zelena"]`.
2. Dodaj `"plava"` na kraj liste pomoću `append()`.
3. Umetni `"žuta"` na početak liste pomoću `insert()`.
4. Dodaj još boja iz liste `["ljubičasta", "narančasta"]` pomoću `extend()`.
5. Ispiši listu nakon svake promjene.

````python
boje = ["crvena", "zelena"]
print(boje)
boje.append("plava")
print(boje)
boje.insert(0, "žuta")
print(boje)
boje.extend(["bijela", "crna"])
print(boje)
````

#### Metode za uklanjanje elemenata

Liste u Pythonu su **promjenjive (mutable)**, pa možemo uklanjati elemente pomoću različitih metoda.

**Pregled metoda za uklanjanje**

| Metoda      | Opis                                                                | Primjer                              | Rezultat                |
| ----------- | ------------------------------------------------------------------- | ------------------------------------ | ----------------------- |
| `remove(x)` | Uklanja **prvo pojavljivanje** elementa `x` iz liste                | `lista = [1,2,3,2]; lista.remove(2)` | `[1, 3, 2]`             |
| `pop(i=-1)`    | Uklanja i vraća element s pozicije `i` (zadnji ako se izostavi `i`) | `lista = [10,20,30]; lista.pop(1)`   | `[10, 30]` (vraća `20`) |
| `clear()`   | Briše **sve elemente** iz liste                                     | `lista = [1,2,3]; lista.clear()`     | `[]`                    |


>  **Napomena**
> * `remove(x)` uklanja **samo prvo pojavljivanje** traženog elementa.
> * `pop(i)` vraća uklonjeni element, što može biti korisno za daljnju obradu.
> * `clear()` potpuno prazni listu, ali **ne briše samu varijablu**.


✏️ **Zadatak:**

1. Definiraj listu `brojevi = [1, 2, 3, 2, 4]`.
2. Ukloni prvo pojavljivanje broja `2` pomoću `remove()`.
3. Ukloni zadnji element liste pomoću `pop()`.
4. Ispiši listu nakon svake promjene.
5. Prazni listu pomoću `clear()` i provjeri njen sadržaj.

````python
brojevi = [1, 2, 3, 2, 4]
brojevi.remove(2)
print(brojevi)
brojevi.pop()   # brojevi.pop(-1)
print(brojevi)
brojevi.clear()
print(brojevi)
````

#### Metode za pretraživanje i brojenje

Liste omogućuju pretraživanje i brojanje elemenata pomoću nekoliko praktičnih metoda.

**Pregled metoda za pretraživanje i brojanje**

| Metoda     | Opis                                                | Primjer                                  | Rezultat |

#### Metode za sortiranje i obrtanje

Liste se mogu **sortirati** i **obrnuti** pomoću ugrađenih metoda.
Važno je znati da obje metode **mijenjaju originalnu listu** i **ne vraćaju novu**.


| Metoda            | Opis                                                                   | Primjer                                | Rezultat    |
| ----------------- | ---------------------------------------------------------------------- | -------------------------------------- | ----------- |
| `lista.sort()`    | Sortira elemente liste **na mjestu** (uzlazno po zadanom redoslijedu). | `brojevi = [3,1,2]; brojevi.sort()`    | `[1, 2, 3]` |
| `lista.reverse()` | Obrće redoslijed elemenata u listi.                                    | `brojevi = [1,2,3]; brojevi.reverse()` | `[3, 2, 1]` |

>  **Napomena**
>
> * Metode `sort()` i `reverse()` **ne vraćaju novu listu**, već mijenjaju postojeću.
> * Ako želiš zadržati originalni redoslijed, napravi kopiju liste prije poziva metode.

**Primjeri:**

```python
brojevi = [4, 1, 7, 3, 9, 2]
brojevi.sort()
print(brojevi)   # [1, 2, 3, 4, 7, 9]

brojevi.reverse()
print(brojevi)   # [9, 7, 4, 3, 2, 1]
```

✏️ **Zadatak:**

1. Definiraj listu `brojevi = [8, 2, 5, 1, 9, 4]`.
2. Sortiraj listu pomoću `sort()`.
3. Obrni redoslijed elemenata pomoću `reverse()`.
4. Provjeri što se događa ako pozoveš `reverse()` dva puta zaredom.

````python
brojevi = [8, 2, 5, 1, 9, 4]
print(brojevi)
brojevi.sort()
print(brojevi)
brojevi.reverse()
print(brojevi)
brojevi.reverse()
brojevi.reverse()
print(brojevi)
````

#### Metoda za kopiranje

Ponekad želimo napraviti **novu listu** koja je **kopija postojeće**, kako bi mogli mijenjati novu listu bez utjecaja na original.
Za to se koristi metoda **`copy()`**.

| Metoda         | Opis                               | Primjer                                        | Rezultat    |
| -------------- | ---------------------------------- | ---------------------------------------------- | ----------- |
| `lista.copy()` | Vraća **plitku kopiju** liste. | `original = [1,2,3]; kopija = original.copy()` | `[1, 2, 3]` |


>  **Napomena**
>
> * Metoda `copy()` pravi **plitku kopiju (shallow copy)**, što znači da **unutarnji objekti** unutar liste nisu kopirani.
> * Ako lista sadrži druge liste ili promjenjive objekte, promjene u njima utječu i na kopiju.

**Primjer plitke kopije:**

```python
original = [1, 2, [3, 4]]
kopija = original.copy()

kopija[0] = 100
kopija[2][0] = 300

print(original)  # [1, 2, [300, 4]]
print(kopija)    # [100, 2, [300, 4]]
```

>  **Napomena**
>
> Osim metode `copy()`,  **plitka kopija** liste se može napraviti i rezanjem liste od početka do kraja, odnosno `lista[:]`


✏️ **Zadatak:**

1. Definiraj listu `brojevi = [5, 10, 15, 20]`.
2. Napravi kopiju liste pomoću metode `copy()`.
3. Promijeni prvi element kopije i provjeri da li se promjena vidi u originalu.

````python
brojevi = [5, 10, 15, 20]
kopija = brojevi.copy()
kopija[0] = "pet"
print(brojevi)
print(kopija)
````

#### Skraćeni oblik liste (list comprehension)

Python omogućuje **brzo stvaranje lista** pomoću **skraćenog oblika**, poznatog kao **list comprehension**.
Ovaj način je **kraći i čitljiviji** od klasičnih petlji `for` s `append()` metodom.

**Sintaksa**

```python
[izraz for element in iterator if uvjet]
```

* `izraz` – vrijednost koja ide u novu listu
* `element` – trenutni element iz iterabilnog objekta (npr. lista, string, `range()`)
* `iterator` – iterabilni objekt
* `if uvjet` – **neobavezni** filter koji uključuje samo elemente koji zadovoljavaju uvjet


**Primjeri**

1. Napravi listu kvadrata brojeva od 0 do 5:

```python
kvadrati = [x**2 for x in range(6)]
print(kvadrati)  # [0, 1, 4, 9, 16, 25]
```

2. Lista parnih brojeva od 0 do 10:

```python
parni = [x for x in range(11) if x % 2 == 0]
print(parni)  # [0, 2, 4, 6, 8, 10]
```

3. Pretvori sve riječi u listi u velika slova:

```python
rijeci = ["python", "je", "super"]
velika = [r.upper() for r in rijeci]
print(velika)  # ['PYTHON', 'JE', 'SUPER']
```

>  **Napomena**
>
> * Skraćeni oblik liste može uključivati **uvjete** i **izraz za transformaciju**.
> * Ako se koristi za složene operacije, može postati teže čitljiv.
> * Pogodno je za jednostavne i srednje složene transformacije lista.

✏️ **Zadatak:**

1. Napravi listu `brojevi` od 0 do 20.
2. Napravi novu listu koja sadrži **sve neparne brojeve** iz `brojevi`.
3. Napravi listu kvadrata tih neparnih brojeva.
4. Napravi listu `voca` iz `['jabuka', 'limun', 'banana', 'kokos']` u **velikim slovima** koja sadrži ona voća čiji naziv sadrži slovo `"a"`.

````python
brojevi = [br for br in range(20)]
print(brojevi)
neparni = [br for br in brojevi if br % 2 == 1]
print(neparni)
kvadrati = [br ** 2 for br in neparni]
print(kvadrati)
voca = ['jabuka','limun','banana','kokos']
velika = [v.upper() for v in voca if "a" in v]
print(velika)
````

#### Sortiranje i obrtanje

Python omogućuje jednostavno sortiranje elemenata i promjenu redoslijeda u listi pomoću ugrađenih metoda `sort()` i `reverse()`, te ugrađenih funkcije `sorted()` i `reversed()`.

Metoda **`sort()`** koristi se nad samom listom i **mijenja originalnu listu**.
Nasuprot tome, funkcija **`sorted()`** vraća **novu sortiranu listu**, a original ostaje nepromijenjen.

```python
brojevi = [5, 2, 9, 1, 7]
brojevi.sort()                # mijenja original
print(brojevi)                # [1, 2, 5, 7, 9]

nove = sorted(brojevi)        # vraća novu listu
print(nove)                   # [1, 2, 5, 7, 9]
print(brojevi is nove)        # False  (različiti objekti)
```

Ako želimo **silazni redoslijed**, u obje verzije možemo dodati argument `reverse=True`.

```python
brojevi.sort(reverse=True)
print(brojevi)  # [9, 7, 5, 2, 1]
```

Redoslijed elemenata možemo jednostavno **okrenuti** metodom **`reverse()`**, koja samo obrće poredak bez sortiranja:

```python
boje = ["crvena", "zelena", "plava"]
boje.reverse()
print(boje)  # ['plava', 'zelena', 'crvena']
```

Ako želimo **dobiti novi redoslijed**, ali **ne mijenjati originalnu listu**, koristimo funkciju **`reversed()`**, koja vraća iterator:

```python
boje = ["crvena", "zelena", "plava"]
obrnute = list(reversed(boje))
print(obrnute)  # ['plava', 'zelena', 'crvena']
print(boje)     # ['crvena', 'zelena', 'plava'] (nepromijenjena)
```

>  **Napomena**
>
> Sortiranje radi samo ako su elementi međusobno usporedivi (npr. svi brojevi ili svi stringovi).  
> Metoda `sort()` i funkcija `sorted()` imaju dva važna argumenta:
>
> * **`reverse`** – ako je `True`, sortira silazno
> * **`key`** – funkcija koja određuje po kojem kriteriju se elementi uspoređuju

Primjer korištenja ugrađene funkcije `len` kao ključa:

```python
rijeci = ["AI", "programiranje", "kod", "python"]
rijeci.sort(key=len)
print(rijeci)  # ['AI', 'kod', 'python', 'programiranje']
```

Možemo definirati i **vlastitu funkciju** koja vraća kriterij sortiranja.
Na primjer, sortirajmo brojeve prema **zadnjoj znamenki**:

```python
def zadnja_znamenka(broj):
    return broj % 10

brojevi = [42, 17, 93, 58, 21]
brojevi.sort(key=zadnja_znamenka)
print(brojevi)  # [21, 42, 93, 17, 58]
```

✏️ **Zadatak:**

1. Definiraj listu `brojevi = [42, 7, 19, 3, 25, 11]`.
2. Sortiraj ju **uzlazno**, zatim **silazno**.
3. Napravi novu sortiranu listu pomoću `sorted()` bez mijenjanja originala.
4. Definiraj listu `rijeci = ["sunce", "snijeg", "oblak", "kiša", "vjetar"]` i:
   * sortiraj ju po duljini (`key=len`),
   * zatim obrni redoslijed pomoću `reversed()`.
5. Napiši funkciju `zadnji_znak(rijec)` koja vraća zadnji znak riječi i koristi ju kao `key` za sortiranje.

````python
brojevi = [42, 7, 19, 3, 25, 11]
print("brojevi:", brojevi)
brojevi.sort()
print("uzlazno:", brojevi)
brojevi.sort(reverse=True)
print("silazno:", brojevi)

rijeci = ["sunce", "snijeg", "oblak", "kiša", "vjetar"]
print("riječi:", rijeci)
rijeci.sort(key=len)
print("po duljini:", rijeci)
rijeci = list(reversed(rijeci))
print("obrnuto:", rijeci)

def zadnji_znak(rijec):
    return rijec[-1]

rijeci = sorted(rijeci, key=zadnji_znak)
print("po zadnjem znaku:", rijeci)

````
