# 1. Varijable

Varijable služe za referenciranje vrijednosti pohranjene u memoriji. One su osnovni način na koji programi pamte i koriste podatke tijekom izvođenja. Referenciranjem se uspostavlja veza između imena varijable i vrijednosti. Ponekad vrijednosti nije potrebno referencirati, pa se u tu svrhu koriste anonimne varijable.

## 1.1. Definiranje varijable

Varijabla je ime koje koristimo za spremanje vrijednosti u programu. Omogućuje nam da čuvamo podatke i kasnije ih koristimo. 

Varijabla se definira tako da imenu varijable dodijelimo vrijednost pomoću operatora pridruživanja `=`.

**Primjeri:**
```python
x = 5        # int
pi = 3.14    # float
ime = "Ana"  # string 
```

Ime varijable mora:
* početi slovom ili donjom crtom `_`
* sastojati se od slova, brojeva i donjih crta
* ne smije biti ključna riječ (`if`, `for`, `def`, ...)

Python je **dinamički tipiziran jezik**:
* Tip varijable se automatski određuje prema dodijeljenoj vrijednosti
* Varijabli možemo pridružiti vrijednost različitog tipa u različito vrijeme

**Primjeri:**
```python
x = 10     # int
x = 3.14   # float
x = "Ana"  # string 
```

✏️ **Zadatak**

1. Definiraj varijablu `ime` i dodijeli joj neko ime.
2. Definiraj varijablu `godine` i dodijeli joj broj godina.
3. Ispiši obje varijable koristeći `print()`.
4. Promijeni vrijednost varijable `godine` u string, npr. `dvadeset sedam`. 
5. Ponovno ispiši obje varijable koristeći `print()`.

````python
ime = "Branko"
godine = 50
print(ime, godine)

godine = "pedeset"
print(ime, godine)
````

## 1.2. Provjera tipa varijable: `type()`

U Pythonu je moguće provjeriti **koji tip podataka** varijabla sadrži pomoću ugrađene funkcije `type()`.

**Primjeri:**

```python
x = 5
print(type(x))    # <class 'int'>

pi = 3.14159
print(type(pi))   # <class 'float'>

ime = "Ana"
print(type(ime))  # <class 'str'>

f = print
print(type(f))    # <class 'builtin_function_or_method'>

# Pozivanje funkcije preko varijable
f("Pozdrav!")       # Pozdrav!
```

> 💡 **Napomena**
>
> * `type()` vraća **tip objekta**.
> * Tip varijable se automatski određuje pri dodjeli vrijednosti.
> * Varijabla može spremiti funkciju ili bilo koji drugi objekt.

✏️ **Zadatak**

1. Definiraj varijable: `a = 5`, `b = 3.5`, `c = "Python"`, `f=print`.
2. Koristeći varijablu `f` ispiši vrijednost i tip svake varijable.
3. Dodijeli varijabli `a` novu vrijednost `"Hello"` i ispiši njenju vrijednost i tip.

````python
a = 5
b = 3.5
c = "Python"
f = print

f(a, type(a))
f(b, type(b))
f(c, type(c))
f(f, type(f))

a = "Hello"
print(a, type(a))
````

# 2. Tipovi podataka

Svaki podatak ima određeni tip koji određuje što se s njim može raditi i kako se pohranjuje u memoriji. Podatke možemo podijeliti na *primitivne* i *kolekcijske*. 

Primitivni tipovi uključuju brojeve (`int`, `float`), logičke vrijednosti (`bool`) i praznu vrijednost (`None`).

Kolekcijski tipovi obuhvaćaju liste (`list`), n-torke (`tuple`), rječnike (`dict`) i skupove (`set`, `frozenset`), koji mogu sadržavati više vrijednosti i omogućuju njihovo organizirano čuvanje i manipulaciju.

Tekstualni tip podataka (`string`) se često kategorizira kao primitivni tip podatka, iako je zapravo kolekcijski tip.

## 2.1. Brojčani tipovi podataka: `int` i  `float`

Python razlikuje cijele brojeve (`int`) i decimalne brojeve (`float`).

---

#### Definiranje

* `int` predstavlja cijele brojeve bez decimalnog dijela, npr. `5`, `-10`.
* `float` predstavlja brojeve s decimalnim dijelom, npr. `3.14`, `-0.001`.
  Float brojevi mogu se zapisivati i u znanstvenom obliku, npr. `1.5e2` (što je 150.0).

Python automatski prepoznaje tip prilikom dodjele vrijednosti, a matematičke operacije između `int` i `float` rezultiraju `float` tipom.

**Primjeri:**

* `a = 7` → `int`
* `b = 2.5` → `float`
* `c = 1.2e3` → `float` (znanstveni zapis, 1200.0)
* `d = a + b` → rezultat je `float`

---

#### Aritmetički operatori

Python podržava osnovne aritmetičke operacije nad brojčanim tipovima (`int` i `float`). Rezultat ovisi o tipovima operanada, a u operacijama između `int` i `float` rezultat je `float`.

| Operator | Opis                   | Primjer  | Rezultat |
| -------- | ---------------------- | -------- | -------- |
| `+`      | Zbrajanje              | `3 + 2`  | `5`      |
| `-`      | Oduzimanje             | `5 - 1`  | `4`      |
| `*`      | Množenje               | `4 * 2`  | `8`      |
| `/`      | Dijeljenje (float)     | `7 / 2`  | `3.5`    |
| `//`     | Cjelobrojno dijeljenje | `7 // 2` | `3`      |
| `%`      | Modulo (ostatak)       | `7 % 2`  | `1`      |
| `**`     | Potenciranje           | `3 ** 2` | `9`      |

> 💡 **Napomena:**
> 
> * Operacije s barem jednim operandom tipa `float` vraćaju `float`.
> * Operator `/` uvijek vraća `float`, čak i ako su oba operanda `int`.
> * Operator `//` vraća rezultat zaokružen prema dolje u cijelom broju (int ili float, ovisno o operandima). Ako je bilo koji operand `a` ili `b` tipa `float` onda vrijedi $a \mathbin{//} b = \lfloor \frac{a}{b} \rfloor$
> * Modulo `%` daje ostatak pri cjelobrojnom dijeljenju.Ako je bilo koji operand `a` ili `b` tipa `float` onda vrijedi $a \bmod b = a - b \times \left\lfloor \frac{a}{b} \right\rfloor$
> * Potenciranje `**` izračunava broj na zadanu potenciju.

✏️ **Zadatak:**

2. Izračunaj i ispiši njihov zbroj, razliku, umnožak i količnik.

````python
x = 3
y = 4.5

print(x + y)
print(x - y)
print(x * y)
print(x / y)
````

#### Operatori pridruživanja

Operatori pridruživanja omogućuju **kombinaciju aritmetičke operacije i dodjele** u jednom izrazu. Oni skraćuju zapis i čine kod preglednijim.


| Operator | Značenje                               | Primjer   | Rezultat   |
| -------- | -------------------------------------- | ----------| ---------- |
| `+=`     | Zbrajanje i pridruživanje              | `x += 5`  | $x = x + 5$  |
| `-=`     | Oduzimanje i pridruživanje             | `x -= 3`  | $x = x - 3$  |
| `*=`     | Množenje i pridruživanje               | `x *= 2`  | $x = x × 2$  |
| `/=`     | Dijeljenje i pridruživanje             | `x /= 4`  | $x = x ÷ 4$  |
| `//=`    | Cjelobrojno dijeljenje i pridruživanje | `x //= 3` | $x = x\ \text{div}\ 3$ |
| `%=`     | Modulo i pridruživanje                 | `x %= 3`  | $x = x\ \text{mod}\ 3$  |
| `**=`    | Potenciranje i pridruživanje           | `x **= 2` | $x = x^2$ |

> 💡 **Napomena**
>
> * Operator pridruživanja izvršava **aritmetičku operaciju i odmah dodjeljuje rezultat** varijabli.
> * Rezultat može biti `int` ili `float`, ovisno o operaciji (`/` uvijek vraća float).

````python
a = 7
print(a)
a += 5
print(a)
a *= 3
print(a)
a -= 4
print(a)
a /= 2
print(a)
a %= 5
print(a)
````

#### Beskonačan broj i ne-broj

U Pythonu možemo raditi s posebnim vrijednostima koje predstavljaju **pozitivnu i negativnu beskonačnost** (`inf` i `-inf`).

```python
poz_inf = float('inf')   # pozitivna beskonačnost
neg_inf = float('-inf')  # negativna beskonačnost
nan = float('nan')  # ne-broj
```

* `inf` i `-inf` su tipa `float`.
* `inf` se smatra većim od svih konačnih brojeva.
* `-inf` je manji od svih konačnih brojeva.
* Zbrajanje ili oduzimanje beskonačnosti ovisi o znakovima (npr. `inf - inf` daje `nan`).
* Dijeljenje beskonačnosti ima posebna pravila:
  * `1 / inf` → `0.0`
  * `1 / -inf` → `-0.0`
  * `inf / inf` → `nan`
* NaN (`nan`) je posebna `float` vrijednost koja označava neodređeni ili nevaljani brojčani rezultat. Definira se sa `float("nan")`.

## 2.2. Logički tip podataka: `bool`

Logički tip podataka `bool` je primitivni tip podataka s vrijednostima `True` i `False`. Logički tip je podtip cijelog broja (`int`), što znači da se `bool` ponaša kao `int` u mnogim situacijama.

---

#### Definiranje

* `bool` predstavlja istinosnu vrijednost `True` ili `False`

**Primjeri:**

* `a = True`
* `b = False`

---

#### Logički operatori

Python podržava osnovne logičke operatore za rad s vrijednostima tipa `bool`. Oni omogućuju spajanje i negaciju logičkih izraza.

| Operator | Opis            | Primjer          | Rezultat |
| -------- | --------------- | ---------------- | -------- |
| `and`    | Logičko **i**   | `True and False` | `False`  |
| `or`     | Logičko **ili** | `True or False`  | `True`   |
| `not`    | Negacija        | `not True`       | `False`  |

* `and` vraća `True` samo ako su oba izraza `True`.
* `or` vraća `True` ako je barem jedan izraz `True`.
* `not` mijenja `True` u `False` i obrnuto.

Ovi operatori često se koriste za kombiniranje uvjeta u kontrolnim strukturama i donošenje odluka.

✏️ **Zadatak:**

1. Definiraj varijable `a` i `b` kao `True` i `False`.
2. Ispiši rezultate osnovnih logičkih operacija (`and`, `or`, `not`) između `a` i `b`.
3. Ispiši rezultate kombiniranih izraza koristeći zagrade za jasnoću.

````python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
print(not b)

print(a or (b and a))
````


---

#### Aritmetički operatori

Pošto je `bool` podtip od `int`, i nad njim se mogu vršiti aritmetičke operacije:

Logičke vrijednosti `True` i `False` su posebna vrsta cijelih brojeva `int`:
* `True` se ponaša kao `1`,
* `False` se ponaša kao `0`.

**Primjer:**

```python
print(True + True)     # 2
print(True + False)    # 1
print(True * 5)        # 5
print(False * 10)      # 0
print(True - False)    # 1
print(True / False)    # Greška: dijeljenje s nulom
```

> 💡 **Napomena**
>
> * Iako se `bool` ponaša kao `int`, u pravilu se koristi **za logičke izraze**, ne za računanje.
> * Usporedbe i logički operatori (`and`, `or`, `not`) su jasniji i preporučeni način za rad s logičkim vrijednostima.
> * Dijeljenje (`/`) s `False` nije dopušteno jer `False` predstavlja `0`, pa uzrokuje `ZeroDivisionError`.

✏️ **Zadatak:**

   * `True + False`
   * `True * 10`
   * `False + 5`
2. Pokušaj izračunati `True / False`.

````python
print(True + False)
print(True * 10)
print(False + 5)
# print(True / False)
````

#### Usporedni operatori

Usporedni (relacijski) operatori služe za **usporedbu dviju vrijednosti** i uvijek vraćaju **logički tip** — `True` ili `False`.
Koriste se kod tipova `int`, `float`, `bool` ali i kod kolekcijskih tipova.


| Operator | Značenje          | Primjer  | Rezultat |
| -------- | ----------------- | -------- | -------- |
| `==`     | jednako           | `5 == 5` | `True`   |
| `!=`     | nije jednako      | `5 != 3` | `True`   |
| `>`      | veće od           | `7 > 2`  | `True`   |
| `<`      | manje od          | `2 < 5`  | `True`   |
| `>=`     | veće ili jednako  | `5 >= 5` | `True`   |
| `<=`     | manje ili jednako | `3 <= 2` | `False`  |


**Primjeri u Pythonu:**

```python

b = 7
c = 10

print(a == b)   # False
print(a != b)   # True
print(a > b)    # True
print(a < b)    # False
print(a >= c)   # True
print(a <= b)   # False
```


> 💡 **Napomena**
>
> * Rezultat usporedbe uvijek je **bool** (`True` ili `False`).
> * Usporedbe s posebnim vrijednostima `float("inf")` i `float("-inf")` također su moguće: svaki cijeli ili decimalni broj manji je od `inf`, a veći od `-inf`.

#### Ugrađene funkcije

Neke ugrađene funkcije kod kojih je logički tip argument ili rezultat funkcije:

| Funkcija              | Opis                                            | Primjer                     | Rezultat         |

## 2.3. Prazni tip podataka: `None`

Python ima poseban prazni tip podataka koji ima jednu jedinu vrijednost `None`.  
`None` označava **„nema vrijednosti”** ili **neinicijalizirano**.

#### Definiranje

* `None` predstavlja praznu vrijednost

**Primjer:**

* `a = None`

#### Usporedni operatori

| Operator | Opis                               | Primjer         | Rezultat                   |
| -------- | ---------------------------------- | --------------- | -------------------------- |
| `==`     | Provjerava jednakost vrijednost    | `x == None`     | `True` ako je `x` `None`   |
| `!=`     | Provjerava različitost vrijednosti | `x != None`     | `True` ako `x` nije `None` |

#### Identitetski operatori

| Operator | Opis                                | Primjer         | Rezultat                   |
| -------- | ----------------------------------- | --------------- | -------------------------- |
| `is`     | Provjerava identitet objekta        | `x is None`     | `True` ako je `x` `None`   |
| `is not` | Provjerava da objekti **nisu isti** | `x is not None` | `True` ako `x` nije `None` |

> 💡 **Napomena:**
> Za `None` se **uvijek preporučuje** koristiti `is` ili `is not`, a ne `==` ili `!=`.
> * `None` je **singleton**, što znači da postoji samo jedna instanca u memoriji.
> * Zato `x is None` uvijek daje ispravan rezultat za provjeru „prazne” vrijednosti.


```python
x = None
y = 10

print(x is None)       # True
print(y is None)       # False
print(x is not None)   # False
print(y is not None)   # True
```

✏️ **Zadatak:**

1. Definiraj varijablu `a` kao `None`.
2. Ispiši rezultate izraza koji koriste `is` i `is not` za provjeru vrijednosti `a`.
3. Ispiši rezultate usporedbi brojeva `3` s `3.0` koristeći `==` i `is`.


````python
a = None
print(a is None)
print(a is not None)
print(3 == 3.0)
print(3 is 3.0)
````

# 3. Kontrola tijeka

Kontrola tijeka određuje redoslijed izvršavanja naredbi u programu.  
Omogućuje da program donosi odluke, pojavlja izvršavanje naredbe ili preusmjerava izvršavanje na temelju uvjeta.

Osnovne vrste kontrole tijeka su:
* **grananje** - omogućuje programima da izvrše različite naredbe ovisno u uvjetu
* **petlje** - omogućuju ponavljanje niza naredbi dok je uvjet zadovoljen ili za svaki element u kolekciji

## 3.1. Uvjeti

Uvjet je izraz koji se evaluira u logičku vrijednost `True` ili `False`.
Uvjeti se često koriste u grananju, petljama i drugim logičkim strukturama za donošenje odluka.

Uvjet može uključivati:
* aritmetičke operatore,
* logičke operatore: `and`, `or`, `not`,
* usporedne operatore: `==`, `!=`, `<`, `<=`, `>`, `>=`
* identitetske operatore: `is`, `not is`.

**Primjeri:**

* `x > 10`
* `a == b`
* `not c`
* `x % 2 == 0 and x > 0`

> **Napomena:** 
>

## 3.3. Grananje: `if-elif-else`

Grananje omogućuje programu da **donese odluku** i izvrši određeni blok koda **samo ako je uvjet istinit**.
Ako uvjet nije zadovoljen, moguće je isprobati druge uvjete ili izvršiti zadani blok kao zadnju opciju.

**Sintaksa**

```python
if uvjet:
    # blok koda koji se izvršava ako je uvjet True
elif drugi_uvjet:
    # blok koda koji se izvršava ako prvi uvjet nije True, a drugi jest
else:
    # blok koda koji se izvršava ako nijedan uvjet nije True
```

* `if` – provjerava prvi uvjet.
* `elif` – opcionalno, provjerava dodatne uvjete ako prethodni nisu istiniti.
* `else` – opcionalno, izvršava se ako nijedan uvjet nije zadovoljen.
* Blokovi koda **moraju biti uvučeni** (indentacija, obično 4 razmaka).



Moguće kombinacije grananja:
* Sam `if` – provjerava jedan uvjet.
* `if` + `else` – dodaje blok za slučaj da uvjet nije zadovoljen.
* `if` + `elif` – provjerava više alternativnih uvjeta.
* `if` + `elif` + `else` – provjerava više uvjeta i ima zadanu alternativu.

**Primjer:**

```python
x = 10

# Sam if
if x > 0:
    print("Broj je pozitivan")

# If s else
if x < 0:
    print("Broj je negativan")
else:
    print("Broj nije negativan")

# If s elif
if x < 0:
    print("Broj je negativan")
elif x == 0:
    print("Broj je nula")

# If s elif i else
if x < 0:
    print("Broj je negativan")
elif x == 0:
    print("Broj je nula")
else:
    print("Broj je pozitivan")
```

>  💡 **Napomena:**
>
> * U Pythonu **indentacija je obavezna** i definira pripadnost bloka koda.
> * Možemo imati **više `elif` grana**, ali samo **jedan `else`** na kraju.

✏️ **Zadatak:**

1. Definiraj varijablu `godine = 16`.
2. Napiši grananje koje ispisuje:
   * `"Maloljetnik"` ako je godina < 18
   * `"Odrasla osoba"` ako je godina >= 18
3. Dodaj `elif` granu koja provjerava ako je godina točno 18 i ispiši `"Punoljetan"`
4. Promijeni vrijednost varijable i provjeri rezultate za različite uvjete.

````python
godine = 16
if godine < 18:
    print("Maloljetnik")
elif godine == 18:
    print("Punoljetan")
else:
    print("Odrasla osoba")
````

## 3.4. Uvjetna petlja: `while-else`

`while` petlja u Pythonu koristi se za ponavljanje bloka koda sve dok je neki uvjet istinit (`True`).
Prvo se provjerava uvjet, pa ako je on istinit, izvršava se tijelo petlje. Nakon toga se uvjet ponovno provjerava i postupak se ponavlja dok uvjet ne postane neistinit (`False`).

**Sintaksa:**

```python
while uvjet:
    # tijelo petlje

```

**Napomena:**

* Ako uvjet nikada ne postane `False`, petlja postaje beskonačna.
* Često se koristi brojač koji se unutar tijela petlje mijenja kako bi petlja završila nakon određenog broja ponavljanja.

**Zadatak:**

1. Postavi početnu vrijednost varijable `x` na 1.
2. Dok je `x` manje od ili jednako 5, ispiši `x` i povećaj ga za 1.
3. Na kraju ispisa dodaj poruku da je petlja završila.



````python
x = 1

while x <= 5:
    print("x =", x)
    x += 1

print("Petlja je završila.")
````

### 3.4.1. `while` petlja s `break` i `continue`

U Pythonu, unutar `while` petlje možemo koristiti naredbe `break` i `continue` za kontrolu toka izvođenja:

* **`break`** odmah prekida izvođenje petlje – koristi se kada želimo izaći iz petlje prije nego uvjet postane `False`.
* **`continue`** preskače ostatak koda u trenutnoj iteraciji i prelazi na sljedeći krug petlje – koristi se kada želimo izbjeći dio logike za određene uvjete.

Ove naredbe omogućuju fleksibilnije upravljanje petljama, osobito kada imamo složenije uvjete.

✏️ **Zadatak:**

1. Postavi brojač `i` na 0.
2. U petlji povećavaj `i` za 1 dok je manje od 10.
3. Ako je `i` djeljivo s 3, preskoči ispis (koristi `continue`).
4. Ako `i` postane veće od 7, prekini petlju (koristi `break`).
5. Inače ispiši vrijednost `i`.


````python
i = 0

while i < 10:
    i += 1
    if i % 3 == 0:
        continue
    if i > 7:
        break
    print(i)
````

### 3.4.2. Beskonačna petlja

**Beskonačna petlja** je petlja koja se izvršava zauvijek jer nikada ne zadovolji uvjet za izlaz.
Najčešće se piše kao `while True`, a koristi se u programima koji stalno osluškuju neki događaj, čekaju unos ili ponavljaju radnju dok ih vanjski čimbenik ne prekine.

> 💡 **Napomena:**
> 
> * Beskonačna petlja može sadržavati tijelo koje privremeno ne radi ništa – tada se koristi naredba `pass` kao zamjena za budući kod.
> * Treba biti pažljiv pri korištenju beskonačnih petlji kako program ne bi "zapeo" bez načina za prekid.


✏️ **Zadatak:**

1. Napiši beskonačnu `while` petlju.
2. U tijelu petlje koristi naredbu `pass`.

````python
while True:
    pass
````

### 3.4.3. `while` petlja s provjerom uvjeta na kraju

U ovom tipu petlje koristi se `while True` petlja koja se zaustavlja samo kada se unutar petlje zadovolji određeni uvjet.
Razlika je u tome što se uvjet za prekid (`break`) provjerava **nakon** izvršavanja ostatka tijela petlje, tj. na samom kraju.

To omogućuje da se neka akcija (npr. smanjenje broja) uvijek prvo izvrši, a tek potom provjeri treba li izaći iz petlje.

> 💡 **Napomena:**
> 
> * Ovakav pristup često se koristi kada želimo barem jednom izvršiti tijelo petlje prije nego što provjerimo uvjet.
> * Pazi na redoslijed operacija da ne bi nastala beskonačna petlja.

✏️ **Zadatak:**

1. Definiraj varijablu `broj` s vrijednošću 5.
2. Pokreni beskonačnu `while True` petlju.
3. Unutar petlje ispiši trenutnu vrijednost `broj`.
4. Na kraju tijela petlje smanji vrijednost `broj` za 1.
5. Nakon smanjenja, provjeri je li `broj` manji od 0 i ako jest, izađi iz petlje pomoću `break`.


````python
broj = 5

while True:
    print("broj =", broj)
    broj -= 1
    if broj < 0:
        break
````

### 3.4.4. `while` petlja s `else` dijelom

`while` petlja u Pythonu može imati **`else`** dio koji se izvršava **samo ako petlja nije prekinuta naredbom `break`**.
To znači da se `else` blok preskače ako je petlja prekinuta prije nego što je uvjet postao netočan.

Ovakav način pisanja petlje koristan je kada želimo razlikovati situacije u kojima je petlja završila „normalno” (uvjet postao `False`) od onih kada je prekinuta „na silu” (`break`).

> 💡 **Napomena**  
> 
> * `else` se odnosi na završetak petlje, a ne na `if` uvjet unutar nje.
> * Često se koristi u pretraživanju — ako se element pronađe, koristi se `break`; ako se ne pronađe, `else` javlja da nije pronađen.

**Sintaksa:**

```python
while uvjet:
    # tijelo petlje
    if neki_uvjet:
        break
else:
    # izvršava se samo ako nije bilo break-a
```

✏️ **Zadatak:**

1. Definiraj varijablu `brojevi` kao listu `[1, 3, 5, 7, 9]`.
2. Postavi varijablu `cilj` na `5`.
3. Postavi indeksnu varijablu `i` na `0`.
4. Napiši `while` petlju koja traje dok je `i` manji od duljine liste `brojevi`.
5. Ako je element `brojevi[i]` jednak `cilj`, ispiši `"Pronađen broj!"` i prekini petlju pomoću `break`.
6. U suprotnom, povećaj `i` za 1 i nastavi.
7. U `else` dijelu ispiši `"Broj nije pronađen."`


````python
brojevi = [1, 3, 5, 7, 9]
cilj = 5
i = 0
while i < len(brojevi):
    if brojevi[i] == cilj:
        print("Pronađen broj:", cilj)
        break
    i += 1
else:
    print("Broj nije pronađen.")
````

# 4. Funkcija

Funkcija je imenovani blok koda koji izvršava određeni zadatak i može se višekratno pozivati.

## 4.1. Pozicijski argumenti

**Pozicijski argumenti** su najosnovniji tip argumenata funkcije.
Vrijednosti se dodjeljuju parametrima **redoslijedom kojim su navedeni** u definiciji funkcije.

**Sintaksa:**

```python
def ime_funkcije(parametar1, parametar2, parametar3):
    # tijelo funkcije
    print(parametar1, parametar2, parametar3)
```

Kod poziva funkcije, vrijednosti se dodjeljuju **prema redoslijedu**:

```python
ime_funkcije(vrijednost1, vrijednost2, vrijednost3)
```

**Primjer:**
```python
def predstavi_osobu(ime, godine, grad):
    print(ime, "ima" godine, "godina i živi u", grad)

predstavi_osobu("Ana", 23, "Zagreb")
# Ana ima 23 godina i živi u gradu Zagreb.

predstavi_osobu("Ivan", 30, "Split")
# Ivan ima 30 godina i živi u gradu Split.
```

> 💡 **Napomena:**
>
> * Redoslijed argumenta je bitan: `predstavi_osobu(23, "Ana", "Zagreb")` bi dao pogrešan rezultat.
> * Pozicijski argumenti se **ne mogu preskočiti** ako nemaju zadane vrijednosti.

✏️ **Zadatak:**

1. Definiraj funkciju `razlika_triju(a, b, c)` koja vraća razliku tri broja.

3. Promijeni redoslijed pozivom `razlika_triju(2, 5, 7)` i provjeri rezultat.

````python
def razlika_triju(a, b, c):
    razlika = a - b - c
    return razlika

print(razlika_triju(7, 5, 2))
print(razlika_triju(2, 5, 7))
````

## 4.2. Argumenti s zadanim vrijednostima

**Argumenti s zadanim vrijednostima** omogućuju funkciji da ima **predefinirane vrijednosti** za neke parametre.
Ako se prilikom poziva funkcije **ne navede vrijednost** za taj parametar, koristi se zadana vrijednost.

**Sintaksa:**

```python
def ime_funkcije(parametar1, parametar2=zadana_vrijednost):
    # tijelo funkcije
    print(parametar1, parametar2)
```

* Pozicijski argumenti bez zadane vrijednosti moraju biti navedeni prije onih sa zadanim vrijednostima.



```python
def pozdrav(ime, pozdrav="Bok"):

```

```python
def predstavi_osobu(ime, godine=18, grad="Zagreb"):
```

> 💡 **Napomena:**
>
> * Argumenti s zadanim vrijednostima **moraju biti na kraju** definicije funkcije.
> * Možemo kombinirati pozicijske i zadane argumente.
> * Zadane vrijednosti mogu biti bilo koji tip podatka, uključujući liste, n-torke ili čak druge funkcije.
