#### 4.7.2.1. Rekurzija nad binarnim stablom

Binarno stablo je **rekurzivna struktura podataka** u kojoj svaki čvor ima najviše dva podstabla:
* lijevo podstablo
* desno podstablo

Formalno:
* stablo je ili **prazno**
* ili **čvor** koji sadrži **vrijednost** i **dva binarna podstabla**

Binarno stablo se u Pythonu može definirati kao:
* `None` - prazno stablo
* trojka oblika `(vrijednost, lijevo, desno)` gdje su:
  * `vrijednost` podatak u broju (npr. broj, string...)
  * `lijevo` i `desno` opet binarna stabla.

**Primjer:**

```python
(1, (2, None, None), 
    (3, (4, None, 
            (5, None, None)), 
        (6, None, None)))
```

je stablo kojeg možemo vizualizirati ovako:

```
    1
   / \
  2   3
     / \
    4   6
     \
      5
```

**Elementi stabla**
* **Korijen** stabla je početni čvor iz kojeg kreće cijela struktura. U gornjem primjeru, korijen je `1`.
* **Listovi** stabla su čvorovi koji nemaju podstabla. U gornjem primjeru, to su `2`, `5` i `6`.

>
> Binarno stablo se može zapisati i pomoću rječnika gdje umjesto trojke `(vrijednost, lijevo, desno)` se može koristiti `{"vrijednost": ..., "lijevo": ..., "desno": ...}`.
>
> Primjer gornjeg stabla prikazanog pomoću rječnika:
> 
> ```python
> {"vrijednost": 1,
>  "lijevo": {"vrijednost": 2,
>             "lijevo": None,
>             "desno": None},
>  "desno": {"vrijednost": 3,
>            "lijevo": {"vrijednost": 4,
>                       "lijevo": None,
>                       "desno": {"vrijednost": 5,
>                                 "lijevo": None,
>                                 "desno": None}},
>            "desno": {"vrijednost": 6,
>                      "lijevo": None,
>                      "desno": None}}}
> ```



✏️ **Zadatak**

1. Napravi rekurzivnu funkciju koja će vratiti listu svih listova zadanog stabla opisanog pomoću trojke oblika `(vrijednost, lijevo, desno)`.
2. Provjeri nad stablom `(1, (2, None, None), (3, (4, None, (5, None, None), (6, None, None))))`.

````python
stablo = (1, (2, None, None), (3, (4, None, (5, None, None)), (6, None, None)))

def listovi(stablo):
    if stablo is None:
        return []
    vrijednost, lijevo, desno = stablo
    if lijevo is None and desno is None:
        return [vrijednost]
    return listovi(lijevo) + listovi(desno)


listovi(stablo)
````

✏️ **Zadatak**

1. Napravi rekurzivnu funkciju koja će vratiti putanju od korijena do zadanog čvora u binarnom stablu gdje su sve vrijednosti čvorova međusobno različite.
2. Provjeri putove nad stablom `(1, (2, None, None), (3, (4, None, (5, None, None), (6, None, None))))` gdje su traženi čvorovi `1`, `2`, `3`, `4`, `5`, `6` i `7`.

````python
def putanja(stablo, trazeni):
    if stablo is None:
        return []
    vrijednost, lijevo, desno = stablo
    if vrijednost == trazeni:
        return [vrijednost]
    
    put = putanja(lijevo, trazeni)
    if put:
        return [vrijednost] + put
    
    put = putanja(desno, trazeni)
    if put:
        return [vrijednost] + put
    
stablo = (1, (2, None, None), (3, (4, None, (5, None, None)), (6, None, None)))

for trazeni in [1, 2, 3, 4, 5, 6, 7]:
    print(f"Putanja do {trazeni}: {putanja(stablo, trazeni)}")

````

#### 4.7.2.2. Rekurzija nad aritmetičkim izrazom

Aritmetički izraz se može prikazati kao **rekurzivna struktura podataka** sastavljene od:
* konstanti (brojeva)
* binarnih operacija (npr. zbrajanje, množenje)

Definirat ćemo aritmetički izraz kao:
* ili **konstanta**
* ili **operacija nad dva podizraza**

### Reprezentacija u Pythonu

Aritmetički izraz možemo definirati kao:

* `int`, `float` kao konstante
* trojka oblika `(operator, lijevo, desno)` gdje su:
  * `operator` string (`"+"`, `"*"`, `"-"`, `"/"`)
  * `lijevo` i `desno` opet aritmetički izrazi

**Primjer**

```python
("+", 1, ("*", 2, 3))
```

predstavlja izraz `1 + (2 * 3)`

kojeg možemo vizualizirati kao stablo:

```
    +
   / \
  1   *
     / \
    2   3
```


**✏️ Zadatak**

1. Napravi rekurzivnu funkciju koja evaluira aritmetički izraz koji može koristiti operatore `+`, `-`, `*`, `/` i `**`.
2. Provjeri vrijednosti funkcije za izraze:
  * `9`
  * `("+", 1, 2)`
  * `("*", ("+", 1, 2), 3)`
  * `("**", 2, 3) `
  * `("/", ("*", 4, 5), 2)`



````python
def evaluiraj(izraz):
    if isinstance(izraz, (int, float)):
        return izraz
    
    operator, lijevo, desno = izraz

    if operator == "+":
        return evaluiraj(lijevo) + evaluiraj(desno)
    elif operator == "-":
        return evaluiraj(lijevo) - evaluiraj(desno)    
    elif operator == "*":
        return evaluiraj(lijevo) * evaluiraj(desno)
    elif operator == "/":
        return evaluiraj(lijevo) / evaluiraj(desno)
    elif operator == "**":
        return evaluiraj(lijevo) ** evaluiraj(desno)



for izraz in [9, 
              ("+", 1, 2), 
              ("*", ("+", 1, 2), 3), 
              ("**", 2, 3), 
              ("/", ("*", 4, 5), 2)]:
    print(f"evaluiraj({izraz}) --> {evaluiraj(izraz)}")
    

        
````

✏️ **Zadatak**

1. Ulaz rekurzivne funkcije `pretvori(izraz)` je izraz koji može biti ili broj ili trojka oblika `(operacija, a, b)` s operacijama `"+"`, `"-"`, `"*"`, `"/"` ili `"**"` i izrazima `a` i `b`. Izlaz funkcije je čitljivi string izraza s zagradama oko svake operacije. Na primjer, `pretvori(("*", ("+", 3, 5), ("-", 2, 4))` je `((3 + 5) * (2 - 4))`.
2. Provjeri vrijednosti funkcije za sljedeće izraze:
  * `9`
  * `("+", 1, 2)`
  * `("*", ("+", 1, 2), 3)`
  * `("**", 2, 3) `
  * `("/", ("*", 4, 5), 2)`

````python
def pretvori(izraz):
    if isinstance(izraz, (int, float)):
        return str(izraz)
    op, a, b = izraz
    novi_a = pretvori(a)
    novi_b = pretvori(b)

    return f"({novi_a} {op} {novi_b})"


for izraz in [9, ("+", 1, 2), 
              ("*", ("+", 1, 2), 3), 
              ("**", 2, 3), 
              ("/", ("*", 4, 5), 2)]:
    print(f"pretvori({izraz}) --> {pretvori(izraz)}")
    
````
