# 1. Varijable

Varijable služe za referenciranje vrijednosti pohranjene u memoriji.  
One su osnovni način na koji programi pamte i koriste podatke tijekom izvođenja.  
Referenciranjem se uspostavlja veza između imena varijable i vrijednosti.  
Ponekad vrijednosti nije potrebno referencirati, pa se u tu svrhu koriste anonimne varijable.

## 1.1. Definiranje varijable

Varijabla je ime za vrijednost; dodjeljuje se operaterom `=`.

Ime mora:
* početi slovom ili `_`
* sadržavati slova, brojke ili `_`
* ne biti ključna riječ

Python je **dinamički tipiziran**:
* tip se određuje pri dodjeli
* varijabla može mijenjati tip

**Primjeri:**
```python
x = 10
x = 3.14
x = "Ana"
```

✏️ **Zadatak**

1. Definiraj varijablu `ime` i dodijeli joj neko ime.
2. Definiraj varijablu `godine` i dodijeli joj broj godina.
3. Ispiši obje varijable koristeći `print()`.
4. Promijeni vrijednost varijable `godine` u string, npr. `dvadeset sedam`. 
5. Ponovno ispiši obje varijable koristeći `print()`.

````python
ime = "Branko"
godine = 50
print(ime, godine)

godine = "pedeset"
print(ime, godine)
````

## 1.2. Provjera tipa varijable: `type()`

U Pythonu je moguće provjeriti **koji tip podataka** varijabla sadrži pomoću ugrađene funkcije `type()`.

**Primjeri:**

```python
x = 5
print(type(x))    # <class 'int'>

pi = 3.14159
print(type(pi))   # <class 'float'>

ime = "Ana"
print(type(ime))  # <class 'str'>

f = print
print(type(f))    # <class 'builtin_function_or_method'>

# Pozivanje funkcije preko varijable
f("Pozdrav!")       # Pozdrav!
```

> 💡 **Napomena**
>
> * `type()` vraća **tip objekta**.
> * Tip varijable se automatski određuje pri dodjeli vrijednosti.
> * Varijabla može spremiti funkciju ili bilo koji drugi objekt.

✏️ **Zadatak**

1. Definiraj varijable: `a = 5`, `b = 3.5`, `c = "Python"`, `f=print`.
2. Koristeći varijablu `f` ispiši vrijednost i tip svake varijable.
3. Dodijeli varijabli `a` novu vrijednost `"Hello"` i ispiši njenju vrijednost i tip.

````python
a = 5
b = 3.5
c = "Python"
f = print

f(a, type(a))
f(b, type(b))
f(c, type(c))
f(f, type(f))

a = "Hello"
print(a, type(a))
````

## 2.1. Brojčani tipovi podataka: `int` i  `float`

* `int` — cijeli brojevi (`5`, `-10`)
* `float` — decimalni (`3.14`, `-0.001`, `1.5e2`)

Operacije s `int` i `float` daju `float` ako je barem jedan operand `float`.

#### Aritmetički operatori

| Operator | Opis                   | Primjer  | Rezultat |
| -------- | ---------------------- | -------- | -------- |
| `+`      | Zbrajanje              | `3 + 2`  | `5`      |

| `%`      | Modulo (ostatak)       | `7 % 2`  | `1`      |
| `**`     | Potenciranje           | `3 ** 2` | `9`      |

> 💡 **Napomena:**
>
> * `/` uvijek vraća `float`.
> * `//` vraća cijeli dio.
> * `%` vraća ostatak.

# 3. Kontrola tijeka

Kontrola tijeka određuje redoslijed izvršavanja naredbi u programu.  
Omogućuje da program donosi odluke, pojavlja izvršavanje naredbe ili preusmjerava izvršavanje na temelju uvjeta.

Osnovne vrste kontrole tijeka su:

* **grananje** - omogućuje programima da izvrše različite naredbe ovisno u uvjetu
* **petlje** - omogućuju ponavljanje niza naredbi dok je uvjet zadovoljen ili za svaki element u kolekciji

## 3.4. Uvjetna petlja: `while-else`

`while` ponavlja blok koda dok je uvjet `True`.

### 3.4.1. `while` petlja s `break` i `continue`

`break` izlazi iz petlje; `continue` preskače ostatak trenutne iteracije.

✏️ **Zadatak:**

1. Postavi brojač `i` na 0.
2. U petlji povećavaj `i` za 1 dok je manje od 10.
3. Ako je `i` djeljivo s 3, preskoči ispis (koristi `continue`).
4. Ako `i` postane veće od 7, prekini petlju (koristi `break`).

### 3.4.3. `while` petlja s provjerom uvjeta na kraju

Petlja `while True` koja se prekida s `break` nakon tijela petlje.
Omogućuje da se tijelo petlje uvijek barem jednom izvrši.

✏️ **Zadatak:**

1. Definiraj varijablu `broj` s vrijednošću 5.

### 3.4.4. `while` petlja s `else` dijelom

`while` petlja može imati **`else`** blok koji se izvršava **samo ako petlja nije prekinuta s `break`**.

Drugim riječima:

* Ako petlja završi jer je uvjet postao `False` → izvršava se `else`.
* Ako se prekine pomoću `break` → `else` se preskače.

Koristi se kada želimo razlikovati „normalan” završetak od prekida.

> **Napomena**
>
> * `else` se odnosi na završetak petlje.
> * Često se koristi kod pretraživanja elemenata.

**Sintaksa:**

```python
while uvjet:
    # tijelo petlje
    if neki_uvjet:
        break
else:
    # izvršava se ako nije bilo break-a
```

---
✏️ **Zadatak:**

1. Postavi `cilj = 5`.
2. Postavi početnu vrijednost `i = 1`.
3. Napiši `while` petlju koja traje dok je `i <= 9`.
4. Ako je `i == cilj`, ispiši `"Pronađen broj!"` i prekini petlju s `break`.
5. Inače povećaj `i` za 2.
6. U `else` dijelu ispiši `"Broj nije pronađen."`


````python
cilj = 5
i = 1

while i <= 9:
    if i == cilj:
        print("Pronađen broj!")
        break
    i += 2
else:
    print("Broj nije pronađen.")
````

# 4. Funkcija

Funkcija je imenovani blok koda koji izvršava određeni zadatak i može se višekratno pozivati.
Korištenje funkcija omogućuje organizaciju koda, smanjuje ponavljanje i povećava čitljivost programa.

* Definiraju se pomoću ključne riječi `def`.
* Mogu imati **nula ili više argumenata**.
* Mogu, ali ne moraju, **vratiti vrijednost** pomoću naredbe `return`.
* Ako funkcija ništa ne vraća, ona implicitno vraća `None`.
* Funkcija se **poziva** navođenjem njenog imena s pripadajućim argumentima u zagradama.

**Sintaksa:**

```python
def ime_funkcije(argumenti):
    # tijelo funkcije
    return rezultat
```

> 💡 **Napomena:**
> 
> Za definiranje funkcije koja ništa ne radi, može se koristiti naredba `pass` kao privremeno tijelo funkcije.
> Na primjer:
> ```python
> def nesto():
>     pass
> ```
> U ovom slučaju funkcija ne vraća ništa, odnosno `None`.

✏️ **Zadatak:**
1. Definiraj funkciju `djeljiv(a, b)` koja prima dva cijela broja:

   * Ako je `a` djeljiv s `b`, funkcija treba vratiti `1`.
   * Inače, treba vratiti `0`.
   * Ako je `b` jednak nuli, funkcija treba vratiti `-1`.

2. Pozovi funkciju s nekoliko različitih parova brojeva i ispiši rezultate.


````python
def djeljiv(a, b):
    if b == 0:
        return -1
    if a % b == 0:
        return 1
    else:
        return 0
    
print(djeljiv(10, 2))
print(djeljiv(10, 3))
print(djeljiv(10, 0))
````

## 4.3. Nazivni argumenti

Pri pozivu se argumenti mogu navoditi po imenu, redoslijed nije bitan.

```python
def func(a, b=1, c=1):
    ...

func(c=3, a=5)
```

✏️ **Zadatak:**

1. Definiraj funkciju `izracunaj_volumen(duljina, sirina=1, visina=1)` koja ispisuje sve parametre.
2. Pozovi funkciju samo s `duljina=5` koristeći nazivne argumente.
3. Pozovi funkciju s `duljina=5, sirina=4"` i provjeri rezultat.
3. Pozovi funkciju s `duljina=5, sirina=4, visina=3"` i provjeri rezultat.
4. Pozovi funkciju s argumentima u različitom redoslijedu koristeći nazivne argumente.

````python
def izracunaj_volumen(duljina, sirina=1, visina=1):
    volumen = duljina * sirina * visina
    return volumen

print(izracunaj_volumen(5))
print(izracunaj_volumen(5, 4))
print(izracunaj_volumen(5, 4, 3))
print(izracunaj_volumen(sirina=4, duljina=5, visina=3))
````
