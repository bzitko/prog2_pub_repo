## 2.7. Skup: `set`

Skup (`set`) je **neindeksirana kolekcija jedinstvenih objekata** u Pythonu.  
Elementi skupa mogu biti bilo kojeg **nepromjenjivog** tipa (brojevi, logički tip, prazni tip, stringovi, n-torke).  
Glavne karakteristike skupova su:
* **Elementi su jedinstveni** – nema ponavljanja
* **Neuređeni su** – ne možemo pristupiti elementima preko indeksa
* **Promjenjivi su** (mutable) – možemo dodavati i uklanjati elemente
* Definiraju se **vitičastim zagradama `{}`**

---

#### Definiranje skupa

Skup se zapisuje pomoću **vitičastih zagrada `{}`**, a elementi se odvajaju zarezima:

```python
boje = {"crvena", "zelena", "plava"}
brojevi = {1, 2, 3, 4, 5}
mjesovito = {1, "dva", 3.0, (4, 5)}
```

Prazan skup se ne može definirati s `{}` (to je prazan rječnik), već se koristi `set()`:

```python
prazan_skup = set()
```

> 💡 **Napomena**
> Skup ne može sadržavati **promjenjive** tipove kao što su liste ili drugi skupovi.

> 💡 **Napomena**
> Skup je neuređen i prilikom ispisa se redoslijed elemenata može mijenjati.  
> Na primjer, ako je definiran skup brojevi = `{1, 2, 3, 4, 5}` on se može ispisati, kao   
> `{1, 3, 2, 4, 5}`,   
> `{5, 2, 3, 1, 2}` itd.


✏️ **Zadatak:**

1. Definiraj skup `ocjene` koji sadrži ocjene `{2, 3, 4, 5}`.
2. Definiraj skup `imena` s vrijednostima `{"Ana", "Marko", "Ivana"}`.
3. Definiraj prazan skup `prazan`.
4. Ispiši skupove i provjeri tip svakog od njih pomoću funkcije `type()`.

````python
# Rješenje zadatka
ocjene = {2, 3, 4, 5}
imena = {"Ana", "Marko", "Ivana"}
prazan = set()

print(ocjene, type(ocjene))
print(imena, type(imena))
print(prazan, type(prazan))
````

#### Dodavanje i uklanjanje elemenata

Za razliku od n-torki, skupovi su **promjenjivi** pa im možemo dodavati i uklanjati elemente:

| Metoda | Opis |
|--------|------|
| `.add(x)` | Dodaje element `x` u skup |
| `.remove(x)` | Uklanja element `x` (baca grešku ako ne postoji) |
| `.discard(x)` | Uklanja element `x` (ne baca grešku ako ne postoji) |
| `.pop()` | Uklanja i vraća proizvoljni element |
| `.clear()` | Uklanja sve elemente iz skupa |

**Primjeri:**

```python
boje = {"crvena", "zelena"}
boje.add("plava")
print(boje)  # {'crvena', 'zelena', 'plava'}

boje.remove("zelena")
print(boje)  # {'crvena', 'plava'}

boje.discard("žuta")  # neće baciti grešku
```

✏️ **Zadatak:**

1. Definiraj skup `brojevi = {10, 20, 30}`.
2. Dodaj broj `40` u skup.
3. Ukloni broj `20` iz skupa.
4. Pokušaj ukloniti broj `50` koristeći `.remove()` i `.discard()` i promotri razliku.

````python
# Rješenje zadatka
brojevi = {10, 20, 30}
brojevi.add(40)
print("Nakon dodavanja:", brojevi)

brojevi.remove(20)
print("Nakon uklanjanja 20:", brojevi)

# Ovo će baciti grešku:
# brojevi.remove(50)

# Ovo neće baciti grešku:
brojevi.discard(50)
print("Nakon discard(50):", brojevi)
````

#### Operacije nad skupovima

Skupovi podržavaju matematičke operacije kao što su unija, presjek, razlika i simetrična razlika:

| Operacija | Operator | Metoda | Opis |
|-----------|----------|--------|------|
| Unija | `\|` | `.union()` | Svi elementi iz oba skupa |
| Presjek | `&` | `.intersection()` | Zajednički elementi |
| Razlika | `-` | `.difference()` | Elementi u prvom, ali ne u drugom |
| Simetrična razlika | `^` | `.symmetric_difference()` | Elementi koji su u jednom ili drugom, ali ne u oba |

**Primjeri:**

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)  # {1, 2, 3, 4, 5, 6}
print(a & b)  # {3, 4}
print(a - b)  # {1, 2}
print(a ^ b)  # {1, 2, 5, 6}
```

✏️ **Zadatak:**

1. Definiraj skupove `A = {1, 2, 3, 4, 5}` i `B = {4, 5, 6, 7, 8}`.
2. Izračunaj uniju, presjek, razliku i simetričnu razliku.
3. Provjeri je li `A` podskup od `B` (operator `<=`).

````python
# Rješenje zadatka
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print("Unija:", A | B)
print("Presjek:", A & B)
print("Razlika (A - B):", A - B)
print("Razlika (B - A):", B - A)
print("Simetrična razlika:", A ^ B)
print("Je li A podskup od B?", A <= B)
````

#### Relacijski operatori nad skupovima

Skupovi podržavaju **relacijske operatore** za usporedbu skupova. Ovi operatori provjeravaju odnose između skupova kao što su jednakost, podskup, nadskup itd.

| Operator | Metoda | Opis |
|----------|--------|------|
| `==`     |        | Provjerava jesu li skupovi jednaki (imaju iste elemente) |
| `!=`     |        | Provjerava jesu li skupovi različiti |
| `<`      | `.issubset()` | Pravi podskup (strogo) |
| `<=`     | `.issubset()` | Podskup (uključujući jednakost) |
| `>`      | `.issuperset()` | Pravi nadskup (strogo) |
| `>=`     | `.issuperset()` | Nadskup (uključujući jednakost) |
| `isdisjoint()` | `.isdisjoint()` | Provjerava nemaju li zajedničkih elemenata |

**Primjeri:**

```python
a = {1, 2, 3, 4}
b = {1, 2, 3}
c = {4, 3, 2, 1}
d = {5, 6, 7}

print(b < a)        # True (b je pravi podskup od a)
print(b <= a)       # True (b je podskup od a)
print(a > b)        # True (a je pravi nadskup od b)
print(a >= c)       # True (a je nadskup od c, jednaki su)
print(a == c)       # True (jednaki su)
print(a != d)       # True (različiti su)

print(a.isdisjoint(d))  # True (nemaju zajedničkih elemenata)
print(a.isdisjoint({4, 5}))  # False (imaju zajednički element 4)
```

> 💡 **Napomena:**
> * Operator `<` provjerava **pravi podskup** – prvi skup mora biti podskup drugog, ali ne smiju biti jednaki.
> * Operator `<=` provjerava podskup (dozvoljava jednakost).
> * Metoda `.isdisjoint()` vraća `True` ako skupovi nemaju niti jedan zajednički element.

✏️ **Zadatak:**

1. Definiraj skupove `A = {1, 2, 3, 4, 5}`, `B = {2, 3, 4}` i `C = {5, 4, 3, 2, 1}`.
2. Provjeri je li `B` podskup od `A` (običan i pravi).
3. Provjeri je li `A` nadskup od `C` (običan i pravi).
4. Provjeri jesu li `A` i `C` jednaki.
5. Definiraj skup `D = {6, 7, 8}` i provjeri je li disjunktan s `A`.

````python
# Rješenje zadatka
A = {1, 2, 3, 4, 5}
B = {2, 3, 4}
C = {5, 4, 3, 2, 1}
D = {6, 7, 8}

print("B <= A (podskup):", B <= A)
print("B < A (pravi podskup):", B < A)

print("A >= C (nadskup):", A >= C)
print("A > C (pravi nadskup):", A > C)

print("A == C (jednaki):", A == C)

print("A.isdisjoint(D):", A.isdisjoint(D))
print("A.isdisjoint({4, 5}):", A.isdisjoint({4, 5}))
````

#### Provjera prisutnosti elementa

Kao i kod drugih kolekcija, možemo provjeriti je li element prisutan u skupu pomoću operatora `in` ili `not in`:


> 💡 **Napomena**
> Provjera prisutnosti u skupu je **vrlo brza**, za razliku od lista.

3. Provjeri je li broj `30` prisutan u skupu.

````python
# Rješenje zadatka
brojevi = {5, 10, 15, 20, 25}

print(15 in brojevi)
print(30 in brojevi)
````

#### Ugrađene funkcije za skupove

Python nudi nekoliko ugrađenih funkcija koje rade sa skupovima:

| Funkcija | Opis |
|----------|------|
| `len(s)` | Vraća broj elemenata u skupu |
| `max(s)` | Vraća najveći element |
| `min(s)` | Vraća najmanji element |
| `sum(s)` | Zbraja sve numeričke elemente |
| `set(iterable)` | Pretvara iterabilni objekt u skup |
| `sorted(s)` | Vraća sortiranu listu elemenata skupa |

**Primjeri:**

```python
s = {10, 20, 30, 40}
print(len(s))       # 4
print(max(s))       # 40
print(min(s))       # 10
print(sum(s))       # 100

lista = [1, 2, 2, 3, 3, 3]
print(set(lista))   # {1, 2, 3} - uklanja duplikate

print(sorted(s))    # [10, 20, 30, 40]
```

✏️ **Zadatak:**

1. Definiraj skup `brojevi = {8, 3, 5, 1, 9}`.
2. Ispiši minimalni, maksimalni i prosječni element (pomoću `sum()` i `len()`).
3. Pretvori listu `[1, 2, 2, 3, 3, 3, 4, 4, 4, 4]` u skup i ispiši ga.

````python
# Rješenje zadatka
brojevi = {8, 3, 5, 1, 9}

print("Min:", min(brojevi))
print("Max:", max(brojevi))
print("Prosjek:", sum(brojevi) / len(brojevi))

lista = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
skup = set(lista)
print("Skup bez duplikata:", skup)
````

#### Skraćeni oblik skupa (set comprehension)

Python omogućuje **brzo stvaranje skupa** pomoću **skraćenog oblika**, poznatog kao **set comprehension**.
Ovaj način je **kraći i čitljiviji** od klasičnih petlji `for` s `add()` metodom.

**Sintaksa**

```python
{izraz for element in iterator if uvjet}
```

* `izraz` – vrijednost koja ide u novi skup
* `element` – trenutni element iz iterabilnog objekta (npr. lista, string, `range()`)
* `iterator` – iterabilni objekt
* `if uvjet` – **neobavezni** filter koji uključuje samo elemente koji zadovoljavaju uvjet


**Primjeri**

1. Napravi skup kvadrata brojeva od 0 do 5:

```python
kvadrati = {x**2 for x in range(6)}
print(kvadrati)  # [0, 1, 4, 9, 16, 25]
```

2. Skup parnih brojeva od 0 do 10:

```python
parni = {x for x in range(11) if x % 2 == 0}
print(parni)  # {0, 2, 4, 6, 8, 10}
```

3. Pretvori sve riječi u skupu u velika slova:

```python
rijeci = {"python", "je", "super"}
velika = {r.upper() for r in rijeci}
print(velika)  # {'PYTHON', 'JE', 'SUPER'}
```

>  **Napomena**
>
> * Skraćeni oblik skupa može uključivati **uvjete** i **izraz za transformaciju**.
> * Ako se koristi za složene operacije, može postati teže čitljiv.
> * Pogodno je za jednostavne i srednje složene transformacije skupa.

✏️ **Zadatak:**

1. Napravi skup `brojevi` od 0 do 20.
2. Napravi novi skup koji sadrži **sve parne brojeve** iz `brojevi`.
3. Napravi skup kvadrata tih neparnih brojeva.
4. Napravi skup `voca` iz `['jabuka', 'limun', 'banana', 'kokos']` u **velikim slovima** koji sadrži ona voća čiji naziv sadrži slovo `"a"`.

````python
brojevi = {br for br in range(20)}
print(brojevi)
neparni = {br for br in brojevi if br % 2 == 1}
print(neparni)
kvadrati = {br ** 2 for br in neparni}
print(kvadrati)
voca = ['jabuka','limun','banana','kokos']
velika = {v.upper() for v in voca if "a" in v}
print(velika)
````

## 2.8. Nepromjenjivi skup: `frozenset`

Pored običnog skupa (`set`), Python nudi i **`frozenset`** – nepromjenjivu verziju skupa.

```python
fs = frozenset([1, 2, 3, 4])
```

Karakteristike `frozenset`-a:
* **Nepromjenjiv** – ne možemo dodavati niti uklanjati elemente
* Može se koristiti kao **ključ u rječniku** ili **element drugog skupa**
* Podržava sve operacije nad skupovima (unija, presjek, ...)

**Primjer:**

```python
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([2, 3, 4])

print(fs1 | fs2)  # frozenset({1, 2, 3, 4})

# Skup frozenset-ova (moguće)
skup_skupova = {fs1, fs2}
```

✏️ **Zadatak:**

1. Definiraj `frozenset` s elementima `{1, 2, 3, 4, 5}`.
2. Definiraj dva frozenseta i izračunaj njihov presjek.
3. Pokušaj dodati element u taj frozenset i promotri grešku.


````python
# Rješenje zadatka
fs = frozenset([1, 2, 3, 4, 5])
print("Frozenset:", fs)

fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])
print("Presjek frozensetova:", fs1 & fs2)

# Ovo će baciti grešku:
fs.add(6)

````
