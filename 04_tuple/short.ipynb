# 2. Tipovi podataka

## 2.6. N-torka: `tuple`

N-torka (`tuple`) je **uređena kolekcija objekata**.
Elementi mogu biti različitih tipova.
Pristupa im se preko **indeksa (od 0)**.
N-torke su **nepromjenjive (immutable)**.

Definira se pomoću **`()`**, elementi se odvajaju zarezima.

---

#### Definiranje

Tuple se definira pomoću **`()`** ili samo zarezima:

```python
par = (3, 5)
boje = ("crvena", "zelena", "plava")
t = 1, 2, 3
```

> **Napomena:**
>
> Jedan element → potreban zarez `(5,)`

---

✏️ **Zadatak:**

1. Definiraj `koordinate = (10, 20)`.
2. Definiraj `student = ("Ana", 23, "Matematika")`.
3. Definiraj tuple s jednim elementom `7`.
4. Provjeri tip (`type()`).


`````python
koordinate = (10, 20)
student = ("Ana", 23, "Makematika")
broj = (7,)

print(broj, type(broj))
`````

---

#### Spajanje i ponavljanje

N-torke se mogu **spajati (`+`)** i **ponavljati (`*`)**.
Rezultat je uvijek **nova n-torka**.

**Primjeri:**

```python id="k2m8qn"
t1 = (1, 2, 3)
t2 = (4, 5)
print(t1 + t2)

t4 = ("A", "B") * 3
print(t4)
```

> **Napomena:**
>
> N-torke su nepromjenjive → `+` i `*` stvaraju novu n-torku.

---

✏️ **Zadatak:**

1. Definiraj `t1 = ("jabuka","banana")`, `t2 = ("trešnja",)`.
2. Spoji u `voće`.
3. Ponovi `voće` 2 puta.
4. Provjeri da su varijable `t1` i `t2` nepromijenjene.


`````python
t1 = ("jabuka", "banana")
t2 = ("trešnja", )

voće = t1 + t2
print(voće)
print(voće * 2)
print(t1)
print(t2)
`````

---

#### Indeksiranje i rezanje

N-torka je **indeksirana kolekcija** → pristup pomoću **indeksa** i **rezanja**.

```python
boje = ("crvena","zelena","plava")
print(boje[0])
print(boje[-1])

brojevi = (10,20,30,40,50)
print(brojevi[1:4])
print(brojevi[::2])
```

> **Napomena:**
>
> N-torke su nepromjenjive.
> Rezanje vraća novu n-torku.

---

✏️ **Zadatak:**

1. Definiraj `t = (5, 10, 15, 20, 25)`.
2. Ispiši prvi i zadnji element.
3. Ispiši elemente od 2. do 4.


`````python
t = (5, 10, 15, 20, 25)
print(t[0], t[-1])
print(t[2:4])
`````

---

#### Provjera prisutnosti elementa

Provjera pomoću **`in`** i **`not in`**.

```python
brojevi = (10, 20, 30, 40)

print(20 in brojevi)
print(50 in brojevi)
print(30 not in brojevi)
```

> **Napomena:**
>
> Provjerava se cijeli element, ne dio.

---

✏️ **Zadatak:**

1. Definiraj `boje = ("crvena","zelena","plava")`.
2. Provjeri `"zelena"` u n-torki.
3. Provjeri `"žuta"` u n-torki.


`````python
boje = ("crvena","zelena","plava")
print("zelena" in boje)
print("žuta" in boje)
`````

---

#### Ugrađene funkcije za n-torke

Najčešće funkcije:

| Funkcija      | Opis                |
| ------------- | ------------------- |
| `len(t)`      | Broj elemenata      |
| `max(t)`      | Najveći element     |
| `min(t)`      | Najmanji element    |
| `sum(t)`      | Zbroj elemenata     |
| `tuple(...)`  | Pretvara u n-torku  |
| `all(t)`      | Svi elementi `True` |
| `any(t)`      | Barem jedan `True`  |
| `sorted(t)`   | Sortirana lista     |
| `reversed(t)` | Obrnuti iterator    |

**Primjeri:**

```python
t = (10, 20, 30, 40)
print(len(t), max(t), min(t), sum(t))

print(tuple([1,2,3]))

t2 = (True, False, True)
print(all(t2), any(t2))

print(sorted(t))
print(list(reversed(t)))
```

---

✏️ **Zadatak:**

1. Definiraj `brojevi = (5, 10, 0, 15)`.
2. Provjeri `all()` i `any()`.
3. Sortiraj (`sorted()`).
4. Obrni (`reversed()`) i pretvori u n-torku.

`````python
brojevi = (5, 10, 0, 15)
print(all(brojevi))
print(any(brojevi))
print(sorted(brojevi))
print(tuple(reversed(brojevi)))
`````

---

#### Metode

N-torke imaju malo metoda jer su **nepromjenjive**:

| Metoda      | Opis                       | Primjer                   | Rezultat |
| ----------- | -------------------------- | ------------------------- | -------- |
| `.index(x)` | Indeks prvog pojavljivanja | `(10,20,30,20).index(20)` | `1`      |
| `.count(x)` | Broj pojavljivanja         | `(1,2,2,3).count(2)`      | `2`      |

> **Napomena:**
>
> `.index(x)` → greška ako element ne postoji
> `.count(x)` → vraća `0` ako ne postoji

---

✏️ **Zadatak:**

1. Definiraj `brojevi = (5, 3, 7, 3, 3, 9)`.
2. Pronađi indeks prvog `3`.
3. Izbroji pojavljivanja `3`.
4. Pronađi indeks drugog `3`.


`````python
brojevi = (5, 3, 7, 3, 3, 9)
print(brojevi.index(3))
print(brojevi.count(3))
print(brojevi.index(3, brojevi.index(3)+1))
`````

# 1. Varijable

## 1.4. Višestruko pridruživanje

Python omogućuje **istovremeno pridruživanje više vrijednosti više varijablama** u jednom izrazu.
Ovo je korisno za **kraći i pregledniji kod**, zamjenu vrijednosti ili ignoriranje nekih vrijednosti.

**Osnovno višestruko pridruživanje**

```python
a, b, c = 1, 2, 3
print(a, b, c)  # 1 2 3
```

**Dodjeljivanje iste vrijednosti**

```python
x = y = z = 0
print(x, y, z)  # 0 0 0
```

**Zamjena vrijednosti bez privremene varijable**

```python
a, b = 5, 10
a, b = b, a
print(a, b)  # 10 5
```

> 💡 **Napomena:**
>
> * Broj varijabli s lijeve strane mora odgovarati broju vrijednosti s desne strane.
> * Višestruko pridruživanje je često korisno u **petljama**, **funkcijama koje vraćaju n-torku** i **zamjeni vrijednosti**.

✏️ **Zadatak:**

1. Dodijeli vrijednosti `10`, `20` i `30` varijablama `a`, `b` i `c`.
2. Zamijeni vrijednosti varijabli `a`, `b` i `c` bez korištenja privremene varijable.
3. Dodijeli istu vrijednost `0` varijablama `i, j, k`.

`````python
a, b, c = 10, 20, 30
a, b, c = c, b, a
print(a, b, c)
i = j = k = 0
print(i, j, k)
`````

## 1.5. Anonimne varijable

Anonimna varijabla je `_` i označava **vrijednost koja nije važna**.

**Primjeri:**

```python
for _ in range(5):
    print("Pozdrav!")

x, _, z = (1, 2, 3)
print(x, z)

_ = input("Pritisni Enter...")
```

> **Napomena:**
>
> `_` je konvencija, nije ključna riječ.

---

✏️ **Zadatak:**

1. Pomoću `for` petlje `Ispiši "Python je super!" 3 puta koristeći `_`.
2. Iz stringa `"abc"` izdvoji drugi znak koristeći `_`.
3. Iz `lista = [("Ana", 25),("Marko", 30),("Lana", 22)]` napravi `imena` (zanemari godine).


`````python
for _ in range(3):
    print("Python je super!")

_, drugi, _ = "abc"
print(drugi)

lista = [("Ana", 25), ("Marko", 30), ("Lana", 22)]
imena = [ime for ime, _ in lista]
print(imena)
`````

# 4. Funkcije

## 4.4. Anonimne funkcije (`lambda` funkcije)

* **Anonimne funkcije** nemaju ime.
* Koriste se za **kratke funkcije na jednom mjestu**, često kao argumenti drugih funkcija (`sorted()`, `min()`, `max()`).

**Sintaksa:**

```python
lambda argumenti: izraz
```

* `lambda` – početak funkcije
* `argumenti` – jedan ili više, odvojeni zarezima
* `izraz` – samo **jedan izraz**, automatski vraća rezultat

---

**Primjeri:**

```python
# Zbrajanje dva broja
zbroj = lambda a, b: a + b
print(zbroj(3, 5))   # 8

# Bez argumenata
pozdrav = lambda: "Bok!"
print(pozdrav())     # Bok!

# Maksimum od dva broja
maks = lambda x, y: x if x > y else y
print(maks(7, 12))   # 12
```

> 💡 **Napomena:**
>
> * Ograničene su na **jedan izraz**.
> * Koriste se kada je definiranje `def` funkcije previše.

---

**Zadaci / Primjena:**

1. Lambda funkcija za **kvadrat broja**.
2. Lambda funkcija koja provjerava je li broj **paran**.
3. Pomoću `lambda` i `min()` pronađi string s **najmanje slova "a"**:
   `["banana", "jabuka", "limun", "trešnja"]`.
4. Sortiraj stringove iz n-torke `("banana", "jabuka", "limun", "trešnja")` po **broju samoglasnika**.


`````python
kvadrat = lambda x: x ** 2
print("Kvadrat od", 3, "je", kvadrat(3))

paran = lambda x: x % 2 == 0
print("Je li 4 paran broj?", paran(4))

print(min(["banana", "jabuka", "limun", "trešnja"], key=lambda voce: voce.count("a")))

print(sorted(("banana", "jabuka", "limun", "trešnja"), key=lambda voce: sum(1 for slovo in voce if slovo.lower() in 'aeiou')))
`````

## 4.5. Varijabilni pozicijski argumenti

- Funkcija može primiti **nepoznat broj pozicijskih argumenata** pomoću `*args`.
- `args` može biti bilo koje ime; važan je znak `*`.
- `*args` prikuplja sve dodatne pozicijske argumente u **n-torku**.

---

### Sintaksa

```python
def ime_funkcije(*args):
    for a in args:
        print(a)
````

---

### Primjer s varijabilnim argumentima

```python
def zbroj(*brojevi):
    rezultat = 0
    for b in brojevi:
        rezultat += b
    return rezultat

print(zbroj(1, 2, 3))       # 6
print(zbroj(5, 10, 15, 20)) # 50
```

---

### Kombinacija s pozicijskim argumentima

```python
def predstavi_osobu(ime, *osobine):
    print(f"Ime: {ime}")
    print("Osobine:", osobine)

predstavi_osobu("Ana", "visoka", "plava kosa", "vesele naravi")
# Ispis:
# Ime: Ana
# Osobine: ('visoka', 'plava kosa', 'vesele naravi')
```

* Pozicijski argumenti **moraju biti prvi** prije `*args`.
* `*args` se **pretvara u n-torku** i može se iterirati ili indeksirati.

---

### Kombinacija s nazivnim argumentima

```python
def predstavi_osobu(ime, *osobine, godine):
    print(f"Ime: {ime}")
    print("Osobine:", osobine)
    print("Godine:", godine)

predstavi_osobu("Ana", "visoka", "plava kosa", "vesele naravi", godine=25)
# Ispis:
# Ime: Ana
# Osobine: ('visoka', 'plava kosa', 'vesele naravi')
# Godine: 25
```

* Nazivni argumenti mogu imati i zadane vrijednosti.

---

### Zadatak

1. Napiši funkciju `mnozenje(*brojevi)` koja vraća **umnožak** svih predanih argumenata.
2. Napiši funkciju `pozdrav(*imena)` koja ispisuje `"Pozdrav, <ime>!"` za svaki argument.
3. Napiši funkciju `prosjek(ime, *ocjene)` koja ispisuje ime i **srednju vrijednost ocjena**.
4. Napiši funkciju `interval(a, *brojevi, b)` koja vraća listu brojeva iz `brojevi` između `a` i `b`.


`````python
def mnozenje(*brojevi):
    p = 1
    for b in brojevi:
        p *= b
    return p

print("Umnožak brojeva 2, 3, 4 je", mnozenje(2, 3, 4))

def pozdrav(*imena):
    for ime in imena:
        print("Pozdrav,", ime)

pozdrav("Ana", "Marko", "Ivana")

def prosjek(ime, *ocjene):
    if len(ocjene) == 0:
        print(f"{ime} nema ocjena.")
    else:
        sredina = sum(ocjene) / len(ocjene)
        print(f"{ime} ima prosječnu ocjenu: {sredina:.2f}")

prosjek("Ana", 5, 4, 3, 5, 4)   

def interval(a, *brojevi, b):
    rezultat = [x for x in brojevi if a <= x <= b]
    return rezultat

print(interval(10, 5, 15, 20, 25, b=22))
`````
