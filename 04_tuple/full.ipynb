## 2.6. N-torka: `tuple`

N-torka (`tuple`) je **indeksirana kolekcija objekata** u Pythonu, namijenjena za pohranu bilo kojih objekata.  
Elementi n-torke mogu biti bilo kojeg tipa (brojevi, stringovi, logičke vrijednosti, pa čak i druge liste i ntorke).  
Elementi u ntorci su **uređeni**, što znači da imaju svoj **indeks** (počinje od 0).    
Ntorke su **nepromjenjive** (immutable), što znači da im elemente ne možemo mijenjati nakon stvaranja.
Liste se definiraju **zagradama `()`**, a elementi se odvajaju **zarezima**.


---

#### Definiranje

N-torka (tuple) se zapisuje pomoću **običnih zagrada `()`**, a elementi se odvajaju zarezima:

```python
par = (3, 5)
boje = ("crvena", "zelena", "plava")
razni_podaci = ("Ana", 23, True, 4.56)
```

Tuple se može definirati i **bez zagrada**, samo pomoću zareza:

```python
t = 1, 2, 3
print(t)  # (1, 2, 3)
```

> 💡 **Napomena**
> Ako tuple ima **samo jedan element**, potrebno je staviti **zarez** iza njega, inače Python ne prepoznaje da je to tuple:
>
> ```python
> x = (5)      # ovo je int
> y = (5,)     # ovo je tuple s jednim elementom
> ```

#### Spajanje i ponavljanje

Slično kao i kod lista, n-torke (tuple) možemo **spajati** i **ponavljati** koristeći operatore `+` i `*`.
Budući da su n-torke **nepromjenjive**, ovi operatori **ne mijenjaju originalne n-torke**, nego stvaraju **nove**.

**Primjeri:**

```python
# Spajanje n-torki
t1 = (1, 2, 3)
t2 = (4, 5)
t3 = t1 + t2
print(t3)   # (1, 2, 3, 4, 5)

# Ponavljanje n-torke
t4 = ("A", "B") * 3
print(t4)   # ('A', 'B', 'A', 'B', 'A', 'B')
```

> 💡 **Napomena**
> Iako se čini da „dodajemo“ ili „mijenjamo“ n-torku, to zapravo nije moguće.
> Operator `+` stvara **novu n-torku** koja sadrži elemente obje, a `*` stvara **ponovljenu kopiju**.

#### Indeksiranje i rezanje

N-torka je **indeksirana kolekcija**, pa joj se elementi mogu dohvatiti pomoću **indeksa** ili **rezanjem** (slicing).

```python
boje = ("crvena", "zelena", "plava")

print(boje[0])    # 'crvena'   – prvi element
print(boje[-1])   # 'plava'    – posljednji element

brojevi = (10, 20, 30, 40, 50)
print(brojevi[1:4])   # (20, 30, 40)
print(brojevi[::2])   # (10, 30, 50)
```

> 💡 **Napomena**
> N-torke su **nepromjenjive** – ne možemo mijenjati njihove elemente (`boje[0] = "žuta"` izaziva grešku).
> Rezanje vraća **novu n-torku**, bez promjene izvorne.

#### Ugrađene funkcije za n-torke

Python n-torke podržavaju nekoliko **ugrađenih funkcija** koje olakšavaju rad s njima:

| Funkcija          | Opis                                            |
| ----------------- | ----------------------------------------------- |
| `len(t)`          | Vraća broj elemenata n-torke                    |
| `max(t)`          | Vraća najveći element                           |
| `min(t)`          | Vraća najmanji element                          |
| `sum(t)`          | Zbraja sve numeričke elemente                   |
| `tuple(iterable)` | Pretvara iterabilni objekt u n-torku            |
| `all(t)`          | Vraća `True` ako su svi elementi istiniti       |
| `any(t)`          | Vraća `True` ako je barem jedan element istinit |
| `sorted(t)`       | Vraća **sortiranu listu** elemenata n-torke     |
| `reversed(t)`     | Vraća **obrnuti iterator** preko n-torke        |

**Primjeri:**

```python
t = (10, 20, 30, 40)
print(len(t))       # 4
print(max(t))       # 40
print(min(t))       # 10
print(sum(t))       # 100

l = [1, 2, 3]
print(tuple(l))     # (1, 2, 3)

t2 = (True, False, True)
print(all(t2))      # False
print(any(t2))      # True

print(sorted(t))    # [10, 20, 30, 40]
print(list(reversed(t)))  # [40, 30, 20, 10]
```

#### Metode

N-torke imaju vrlo malo metoda jer su **nepromjenjive**.
Dvije glavne metode su:

| Metoda     | Opis                                                | Primjer                                  | Rezultat |
| ---------- | --------------------------------------------------- | ---------------------------------------- | -------- |
| `.index(x, start=0)` | Vraća **indeks prvog pojavljivanja** elementa `x` (gledajući od indeksa `start`)   | `t = (10,20,30,20); t.index(20)` | `1`      |
| `.count(x)` | Vraća koliko puta se element `x` pojavljuje u listi | `t = (1,2,2,3); t.count(2)`      | `2`      |

---

> 💡 **Napomena**
> * `.index(x)` će baciti **ValueError** ako element nije prisutan u listi.
> * `.count(x)` vraća `0` ako element **nije prisutan**, ali ne izaziva grešku.

# 1. Varijable

## 1.3. Anonimne varijable

U Pythonu se ponekad koristi **anonimna varijabla**, odnosno varijabla kojoj je ime **donja crta `_`**.
Ona označava **vrijednost koja nije važna** ili **koju nećemo dalje koristiti**.
Takva varijabla se koristi samo radi **čitljivosti koda**, da pokaže da nam rezultat nije bitan.

**Primjeri:**

```python
# Kada nas ne zanima vrijednost iz petlje
for _ in range(5):
    print("Pozdrav!")   # ponavlja se 5 puta

# Kada funkcija vraća više vrijednosti, ali nam treba samo jedna
x, _, z = (1, 2, 3)
print(x, z)   # Rezultat: 1 3

# Kada želimo zanemariti unos
_ = input("Pritisni Enter za nastavak...")
```
Anonomna varijabla se može koristiti u bilo kojim indeksiranim kolekcijama (string, lista, ntorka).

> 💡 **Napomena:**
> `_` je **samo dogovorena konvencija**, nije posebna ključna riječ u Pythonu.
> Može se koristiti `_` kao bilo koje drugo ime varijable, ali to se **ne preporučuje** jer zbunjuje čitatelja koda.

## 1.4. Višestruko pridruživanje

Python omogućuje **istovremeno pridruživanje više vrijednosti više varijablama** u jednom izrazu.
Ovo je korisno za **kraći i pregledniji kod**, zamjenu vrijednosti ili ignoriranje nekih vrijednosti.

**Osnovno višestruko pridruživanje**


**Dodjeljivanje iste vrijednosti**

```python
x = y = z = 0
print(x, y, z)  # 0 0 0
```

**Zamjena vrijednosti bez privremene varijable**

```python

a, b = b, a

```

> 💡 **Napomena:**
>
> * Broj varijabli s lijeve strane mora odgovarati broju vrijednosti s desne strane.

## 1.5. Pakiranje i raspakiravanje varijabli

Python omogućuje **spremanje više vrijednosti u jednu varijablu** (pakiranje) i **raspoređivanje tih vrijednosti u više varijabli** (raspakiravanje).
Ovo je korisno za **kraći, pregledniji i fleksibilniji kod** i radi sa **n-torkama, listama, stringovima i range objektima**.

---

**Pakiranje**

**Tuple** se može stvoriti jednostavno stavljanjem više vrijednosti odvojenih zarezom:

```python
t = 1, 2, 3       # n-torka s tri vrijednosti
l = [4, 5, 6]     # lista s tri vrijednosti
r = range(7, 10)  # range objekti
s = "Ana"         # string
```

* Sve ove strukture se mogu **raspakirati** u varijable.

---

**Raspakiravanje**

Vrijednosti iz n-torke, liste, stringa ili range objekta možemo **rasporediti u više varijabli**:

```python
# Tuple
a, b, c = t
print(a, b, c)  # 1 2 3

# Lista
x, y, z = l
print(x, y, z)  # 4 5 6

# String
p, q, r = s
print(p, q, r)  # A n a

# Range
i, j, k = r
print(i, j, k)  # 7 8 9
```

Broj varijabli s lijeve strane **mora odgovarati broju elemenata** ili se mogu koristiti **anonimne varijable `_`**:

```python
a, _, c = t
print(a, c)  # 1 3
```

---

**Raspakiravanje s `*`**

Možemo dohvatiti **preostale elemente** koristeći `*`:

```python
a, *ostatak, b = [1, 2, 3, 4, 5]
print(a, ostatak, b)  # 1 [2, 3, 4] 5

p, *sredina, q = range(10, 15)
print(p, sredina, q)  # 10 [11, 12, 13, 14] 14
```

Možemo kombinirati * i _ da ignoriramo elemente koje ne želimo

```python
l = [10, 20, 30, 40, 50]

_, *sredina, _ = l
print(sredina)  # [20, 30, 40]
```
Prvi i zadnji element se preskaču pomoću `_`, a srednji se sprema u listu `sredina`.

```python
l = [10, 20, 30, 40, 50]

prvi, *_, zadnji = l
print(prvi, zadnji)  # 10, 50
```

Svi elementi osim prvog i zadnjeg se preskači pomoću `*_`.


> 💡 **Napomena:**
>
> * Pakiranje i raspakiravanje radi s **iterabilnim objektima**: n-torka, lista, string, range.
> * Funkcije koje vraćaju više vrijednosti zapravo vraćaju **n-torku**, koja se zatim može raspakirati u varijable.
> * Anonimna varijabla `_` može se kombinirati s `*` kako bi se ignorirale neželjene vrijednosti.

# 4. Funkcije

## 4.4. Anonimne funkcije (lambda funkcije)

U Pythonu se anonimne funkcije definiraju pomoću ključne riječi `lambda`.
Zovu se **anonimne** jer nemaju ime — često se koriste kada nam treba **kratka funkcija na jednom mjestu**, obično kao **argument druge funkcije** (npr. `sorted()`, `min()`, `max()`).

**Sintaksa:**

```python
lambda argumenti: izraz
```

* `lambda` označava početak funkcije
* navode se **argumenti** (mogu biti više njih, odvojeni zarezima)
* nakon dvotočke slijedi **izraz** (ne može sadržavati više naredbi)
* rezultat izraza se **automatski vraća** (nema `return`)

**Primjeri:**

```python
# Lambda funkcija koja zbraja dva broja
zbroj = lambda a, b: a + b
print(zbroj(3, 5))   # Rezultat: 8

# Lambda funkcija bez argumenata
pozdrav = lambda: "Bok!"
print(pozdrav())     # Rezultat: Bok!

# Lambda funkcija koja vraća maksimum od dva broja
maks = lambda x, y: x if x > y else y
print(maks(7, 12))   # Rezultat: 12
```

> 💡 **Napomena:**
>
> * `lambda` funkcije su ograničene na **jedan izraz** – ne mogu sadržavati više naredbi.
> * Koriste se kada bi definiranje zasebne funkcije pomoću `def` bilo nepotrebno.

✏️ **Zadatak:**

1. Definiraj `lambda` funkciju koja vraća kvadrat broja.
2. Definiraj `lambda` funkciju koja provjerava je li broj paran.
3. Pomoću `lambda` funkcije i `min()` funkcije  pronađi string iz liste `["banana", "jabuka", "limun", "trešnja"]` koji sadrži najmanje slova `"a"`.

````python
kvadrat = lambda x: x ** 2
print("Kvadrat od", 3, "je", kvadrat(3))

paran = lambda x: x % 2 == 0
print("Je li 4 paran broj?", paran(4))

print(min(["banana", "jabuka", "limun", "trešnja"], key=lambda voce: voce.count("a")))

print(sorted(("banana", "jabuka", "limun", "trešnja"), key=lambda voce: sum(1 for slovo in voce if slovo.lower() in 'aeiou')))
````

## 4.5. Varijabilni pozicijski argumenti

Ponekad funkcija treba primiti **nepoznat broj pozicijskih argumenata**.
U Pythonu se za to koristi **`*args`**, gdje `args` može biti bilo koje ime (važan je samo znak `*`).

**Sintaksa**

```python
def ime_funkcije(*args):
    # args je n-torka svih proslijeđenih argumenata
    for a in args:
        print(a)
```

`*args` prikuplja **sve dodatne pozicijske argumente** u **n-torku**.

**Primjer**

```python
def zbroj(*brojevi):
    rezultat = 0
    for b in brojevi:
        rezultat += b
    return rezultat

print(zbroj(1, 2, 3))       # 6
print(zbroj(5, 10, 15, 20)) # 50
```

---

**Kombinacija s pozicijskim argumentima**

Funkcija može imati i **pozicijske argumente prije `*args`**.

```python
def predstavi_osobu(ime, *osobine):
    print(f"Ime: {ime}")
    print("Osobine:", osobine)

predstavi_osobu("Ana", "visoka", "plava kosa", "vesele naravi")
# Ispis:
# Ime: Ana
# Osobine: ('visoka', 'plava kosa', 'vesele naravi')
```

Pozicijski argumenti **moraju biti prvi** prije `*args`.  
`*args` se **pretvara u n-torku**, pa se može iterirati, indeksirati i koristiti u petljama.


> 💡 **Napomena:**
>
> * Ime `args` je konvencija; može biti bilo koje drugo.
> * `*args` je uvijek **n-torka**, iako funkcija prima varijabilan broj argumenata.

---

**Kombinacija s nazivnim argumentima**

Funkcija može imati i **nazivne argumente poslije `*args`**.

```python
def predstavi_osobu(ime, *osobine, godine):
    print(f"Ime: {ime}")
    print("Osobine:", osobine)
    print("Godine:", godine)

predstavi_osobu("Ana", "visoka", "plava kosa", "vesele naravi", godine=25)
# Ispis:
# Ime: Ana
# Osobine: ('visoka', 'plava kosa', 'vesele naravi')
# Godine: 25
```

> 💡 **Napomena:**
>
> * Nazivni argumenti mogu imati i zadane vrijednosti.


✏️ **Zadatak:**

1. Napiši funkciju `mnozenje(*brojevi)` koja vraća umnožak svih predanih argumenata.
2. Napiši funkciju `pozdrav(*imena)` koja ispisuje `"Pozdrav, <ime>!"` za svaki proslijeđeni argument.
3. Napiši funkciju `prosjek()` koja prima obavezni argument `ime` i **neograničen broj ocjena**, te ispiši ime i srednju vrijednost ocjena.
4. Napiši funkciju `interval()` koja prima obavezni argument `a`, varijabilne pozicijske argumente `brojevi` i nazivni argument `b` koji će vratiti listu brojeva iz `brojevi` koji se nalaze između `a` i `b`


````python
def mnozenje(*brojevi):
    p = 1
    for b in brojevi:
        p *= b
    return p

print("Umnožak brojeva 2, 3, 4 je", mnozenje(2, 3, 4))

def pozdrav(*imena):
    for ime in imena:
        print("Pozdrav,", ime)

pozdrav("Ana", "Marko", "Ivana")

def prosjek(ime, *ocjene):
    if len(ocjene) == 0:
        print(f"{ime} nema ocjena.")
    else:
        sredina = sum(ocjene) / len(ocjene)
        print(f"{ime} ima prosječnu ocjenu: {sredina:.2f}")

prosjek("Ana", 5, 4, 3, 5, 4)   

def interval(a, *brojevi, b):
    rezultat = [x for x in brojevi if a <= x <= b]
    return rezultat

print(interval(10, 5, 15, 20, 25, b=22))
````
