## 2.8. Rječnik: `dict`

Rječnik `dict` je neuređena asocijativna kolekcija koja pohranjuje podatke u obliku **parova ključ–vrijednost** (*key–value*). Za razliku od lista, gdje elementima pristupamo pomoću indeksa, u rječniku elementima pristupamo pomoću **ključa**.

Ključevi moraju biti **jedinstveni** i nepromjenjivog tipa (npr. `int`, `str`, `tuple`), dok vrijednosti mogu biti bilo kojeg tipa.

---

#### Definiranje rječnika

Rječnik definiramo pomoću vitičastih zagrada `{}`. Unutar njih navodimo parove **ključ–vrijednost**, gdje su ključ i vrijednost odvojeni dvotočkom `:`, a pojedini parovi zarezom `,`.

Opći oblik:

```python
naziv_rjecnika = {
    kljuc1: vrijednost1,
    kljuc2: vrijednost2
}
```

Primjer:

Primjer rječnika:

```python
osoba = {
    "ime": "Ana",
    "prezime": "Anić",
    "godine": 25
}
```

U ovom primjeru:

* `"ime"`, `"prezime"` i `"godine"` su ključevi
* `"Ana"`, `"Anić"` i `25` su pripadne vrijednosti

Moguće je definirati i prazan rječnik:

```python
prazan = {}
```

ili pomoću ugrađene funkcije:

```python
prazan = dict()
```

Ključevi moraju biti jedinstveni. Ako se isti ključ ponovi, zadržat će se samo posljednja dodijeljena vrijednost:

```python
primjer = {
    "a": 1,
    "a": 2
}

print(primjer)  # {'a': 2}
```

Vrijednosti u rječniku mogu biti različitih tipova (brojevi, stringovi, logičke vrijednosti itd.).

#### Pristupanje vrijednostima pomoću ključa

Vrijednost iz rječnika dohvaćamo pomoću **ključa**, a ne pomoću indeksa kao kod lista.

Za pristup koristimo uglate zagrade `[]` i unutar njih navodimo ključ:

```python
student = {
    "ime": "Ana",
    "prezime": "Anić",
    "godine": 20
}

print(student["ime"])      # Ana
print(student["godine"])   # 20
```

Ako pokušamo pristupiti ključu koji ne postoji, Python će javiti grešku (`KeyError`):

```python
print(student["adresa"])  # greška
```

Sigurniji način pristupa je pomoću metode `get()`, koja vraća `None` ako ključ ne postoji:

```python
print(student.get("ime"))      # Ana
print(student.get("adresa"))   # None
```

Možemo zadati i zadanu vrijednost ako ključ ne postoji:

```python
print(student.get("adresa", "Nema podatka"))
```

✏️ **Zadatak:**

1. Definiraj rječnik `osoba` s ključevima `"ime"`, `"prezime"` i `"godine"`.
2. Ispiši vrijednost ključa `"ime"` koristeći `[]`.
3. Pokušaj ispisati nepostojeći ključ koristeći `get()`.
4. Ponovno ispiši isti ključ, ali ovaj put postavi zadanu vrijednost `"Nepoznato"`.


````python
osoba = {"ime": "Ana", "prezime": "Anić", "godine": 25}
print(osoba["ime"])
print(osoba.get("zanimanje"))
print(osoba.get("zanimanje", "Nepoznato"))
````

#### Postavljanje vrijednosti

Vrijednosti u rječniku možemo **postaviti** (dodati ili promijeniti) pomoću ključa i operatora pridruživanja `=`.

Ako ključ već postoji, njegova vrijednost će se **promijeniti**. Ako ključ ne postoji, bit će **dodan novi par ključ–vrijednost**.

**Primjer**:

```python
student = {
    "ime": "Ana",
    "godine": 20
}

# promjena postojeće vrijednosti
student["godine"] = 21

# dodavanje novog elementa
student["prezime"] = "Anić"

print(student)
```

Rezultat:

```python
{'ime': 'Ana', 'godine': 21, 'prezime': 'Anić'}
```

4. Ispiši rječnik.


````python
auto = {"marka": "Toyota", "godina": 2020}
auto["godina"] = 2021
auto["boja"] = "crvena"
print(auto)
````

#### Iteracija kroz rječnik

Rječnik možemo prolaziti pomoću `for` petlje.
žZa razliku od liste, gdje prolazimo kroz elemente, kod rječnika najčešće prolazimo kroz:

* ključeve
* vrijednosti
* ili parove (ključ, vrijednost)

**Iteracija kroz ključeve**

Ako koristimo samo `for`, prolazimo kroz **ključeve**.

**Primjer:**

```python
osoba = {
    "ime": "Ana",
    "godine": 25,
    "grad": "Split"
}

for kljuc in osoba:
    print(kljuc)
```

**Iteracija kroz vrijednosti**

Za prolazak kroz vrijednosti koristimo `values()`.

**Primjer:**
```python
for vrijednost in osoba.values():
    print(vrijednost)
```

**Iteracija kroz ključ i vrijednost**

Ako želimo i ključ i vrijednost, koristimo `items()`.

**Primjer:**

```python
for kljuc, vrijednost in osoba.items():
    print(kljuc, "->", vrijednost)
```

✏️ **Zadatak:**

1. Definiraj rječnik `osoba = {"ime": "Ana", "godine": 25, "grad": "Split"}`
2. Napravi `for` petlju koja prolazi kroz rječnik pomoću `items()`
3. Ispiši ključ i vrijednost u formatu: `kljuc: vrijednost`


````python
osoba = {"ime": "Ana", "godine": 25, "grad": "Split"}
for kljuc, vrijednost in osoba.items():
    print(f"{kljuc}: {vrijednost}") 
````

#### Operatori usporedbe

Za rječnike u Python definirani su samo operatori `==` i `!=`. Oni uspoređuju jesu li svi parovi **ključeva i vrijednosti** identični u oba rječnika, **neovisno o redoslijedu unosa**.

**Primjer:**

```python
d1 = {"ime": "Ana", "godine": 25}
d2 = {"godine": 25, "ime": "Ana"}
d3 = {"ime": "Ana", "godine": 20}

print(d1 == d2)  # True
print(d1 != d3)  # True
```

Operatori `<`, `>`, `<=` i `>=` **nisu definirani nad rječnicima** i direktna primjena na `dict` objekt daje `TypeError`.

Operatori `<`, `>`, `<=` i `>=` **jesu definirani nad `keys()`, i `items()` pogledima**, ali tada se  uspoređuju kao **skupovi**.


**Primjer:**
```python
d1 = {"a": 1, "b": 2}
d2 = {"a": 1, "b": 2, "c": 3}

print(d1.keys() <= d2.keys())  # True jer {"a", "b"} <= {"a", "b", "c"}
print(d1.keys() < d2.keys())   # True jer {"a", "b"} < {"a", "b", "c"}

print(d1.items() < d2.items())  # True jer {("a", 1), ("b", 2)} <= {("a", 1), ("b", 2), ("c": 3)}
```

✏️ **Zadatak:**

1. Definiraj `d1 = {"x": 10, "y": 20}` i `d2 = {"x": 10, "y": 20, "z": 30}`.
2. Ispiši `d1 == d2`, `d1 != d2`, te `d1.keys() <= d2.keys()` i `d1.items() <= d2.items()`.

````python
d1 = {"x": 10, "y": 20}
d2 = {"x": 10, "y": 20, "z": 30}
print(d1 == d2)
print(d1 != d2)
print(d1.keys() <= d2.keys())
print(d2.items() <= d1.items())
````

#### Operator pripadnosti

Python omogućuje provjeru **pripadnosti elemenata u rječniku** pomoću operatora `in` i `not in`.

Kod rječnika, operator `in` provjerava **ključeve**, a ne vrijednosti.

| Operator | Opis                                      | Primjer        |
| -------- | ----------------------------------------- | -------------- |
| `in`     | Provjerava postoji li ključ u rječniku    | `"a" in d`     |
| `not in` | Provjerava nepostojanje ključa u rječniku | `"x" not in d` |


> **Napomena**
>
> * Operator `in` kod rječnika provjerava **ključeve (keys)**, ne vrijednosti.
> * Ako želimo provjeriti vrijednosti, moramo koristiti `values()`.
> * Ako želimo provjeriti parove, koristimo `items()`.

**Primjer:**

```python
d = {"a": 1, "b": 2}
print("a" in d)            # True (ključ postoji)
print(1 in d)              # False (vrijednosti se ne provjeravaju)
print(1 in d.values())     # True
print(("a", 1) in d.items())  # True 
```



✏️ **Zadatak:**

1. Definiraj rječnik `student = {"ime": "Ana", "godine": 20, "smjer": "Informatika"}`.
2. Provjeri:

   * postoji li ključ `"ime"` u rječniku,
   * ne postoji li ključ `"prezime"`.
3. Ako postoji ključ `"godine"`, ispiši `"Podatak o godinama postoji."`.
4. Provjeri nalazi li se vrijednost `"Ana"` u rječniku koristeći odgovarajuću metodu.


````python
student = {"ime": "Ana", "godine": 20, "smjer": "Informatika"}
print("ime" in student)
print("prezime" in student)
print("prezime" not in student)
if "godine" in student:
    print(f"Podatak o godinama postoji.")
print("Ana" in student.values())
````

#### Ugrađene funkcije za rječnike

Osim metoda rječnika (`dict`), Python nudi i **ugrađene funkcije** koje se mogu primijeniti na rječnike ili njihove poglede (`keys()`, `values()`, `items()`). Koriste se za **pristup, konverziju i analizu podataka**.

| Funkcija           | Opis                                                       | Primjer                                   |
| ------------------ | ---------------------------------------------------------- | ----------------------------------------- |
| `len(d)`           | Vraća broj parova (ključ–vrijednost) u rječniku            | `len({"a":1,"b":2}) → 2`                  |
| `str(d)`           | Pretvara rječnik u string prikaz                           | `str({"a":1})`                            |
| `list(d)`          | Vraća listu **ključeva**                                   | `list({"a":1,"b":2}) → ['a','b']`         |
| `dict(iterable)`   | Stvara ili kopira rječnik iz iterabilnog objekta           | `dict([("a",1),("b",2)]) → {"a":1,"b":2}` |
| `list(d.keys())`   | Vraća listu ključeva                                       | `['a','b']`                               |
| `list(d.values())` | Vraća listu vrijednosti                                    | `[1,2]`                                   |
| `list(d.items())`  | Vraća listu parova `(ključ, vrijednost)`                   | `[('a',1),('b',2)]`                       |
| `sorted(d)`        | Vraća sortirane ključeve                                   | `['a','b']`                               |
| `any(d)`           | `True` ako barem jedan ključ ima logičku vrijednost `True` | `any({"a": 1, 0: "b"}) → True`                                    |
| `all(d)`           | `True` ako su svi ključevi logički `True`                  | `all({"a": 1, 0: "b"}) → False`                                    |

---

> **Napomena**
>
> * `list(d)` vraća **ključeve rječnika**, ne vrijednosti.
> * `dict()` se može koristiti za:
>
>   * konverziju iz parova `(ključ, vrijednost)`
>   * stvaranje praznog rječnika `dict()`
>   * kopiranje postojećeg rječnika `dict(d)`
> * `sorted(d)` sortira po ključevima, ne po vrijednostima.

**Korištenje `dict()` funkcije:**

```python
parovi = [("a", 1), ("b", 2)]
d = dict(parovi)
print(d)
{"a": 1, "b": 2}
```

#### Metode za pristup podacima

Podacima se pristupa preko ključeva.

| Metoda / pristup | Opis                                                            | Primjer                         | Rezultat             |
| ---------------- | --------------------------------------------------------------- | ------------------------------- | -------------------- |
| `get(k)`         | Vraća vrijednost za ključ `k`, ili `None` ako ne postoji        | `d = {"a": 1}; d.get("b")`      | `None`               |
| `get(k, x)`      | Vraća vrijednost ili zadanu vrijednost `x` ako ključ `k` ne postoji | `d = {"a": 1}; d.get("b", 0)`   | `0`                  |
| `keys()`         | Vraća sve ključeve rječnika                                     | `d = {"a":1,"b":2}; d.keys()`   | `dict_keys([...])`   |
| `values()`       | Vraća sve vrijednosti rječnika                                  | `d = {"a":1,"b":2}; d.values()` | `dict_values([...])` |
| `items()`        | Vraća parove (ključ, vrijednost)                                | `d = {"a":1,"b":2}; d.items()`  | `dict_items([...])`  |


> **Napomena**
>
> * `d[k]` je najizravniji način pristupa, ali može baciti `KeyError` ako ključ ne postoji.
> * `get()` je sigurnija opcija jer omogućuje kontrolu nad slučajem nepostojećeg ključa.
> * `keys()`, `values()` i `items()` koriste se pri iteraciji kroz rječnik.

✏️ **Zadatak:**

1. Definiraj rječnik `osobe = {"Ana": 25, "Marko": 30, "Ivana": 22}`.
2. Ispiši dob osobe `"Ana"` pomoću `[]`.
3. Pokušaj dohvatiti ključ `"Petar"` pomoću `get()` i ispiši rezultat.
4. Dohvati `"Petar"` pomoću `get()` uz zadanu vrijednost `0`.
5. Ispiši sve ključeve, sve vrijednosti i sve parove rječnika.


````python
osobe = {"Ana": 25, "Marko": 30, "Ivana": 22}
print(osobe["Ana"])
print(osobe.get("Petar"))
print(osobe.get("Petar", 0))

for k in osobe:
    print(k)

for v in osobe.values():
    print(v)

for kv in osobe.items():
    print(kv)
````

#### Metode za postavljanje podataka

Rječnici omogućuju **dodavanje novih parova**, **izmjenu postojećih vrijednosti** te **sigurno postavljanje vrijednosti** pomoću ugrađenih metoda.

**Pregled metoda za postavljanje podataka**

| Metoda / način    | Opis                                                       | Primjer                              | Rezultat        |
| ----------------- | ---------------------------------------------------------- | ------------------------------------ | --------------- |
| `update({...})`   | Ažurira rječnik s više parova odjednom                     | `d = {"a":1}; d.update({"b":2})`     | `{"a":1,"b":2}` |
| `setdefault(k,v)` | Postavlja `k` na `v` samo ako ključ ne postoji             | `d = {"a":1}; d.setdefault("a", 99)` | `{"a":1}`       |
| `setdefault(k,v)` | Ako ključ ne postoji, dodaje ga                            | `d = {}; d.setdefault("a", 1)`       | `{"a":1}`       |

> **Napomena**
>
> * `d[k] = v` je najčešći i najizravniji način postavljanja vrijednosti.
> * `update()` je praktičan za dodavanje više elemenata odjednom.
> * `setdefault()` ne mijenja postojeće vrijednosti — koristi se za inicijalizaciju ključeva.

✏️ **Zadatak:**

1. Kreiraj prazan rječnik `student = {}`.
2. Dodaj ključ `"ime"` s vrijednošću `"Luka"` koristeći `[]`.
3. Dodaj ključ `"godine"` s vrijednošću `21` koristeći `update()`.
4. Pokušaj postaviti `"ime"` na `"Ivan"` pomoću `setdefault()` i provjeri rezultat.
5. Dodaj novi ključ `"smjer"` s vrijednošću `"Informatika"` pomoću `setdefault()`.
6. Ispiši rječnik nakon svake operacija.


````python
student = {}
print(student)

student["ime"] = "Luka"
print(student)

student.update({"godine": 21})
print(student)

student.setdefault("ime", "Ivan")
print(student)

student.setdefault("smjer", "Informatika")
print(student)
````

#### Metode za brisanje podataka

Rječnici omogućuju **uklanjanje pojedinih parova ključ–vrijednost** ili cijelog sadržaja rječnika.

**Pregled metoda za brisanje podataka**

| Metoda / operacija | Opis                                           | Primjer                          | Rezultat                |
| ------------------ | ---------------------------------------------- | -------------------------------- | ----------------------- |
| `pop(k)`           | Briše i vraća vrijednost za ključ `k`          | `d = {"a":1,"b":2}; d.pop("a")`  | `{"b":2}`, vraća `1`    |
| `pop(k, x)`        | Briše ključ ili vraća `x` ako ključ ne postoji | `d = {"a":1}; d.pop("b", 0)`     | `0`, rječnik ne mijenja |
| `popitem()`        | Briše i vraća **posljednji umetnuti par**      | `d = {"a":1,"b":2}; d.popitem()` | npr. `("b",2)`          |
| `clear()`          | Briše sve elemente rječnika                    | `d = {"a":1}; d.clear()`         | `{}`                    |

> **Napomena**
>
> * `del d[k]` je osnovna operacija, ali baca grešku ako ključ ne postoji.
> * `pop()` je sigurniji jer vraća vrijednost i može imati zadanu rezervnu vrijednost.
> * `popitem()` se često koristi u strukturama gdje je redoslijed umetanja bitan (LIFO ponašanje u novijim verzijama Pythona).
> * `clear()` prazni rječnik, ali ne briše varijablu.

✏️ **Zadatak:**

1. Definiraj rječnik `auto = {"marka": "VW", "model": "Golf", "godina": 2010}`.
2. Ukloni ključ `"model"` pomoću `del`.
3. Ukloni ključ `"godina"` pomoću `pop()` i ispiši vraćenu vrijednost.
4. Pokušaj ukloniti nepostojeći ključ `"boja"` pomoću `pop()` s default vrijednošću `0`.
5. Ukloni posljednji preostali element pomoću `popitem()`.
6. Isprazni rječnik pomoću `clear()` i provjeri rezultat.


````python
auto = {"marka": "VW", "model": "Golf", "godina": 2010}
print(auto)

del auto["model"]
print(auto)

print(auto.pop("godina"))
print(auto)

print(auto.pop("boja", 0))
print(auto)

print(auto.popitem())
print(auto)

auto.clear()
print(auto)
````

#### Metoda za kopiranje

Ponekad želimo napraviti **novi rječnik** koji je **kopija postojećeg**, kako bi promjene na novom rječniku ne bi utjecale na original.
Za to se koristi metoda **`copy()`**.

| Metoda     | Opis                             | Primjer                          | Rezultat   |
| ---------- | -------------------------------- | -------------------------------- | ---------- |
| `d.copy()` | Vraća **plitku kopiju** rječnika | `d = {"a":1}; kopija = d.copy()` | `{"a": 1}` |

---

> **Napomena**
>
> * Metoda `copy()` stvara **plitku kopiju (shallow copy)** rječnika.
> * To znači da se **samo vanjska struktura kopira**, dok se unutarnji mutabilni objekti (npr. liste) dijele između originala i kopije.
> * Promjena običnih (ne-ugnježđenih) vrijednosti ne utječe na original.

---

**Primjer plitke kopije:**

```python
original = {"a": 1, "b": [2, 3]}
kopija = original.copy()

kopija["a"] = 100
kopija["b"][0] = 999

print(original)  # {"a": 1, "b": [999, 3]}
print(kopija)    # {"a": 100, "b": [999, 3]}
```

✏️ **Zadatak:**

1. Definiraj rječnik `student = {"ime": "Ana", "godine": 20}`.
2. Napravi kopiju rječnika pomoću metode `copy()`.
3. Promijeni ime u kopiji i provjeri originalni rječnik.
4. Dodaj u original novi ključ `"smjer"` i provjeri kopiju.


````python
student = {"ime": "Ana", "godine": 20}

kopija = student.copy()
kopija["ime"] = "Ivana"
print(student)
print(kopija)

student["smjer"] = "Informatika"
print(student)
print(kopija)
````

#### Skraćeni oblik rječnika (dictionary comprehension)

Python omogućuje **brzo stvaranje rječnika** pomoću **skraćenog oblika**, poznatog kao **dictionary comprehension**.
Ovaj pristup je **kraći i čitljiviji** od klasičnih `for` petlji s ručnim dodavanjem elemenata.



**Sintaksa**

```python
{ključ_izraz: vrijednost_izraz for element in iterator if uvjet}
```

* `ključ_izraz` – izraz koji određuje ključ u rječniku
* `vrijednost_izraz` – izraz koji određuje vrijednost
* `element` – trenutni element iz iterabilnog objekta
* `iterator` – iterabilni objekt (npr. lista, `range()`)
* `if uvjet` – **neobavezni filter** za uključivanje elemenata



**Primjeri**

1. Kvadrati brojeva

```python
kvadrati = {x: x**2 for x in range(6)}
print(kvadrati)
{0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

2. Rječnik parnih brojeva

```python
parni = {x: x for x in range(11) if x % 2 == 0}
print(parni)
```

#### Sortiranje rječnika

Rječnici **nisu inherentno sortirani**, ali se mogu sortirati pomoću funkcije `sorted()`. Važno je razumjeti da `sorted()` uvijek vraća **listu**, a ne rječnik, pa se po potrebi rezultat dodatno konvertira.

**Sortiranje po ključevima (osnovno ponašanje)**

```python
d = {"b": 2, "a": 1, "c": 3}
print(sorted(d))
['a', 'b', 'c']
```

**Sortiranje parova (ključ–vrijednost)**

```python
d = {"b": 2, "a": 1, "c": 3}
print(sorted(d.items()))
[('a', 1), ('b', 2), ('c', 3)]
```

**Sortiranje po vrijednostima**

```python
d = {"a": 3, "b": 1, "c": 2}
print(sorted(d.items(), key=lambda x: x[1]))
[('b', 1), ('c', 2), ('a', 3)]
```

**Pretvaranje u rječnik nakon sortiranja**

```python
d = {"a": 3, "b": 1, "c": 2}
print(dict(sorted(d.items(), key=lambda x: x[1])))
{'b': 1, 'c': 2, 'a': 3}
```
> **Napomena**:
> 
> Od Python 3.7+ redoslijed umetanja u rječniku se čuva, pa se dobiva “sortirani rječnik”.


**Obrnuto sortiranje po vrijednostima**

```python
d = {"a": 3, "b": 1, "c": 2}
print(sorted(d.items(), key=lambda x: x[1], reverse=True))
[('a', 3), ('c', 2), ('b', 1)]
```

> `reverse=True` mijenja smjer sortiranja u opadajući.


✏️ **Zadatak:**

1. Definiraj rječnik `ocjene = {"Ana": 5, "Marko": 3, "Ivana": 4}`.
2. Sortiraj rječnik po imenima (ključevima).
3. Sortiraj rječnik po ocjenama (vrijednostima).
4. Kreiraj novi rječnik sortiran po ocjenama.
5. Sortiraj rječnik `ocjene` po ocjenama u opadajućem redoslijedu.


````python
ocjene = {"Ana": 5, "Marko": 3, "Ivana": 4}
print(sorted(ocjene))
print(sorted(ocjene, key=lambda x: ocjene[x]))
novi = {k: v for k, v in sorted(ocjene.items(), key=lambda item: item[1])}
print(novi)
print(sorted(ocjene, key=lambda x: ocjene[x], reverse=True))
````
