## 2.8. Rječnik: `dict`

Rječnik `dict` je neuređena kolekcija parova **ključ–vrijednost**. Ključevi moraju biti jedinstveni i nepromjenjivog tipa, a vrijednosti mogu biti bilo kojeg tipa.

---

#### Definiranje rječnika

Rječnik se definiše pomoću `{}` sa parovima `ključ: vrijednost`.

```python
osoba = {"ime": "Ana", "godine": 25}
```

✏️ **Zadatak:**

1. Definiraj rječnik `knjiga` sa ključevima `"naslov"`, `"autor"` i `"broj_stranica"`.
2. Ispiši rječnik.

````python
knjiga = {"naslov": "Zločin i kazna", 
          "autor": "Fjodor Dostojevski",
          "broj_stranica": 430}

print(knjiga)
````


---

#### Pristupanje vrijednostima pomoću ključa

Vrijednost se dohvaća pomoću `[]` sa ključem, ili `get()` metodom.

```python
student = {"ime": "Ana",
           "prezime": "Anić",
           "godine": 20}

student["ime"]            # Ana
student.get("godine")     # 25
student.get("adresa", "Nema")  # Nema (zadana vrijednost)
```

✏️ **Zadatak:**

1. Definiraj rječnik `osoba = {"ime": "Marko", "godine": 25}`.
2. Ispiši vrijednost `"ime"` pomoću `[]`.
3. Dohvati nepostojeći ključ `"grad"` pomoću `get()` sa zadanom vrijednosti `"Nema"`.

````python
osoba = {"ime": "Marko", "godine": 25}
print(osoba["ime"])
print(osoba.get("grad", "Nema"))
````

#### Postavljanje vrijednosti

Vrijednosti se postavljaju ili mijenjaju pomoću `[ključ] = vrijednost`. Nove vrijednosti se dodaju, postojeće se zamijenjuju.

✏️ **Zadatak:**

1. Definiraj rječnik `auto = {"marka": "Toyota", "godina": 2020}`.
2. Promijeni vrijednost `"godina"` na `2021`.
3. Dodaj novi ključ `"boja"` sa vrijednosti `"crvena"`.
4. Ispiši rječnik.

````python
auto = {"marka": "Toyota", "godina": 2020}
auto["godina"] = 2021
auto["boja"] = "crvena"
print(auto)
````

#### Iteracija kroz rječnik

Rječnik se može prolaziti kroz ključeve, vrijednosti ili parove.

```python
for kljuc in osoba:              # kroz ključeve
    print(kljuc)
for vrijednost in osoba.values(): # kroz vrijednosti  
    print(vrijednost)
for k, v in osoba.items():       # kroz parove
    print(k, v)
```

✏️ **Zadatak:**

1. Definiraj rječnik `osoba = {"ime": "Ana", "godine": 25, "grad": "Split"}`.
2. Napravi `for` petlju koja prolazi kroz parove `items()` i ispiši u formatu `kljuc: vrijednost`.

````python
osoba = {"ime": "Ana", "godine": 25, "grad": "Split"}
for kljuc, vrijednost in osoba.items():
    print(f"{kljuc}: {vrijednost}")
````

#### Operatori usporedbe

Samo `==` i `!=` su definirani za usporedbu rječnika (neovisno o redoslijedu unosa).

```python
d1 = {"ime": "Ana", "godine": 25}
d2 = {"godine": 25, "ime": "Ana"}
print(d1 == d2)  # True
```

Operatori `<`, `>`, `<=` i `>=` se mogu koristiti na `keys()` i `items()`, gdje se uspoređuju kao skupovi.

```python
d1 = {"a": 1, "b": 2}
d2 = {"a": 1, "b": 2, "c": 3}
print(d1.keys() < d2.keys())   # True jer {"a", "b"} < {"a", "b", "c"}
print(d1.items() <= d2.items()) # True jer {("a", 1), ("b", 2)} <= {("a", 1), ("b", 2), ("c": 3)}
```

✏️ **Zadatak:**

1. Definiraj `d1 = {"x": 10, "y": 20}` i `d2 = {"x": 10, "y": 20, "z": 30}`.
2. Ispiši `d1 == d2`, `d1 != d2`, te `d1.keys() <= d2.keys()` i `d1.items() <= d2.items()`.

````python
d1 = {"x": 10, "y": 20}
d2 = {"x": 10, "y": 20, "z": 30}
print(d1 == d2)
print(d1 != d2)
print(d1.keys() <= d2.keys())
print(d2.items() <= d1.items())
````

#### Operator pripadnosti

Operator `in` provjerava postoje li **ključevi** (ne vrijednosti).

```python
osoba = {"ime": "Ana",
         "godine": 25,
         "grad": "Split"}
"ime" in osoba           # True
"Ana" in osoba.values()  # True
```

> **Napomena**
>
> * Operator `in` kod rječnika provjerava **ključeve (keys)**, ne vrijednosti.
> * Ako želimo provjeriti vrijednosti, moramo koristiti `values()`.
> * Ako želimo provjeriti parove, koristimo `items()`.

✏️ **Zadatak:**

1. Definiraj rječnik `student = {"ime": "Ana", "godine": 20, "smjer": "Informatika"}`.
2. Provjeri postoji li ključ `"ime"` i ne postoji li ključ `"prezime"`.
3. Provjeri postoji li vrijednost `"Ana"` u vrijednostima rječnika.

````python
student = {"ime": "Ana", "godine": 20, "smjer": "Informatika"}
print("ime" in student)
print("prezime" in student)
print("prezime" not in student)
if "godine" in student:
    print(f"Podatak o godinama postoji.")
print("Ana" in student.values())
````

#### Ugrađene funkcije za rječnike

| Funkcija | Opis | Primjer |
|---|---|---|
| `len(d)` | Broj elemenata | `len({"a":1,"b":2}) → 2` |
| `str(d)` | Rječnik kao string | `str({"a":1})` |
| `list(d)` | Lista ključeva | `list({"a":1,"b":2}) → ['a','b']` |
| `list(d.keys())` | Ključevi | `['a','b']` |
| `list(d.values())` | Vrijednosti | `[1, 2]` |
| `list(d.items())` | Parovi | `[('a', 1),('b', 2)]` |
| `sorted(d)` | Sortirani ključevi | `['a', 'b']` |

> **Napomena**
> * mogu se koristiti `min()`, `max()`, `sum()`, `any()`, `all()`, `set()`, ... 
> * parametar mogu biti ključevi, vrijednosti i parovi, ako je moguće.
> * `dict()` se može koristiti za:
> 
>   * konverziju iz parova `(ključ, vrijednost)`
>   * stvaranje praznog rječnika `dict()`
>   * kopiranje postojećeg rječnika `dict(d)`
> 

✏️ **Zadatak:**

1. Definiraj rječnik `podaci = {"a": 1, "b": 2, "c": 3}`.
2. Ispiši broj elemenata sa `len()`.
3. Napravi i ispiši listu ključeva.
4. Napravi i ispiši listu parova.
5. Kreiraj novi rječnik iz liste `[("x", 10), ("y", 20)]`.
6. Napravi kopiju od `podaci` i usporedi s originalom.

````python
podaci = {"a": 1, "b": 2, "c": 3}
print(len(podaci))
print(list(podaci))
print(list(podaci.items()))
novi = dict([("x", 10), ("y", 20)])
print(novi)
kopija = dict(podaci)
print(podaci, kopija)
print(podaci == kopija)
````

#### Metode za pristup podacima

| Metoda | Opis | Primjer |
|---|---|---|
| `get(k)` | Vrijednost ili `None` | `d.get("b")` |
| `get(k, x)` | Vrijednost ili zadana | `d.get("b", 0)` |
| `keys()` | Ključevi | `d.keys()` |
| `values()` | Vrijednosti | `d.values()` |
| `items()` | Parovi | `d.items()` |

✏️ **Zadatak:**

1. Definiraj rječnik `osobe = {"Ana": 25, "Marko": 30}`.
2. Ispiši dob osobe `"Ana"` pomoću `[]`.
3. Dohvati nepostojeću osobu `"Petar"` sa `get()`.
3. Dohvati nepostojeću osobu `"Petar"` sa `get()` s zadanom vrijednošću `0`.
4. Ispiši sve ključeve, vrijednosti i parove.

````python
osobe = {"Ana": 25, "Marko": 30, "Ivana": 22}
print(osobe["Ana"])
print(osobe.get("Petar"))
print(osobe.get("Petar", 0))

for k in osobe:
    print(k)

for v in osobe.values():
    print(v)

for kv in osobe.items():
    print(kv)
````

#### Metode za postavljanje podataka

| Metoda | Opis | Primjer |
|---|---|---|
| `update({...})` | Dodaj više parova | `d.update({"b":2})` |
| `setdefault(k,v)` | Postavi ako ne postoji | `d.setdefault("a", 1)` |

✏️ **Zadatak:**

1. Kreiraj prazan rječnik `student = {}`.
2. Dodaj `"ime"` sa vrijednosti `"Luka"` koristeći `[]`.
3. Dodaj `"godine"` s vrijednošću `21` pomoću `update(})`.

#### Metode za brisanje podataka

| Metoda | Opis | Primjer |
|---|---|---|
| `pop(k)` | Obriši i vrati | `d.pop("a")` |
| `pop(k, x)` | Obriši ili vrati zadanu | `d.pop("a", 0)` |
| `popitem()` | Obriši zadnji par | `d.popitem()` |
| `clear()` | Obriši sve | `d.clear()` |

> **Napomena**
>
> * `del d[k]` je osnovna operacija, ali daje grešku ako ključ ne postoji.

✏️ **Zadatak:**

1. Definiraj rječnik `auto = {"marka": "VW", "model": "Golf", "godina": 2010}`.
2. Obriši `"model"` sa `del`.
3. Obriši i ispiši `"godina"` sa `pop()`.
4. Pokušaj obrisati nepostojeći ključ `"boja"` sa `pop()` i zadanom vrijednosti `0`.
5. Ukloni posljednji element s `popitem()`.
6. Isprazni rječnik pomoću `clear()`.


````python
auto = {"marka": "VW", "model": "Golf", "godina": 2010}
print(auto)

del auto["model"]
print(auto)

print(auto.pop("godina"))
print(auto)

print(auto.pop("boja", 0))
print(auto)

print(auto.popitem())
print(auto)

auto.clear()
print(auto)
````

#### Metoda za kopiranje

Metoda `copy()` pravi **plitku kopiju** rječnika.  
Promjenjivi elementi (npr. liste) se dijele između kopije i originala.

```python
original = {"a": 1, "b": 2}
kopija = original.copy()
kopija["a"] = 100
print(original)  # {"a": 1, "b": 2}
print(kopija)    # {"a": 100, "b": 2}
```

✏️ **Zadatak:**

1. Definiraj rječnik `student = {"ime": "Ana", "godine": 20}`.
2. Napravi kopiju sa `copy()`.
3. Promijeni ime u kopiji i provjeri original.
4. Dodaj u original novi ključ `"smjer"` i provjeri kopiju.

````python
student = {"ime": "Ana", "godine": 20}

kopija = student.copy()
kopija["ime"] = "Ivana"
print(student)
print(kopija)

student["smjer"] = "Informatika"
print(student)
print(kopija)
````

#### Skraćeni oblik rječnika (dictionary comprehension)

Brz način kreiranja rječnika: `{ključ: vrijednost for element in iterator if uvjet}`

```python
kvadrati = {x: x**2 for x in range(5)}
parni = {x: x for x in range(10) if x % 2 == 0}
```

✏️ **Zadatak:**

1. Napravi rječnik `kvadrati` gdje su ključevi brojevi od 1 do 10, a vrijednosti njihovi kvadrati.
2. Napravi rječnik  **neparnih** brojeva od 0 do 20 gdje je vrijednost = broj × 2.
3. Napravi rječnik iz liste `['jabuka', 'limun', 'banana', 'kokos']` gdje je ključ=riječ, vrijednost=duljina riječi ali samo za riječi koje sadrže slovo `"a"`.


````python
brojevi = {br: br**3 for br in range(10)}
print(brojevi)

brojevi_parni = {br: br*2 for br in range(10) if br % 2 == 0}
print(brojevi_parni)

lista_voca = ["jabuka", "limun", "banana", "kokos"]
rjecnik_voca = {voce: len(voce) for voce in lista_voca if "a" in voce}
print(rjecnik_voca)
````

#### Sortiranje rječnika

Rječnik se sortira pomoću `sorted()`, koja vraća listu.

```python
d = {"b": 2, "a": 1, "c": 3}
sorted(d)                           # po ključevima: ['a', 'b', 'c']
sorted(d.items(), key=lambda x: x[1])  # po vrijednostima
```

✏️ **Zadatak:**

1. Definiraj rječnik `ocjene = {"Ana": 5, "Marko": 3, "Ivana": 4}`.
2. Sortiraj po imenima (ključevima).
3. Sortiraj po ocjenama (vrijednostima).
4. Napravi novi rječnik od `ocjene` sortiran po ocjenama.
5. Sortiraj rječnik `ocjene` po ocjenama u opadajućem redoslijedu.

````python
ocjene = {"Ana": 5, "Marko": 3, "Ivana": 4}
print(sorted(ocjene))
print(sorted(ocjene, key=lambda x: ocjene[x]))
novi = {k: v for k, v in sorted(ocjene.items(), key=lambda item: item[1])}
print(novi)
print(sorted(ocjene, key=lambda x: ocjene[x], reverse=True))
````
